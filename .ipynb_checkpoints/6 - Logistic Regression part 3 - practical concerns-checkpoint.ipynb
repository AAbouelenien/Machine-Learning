<h1>Math</h1>

now we want to know how to interpret the weights obtained from logistic regression

recall that for linear regression : 

$w_i$ is the amount y will increase if $x_i$ is increased by 1 and all other x's remain constant

lets focus first on the binary scenario :

output of logistic model is either 1 or 0

so a weight will bring the prediction closer to 1 or closer to 0

Intuitively we know a bigger weight $\to$ bigger effect


if $w_i$ is big and positive , then a small increase in $x_i$ (assuming $x_i$ is also positive) will push the output closer to 1

if $w_i$ is big and negative , then a small increase in $x_i$  will push the output closer to 0

 so large magnitude $\to$ large impact
 
 now lets recall the output of the logistic model :
 
 $$p(y=1|x) = \sigma(w^Tx) = \frac{1}{1+\exp(-w^Tx)}$$
 
 $$p(y=0|x) = 1 - \sigma(w^Tx) = 1 - \frac{1}{1+\exp(-w^Tx)} = \frac {\exp(-w^Tx)} {1+\exp(-w^Tx)}$$
 
Now we calculate the 'odds' 

The 'odds' refer to the ratio of two probabilities

odds = P(Event) / P(~ Event)

in our case $p(y=1|x) / p(y=0|x)$

$$\frac {p(y=1|x)} {p(y=0|x)} = \frac {1} {1+\exp(-w^Tx)} \times \frac {1+\exp(-w^Tx)} {\exp(-w^Tx)} = \exp(w^Tx)$$

now lets take the logs of both sides :

$$ \log \left( \frac {p(y=1|x)} {p(y=0|x)} \right) = w^Tx = w_1x_1 + ... + w_Dx_D$$

we took the log of the odds , this quantity is called .... 'log odds' just that 

so basically we are doing linear regression on the log odds

so we interpret the weight $w_i$ as the amoint the log odds increase if $x_i$ increases by 1

<h1>Math</h1>

Now we want to talk about regularisation

Recall the LDA example from 5 where we had 2 gaussian clouds one centered at (2,2) and the other at (-2,-2)

the solution was , w = [0,4,4] m plugging this into the equation of a line $0 + 4x + 4y = 0$ we arrive at the line y = -x

but why (4,4) why not (1,1) or (10,10) , when all arrive at the same line y = -x

we need to reconsider the objective function 

$$J = t \log(y) + (1-t) \log (1-y)$$

take $(x_1,x_2) = (1,1)$ and weights (0,4,4)

should be classified as 1

$\sigma(0+4+4) = \sigma(9) = 0.99966$

well great , but Exactly 1 would be better !!!

with $y = \sigma{8} , J = -0.0003354... $

for weights (0,1,1) $\to$ J = -0.1269

for weights J = (0,10,10) , $J = -2.0611^{-9}$

so the best weights are thus (0,infinity,infinity) , but in the computer thats an error

so the logistic regression model tries to go to (0,infinity,infinity) , to solve this problem we use regularisation which penalises very large weights

so we add a penalty to big weights in cost function , so our cost grows larger if any of our weights get bigger :

$$J = -[t \log(y) + (1-t) \log(1-y)]$$
$$J_{reg} = J + (\lambda /2) {\Vert w \Vert}^2 = J + (\lambda /2) w^Tw $$

now we need  to do gradient descent to optimise w

this time we want $dJ_{reg} / dw$ instead of $dJ / dw$

since addition doesnot affect other terms in derivative lets look at the penalty term

reg_cost $ = (\lambda/2)(w_0^2 + w_1^2 + w_2^2 + ...)$

d(reg_cost)/d$w_i = \lambda_i$

In vector form d(reg_cost)/dw = $ \lambda w$

$$\frac {dJ_{reg}} {dw} = X^T(Y-T) + \lambda w$$

Now lets look at this from a probabilistic perspective :

we know that cross-entropy maximises the likelihood , since J = -log(likelihood)

lets make $J_reg$ something we want to maximise and see what we are maximising , we will multiply by -1 and whatever we happen to maximise is maximised when we minimise $J_reg$ , so let J = - J_{reg} (not to confuse this with the J for cross-entropy)

$$J = + \left[ t \log y + (1-t) \log(1-y) \right] - (\lambda /2) {\Vert w \Vert}^2$$

This follows that :

$$exp(J) = y^t (1-y)^{(1-t)} \exp(-\lambda {\Vert w \Vert}^2 /2)$$
 
The first part is a Bernoulli (likelihood)

The second term is a Gaussian (prior)

t ~ Bernoulli(y)

w ~ N(0,1/$\lambda$) , $\lambda$ is also the called the precision which is the inverse variance

recall bayes rule 

$$P(B|A) = \frac {P(A|B)P(B)} {P(A)}$$

we call :

P(B|A) the posterior

P(A|B) the likelihood

P(B) the prior

P(A) : the bottom part is just the top part marganalised over B , since $\sum_B P(A,B) = \sum_B P(A|B)P(B) = P(A)$ , this is just a normalising constants so we obtain probabilities that sum up to one

we know that

posterior ~ likelihood x prior



without regularisation we are maximising the lokelihood

with regularisation we maximise the posterior

also called maximum a posteriori , MAP











<h1>code</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
D = 2

X = np.random.randn(N,D)

# the first 50 points belong to class 0 and are centered around -2
X[:50,:] = X[:50,:] - 2*np.ones((50,D))
# the last 50 points belong to class 1 and are centered around 2
X[50:,:] = X[50:,:] + 2*np.ones((50,D))

# here we initialise the targets
T = np.array([0]*50 + [1]*50)

# now we concatenate a column of ones with X
ones = np.atleast_2d(np.ones(N)).T
X_b = np.hstack((ones,X))

# initialise the weights , we will with and without regularisation 
w = np.random.randn(D+1)
w_reg = w.copy()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def cross_entropy(T,Y):
    E = 0
    for i in range(N):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1-Y[i])
    return E
```

```python
z = X_b @ w
Y = sigmoid(z)
# gradient descent
learning_rate = 0.1
costs = [] 
for i in range(100) : # steps
    e = cross_entropy(T,Y)
    costs.append(e)
    w -= learning_rate * (X_b.T @ (Y-T))
    Y = sigmoid(X_b@w)    
```

```python
plt.plot(costs)
plt.show()
```

```python
z = X_b @ w_reg
Y = sigmoid(z)
# gradient descent + L2 regularisation
learning_rate = 0.1
lambda_ = 0.1
costs = []

for i in range(100) : # steps
    e = cross_entropy(T,Y)
    costs.append(e)
    w_reg -= learning_rate * ((X_b.T @ (Y-T)) + lambda_*w_reg)
    Y = sigmoid(X_b@w_reg)
```

```python
plt.plot(costs)
plt.show()
```

```python
# gradient descent weights
print(w)
```

```python
# gradient descent + regularisation weights
print(w_reg)
```

```python
# we can clealy see how regularisation causes weights to be smaller 
```

<h1>Math</h1>

we discussed previously how we prefer a skinny matrix over a fat one , that is a matrix where $D << N$ is preferable than a mtrix $D>N$

Now we will discuss L1 regularisation for logistic regression , as a reminder this should help us acheive sparsity , that is most weights will be 0


So for L2 Regularisation we had :

$$J_{RIDGE} = - \sum^{N}_{n=1}(t_n\log y_n + (1-t_n) \log(1-y_n)) + \lambda {\Vert w \Vert}_2^2$$

for L1 we have :

$$J_{LASSO} = - \sum^{N}_{n=1}(t_n\log y_n + (1-t_n) \log(1-y_n)) + \lambda {\vert w \vert}_1$$

As shown before we , we can take $\exp(J_LASSO)$ , the likelihood term remaing the same , a gaussian likelihood , our prior term becomes:

$$p(w) = \frac{\lambda}{2} \exp(-\lambda \vert w \vert)$$

which follows a laplace distribution , so we know that L1 has a Bernoulli likelihood and a laplace prior

next we take derivative and try to solve for w , since we tried this before we know that we fail and we need to use gradient descent

$$J = - \sum^{N}_{n=1}(t_n\log y_n + (1-t_n) \log(1-y_n)) + \lambda {\vert w \vert}_1$$

$$\frac{\partial J}{\partial w} = X^T(Y-T) + \lambda sign(w)$$

<h1>code</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
N = 50
D = 50 

# x is uniformally dstributed between -5 , +5
X = (np.random.random((N,D)) - 0.5)*10

# actually only the first 3 dimensions matter
true_w = np.array([1,0.5,-0.5] + [0]*(D-3))

# generate y , add noise with variance 0.5
Y = sigmoid((X@true_w + np.random.randn(N)*0.5))
```

```python
# now lets plot the data
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2],c=Y)
plt.show()
```

```python
# now we perform gradient descent to opptimise w
# note that since we created Y by dotting X and w we already know that there is no bias term so we dont bother adding it
costs = [] # keep track of mean error
w = np.random.randn(D) / np.sqrt(D) # randomly initialise w
learning_rate = 0.001
l1 = 3.0 

for i in range(5000):
    Yhat = sigmoid(X@w)
    delta = Yhat - Y
    w = w - learning_rate*(X.T@delta + l1*np.sign(w))
    cost = -(Y*np.log(Yhat) + (1-Y)*np.log(1-Yhat)).mean() + l1*np.abs(w).mean()
    costs.append(cost)
    
```

```python
# plot the costs
plt.plot(costs)
plt.show()
```

```python
print("final w: ",w)
```

```python
# plot true_w vs predicted_w
plt.plot(true_w,label = 'true weights')
plt.plot(w,label = 'predicted weights')
plt.legend()
plt.show()
```

```python
# now we discuss several problems that appear to be non-solvable by logisitc regression 
# but by tweaking the datset a bit , we can solve the problem
```

```python
# The Donut Problem
N = 1000
D = 2
R_inner = 5
R_outer = 10

# in a polar system , this is the distance from origin
# we sample from a uniform distribution , set mean to 0
R1 = np.random.randn((N//2)) + R_inner
# and this is the angle
# 2pi * (number from 0-1) so we get thetas between 0-2*pi
theta = 2*np.pi*np.random.random(N//2)
# in a cartesian system x component is R cos theta , y component is R sin theta
X_inner = np.concatenate([[R1*np.cos(theta)],[R1*np.sin(theta)]]).T

R2 = np.random.randn((N//2)) + R_outer
theta = 2*np.pi*np.random.random(N//2)
X_outer = np.concatenate([[R2*np.cos(theta)],[R2*np.sin(theta)]]).T

X = np.concatenate([X_inner , X_outer])
T = np.array([0]*(N//2) + [1]*(N//2))
```

```python
# lets plot the Donuts !!!
plt.scatter(X[:,0],X[:,1],c=T)
plt.show()
```

```python
# at first sight this cant be solved by logisitc regression
# since data is not linearly sperable
# we can use the fact that each donut has a different radius
# if we add the radius as another dimension , instead of having point at (i,j) we have points at (i,j,k) 
# now points that belong to smaller donut will have k values close to R1 (5) , same for big donut but with R2
# Imagine having one donut in space at height 5 and another at height 10 , surely you can put a sheet of paper between
# in the same sense logistic regression will now find a plane that seperates the donuts in 3d space
```

```python
ones = np.ones((N,1))
# distance of each point
r = np.zeros((N,1))

# if this was some data we wont have R
# but what we have for each point is y co-ordinate : r sin theta and x-cordinate : r cos theta  
# so for a point x , x@x = r^2 sin theta + r^2 cos theta = r^2 (sin theta + cos theta) = r^2
# so sqrt(x@x) = r , the distance/radius of that point from origin

for i in range(N):
    r[i] = np.sqrt(X[i,:]@X[i,:])

Xb = np.concatenate((ones,r,X),axis=1)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def cross_entropy(T,Y):
    E = 0
    # since always one of the terms is 0 , only compute the non-zero term   
    # just a more effecient version 
    for i in range(N):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1-Y[i])
    return E
```

```python
w = np.random.randn(D+2)
z = Xb @ w
```

```python
# now lets do gradient descent

learning_rate = 0.0001
errors = []

for i in range(5000):
    Y = sigmoid(Xb @ w)
    delta = Y - T
    error = cross_entropy(T,Y)
    errors.append(error)
    
    if i %100 == 0:
        print(error)
        
    w -= learning_rate * ((delta @ Xb) - 0.01*w)
    
```

```python
plt.plot(errors)
plt.show()
```

```python
print('final w: ',w)
Y = sigmoid(Xb @ w)
print('Final classification rate:',np.mean(T==np.round(Y)))
```

```python
# its interesting how the weights for x and y co-ordinates are almost 0
# this means that the classification has nothing to do with the coordinates of the point
# expected since we now that the classification depends mainly on the height of the point (its radius)
```

```python
# now here is another one
# its called the XOR problem
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 2000
X = np.random.random((N, 2))*2 - 1

# generate labels
T = np.zeros(N)
T[(X[:,0] < 0) & (X[:,1] > 0)] = 1
T[(X[:,0] > 0) & (X[:,1] < 0)] = 1

# plot it
plt.scatter(X[:,0], X[:,1], c=T)
plt.show()
```

```python
# as easy as it seems no line can seperate both classes
# we follow the previous approach of adding extra dimensions
ones = np.ones((N,1))
# the dimension we add is simpy xy
xy = np.atleast_2d(X[:,0]*X[:,1]).T
Xb = np.array(np.concatenate((ones,xy,X),axis=1))
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def cross_entropy(T,Y):
    E = 0
    # since always one of the terms is 0 , only compute the non-zero term   
    # just a more effecient version 
    for i in range(N):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1-Y[i])
    return E
```

```python
w = np.random.randn(D+2)
z = Xb @ w
```

```python
# now lets do gradient descent

learning_rate = 0.0001
errors = []

for i in range(5000):
    Y = sigmoid(Xb @ w)
    delta = Y - T
    error = cross_entropy(T,Y)
    errors.append(error)
    
    if i %100 == 0:
        print(error)
        
    w -= learning_rate * ((delta @ Xb) - 0.01*w)
```

```python
plt.plot(errors)
plt.show()
```

```python
print('final w: ',w)
Y = sigmoid(Xb @ w)
print('Final classification rate:',np.mean(T==np.round(Y)))
```
