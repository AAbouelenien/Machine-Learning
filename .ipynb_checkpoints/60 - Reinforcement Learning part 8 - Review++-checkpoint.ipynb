<h3>Episode</h3>

Next, let's continue defining more terms

So far, we know about the agent and the environment

The next term we want to define is episode

What happens when we play a game of tic-tac-toe or breakout?

Well, some sequence of events will occur, and then at the end, we will win or lose 

With our math exam example, we will take our math exam and then we'll get your grade

Now, we know that with learning algorithms, the way that they learn is with data

So if we're training a dog versus cat classifier, we'll need lots of labelled images of dogs and cats

Similarly with tic-tac-toe or breakout, once the game is over, we can opt to play again

This is the method through which we will gain experience or to be more technical data

We might call these games or rounds or matches, but in reinforcement learning, the official term is episode

So when we're training an agent to play tic-tac-toe, we're going to play multiple episodes, and at the end of each episode, our agent will have won or lost.

Hopefully by the end of the training process or in other words, after many episodes, our agent will be winning more than losing

---

Of course, not all reinforcement learning environments are episodic

To say an environment is episodic means that they end at some point, and you can start again with a new, fresh episode

Furthermore, there is no relationship between one episode and the next

So the fact that we lost the previous tic-tac-toe episode will have no effect on the environment in the next episode

However, there are examples of non episodic environments

Take, for example, the stock market

For all intents and purposes, this can go on forever

There's no real notion of the end

Well, if you lose all our money, then technically there is nothing more you can do

But it's not the same as losing a game of tic-tac-toe and starting again

We can't go back in time and restart the stock market

Another example is an online advertising system

Our agent's job will be to choose the right advertisements, to show to users at any given moment in order to maximize our company's revenue

It should do this continuously

There is no concept of an end to an online advertising service

All right, so those are some examples of some non episodic environments

We can refer to such environments as having infinite horizons

note : We can have an episodic task with an infinite horizon, but we wont bother to distinguish tasks with this level of granularity

<h3>States, Actions, Rewards</h3>

<h3>Example: Maze</h3>

Here's another example of states actions and rewards

Think of a maze 

In this maze, the state is our position in the maze 

Our actions may consist of the various directions we can go

For example, up, down, left or right

The reward is tricky

Remember, we said it's up to us to think of a good reward to assign to our agent to encourage it to learn how to solve the environment

We might say $+1$ for solving the maze and $0$ otherwise

But we ask oursleves, is this a good strategy?

Imagine we throw your agent into this maze and it has to learn what to do

Imagine your agent has played this game 10000 times and has never solved the maze

We can pretend the environment is episodic so that after taking 100 steps, we reach a terminal state and the game is over

What happens if we get zero reward each time?

Well, in that case, the agent learns that it does not matter at all what it does because doing anything always leads to the same reward, aero

In this case, the agent has no incentive to solve the maze

Our agent will never prioritize one action over another because it knows that no matter what it does, it always gets zero reward 

In this case all actions are equal

---

Perhaps a better reward structure would be to assign a $-1$ reward at every state

In this case, we can maximize our reward by solving the maze as fast as possible

Doing any extraneous actions will lead to a more negative reward

Not solving the maze will lead to the most negative reward

So in this case, assigning a negative reward upon reaching any state will allow our agent to solve
the environment

---

<h3>This is not an English class</h3>

Now, we have to keep in mind this is not English class, so we have to remove any bias we may have about what connotations are associated with the term reward

We might think of reward as a good thing, like a prize

For example, if we have a dog and it just successfully completed a trick, we  may give it a treat as a reward

But in reinforcement learning, this is not what we mean by reward

The only constraint is that the reward is a real number

It can be positive, negative or zero

We will also receive this number at every step in the environment, not just when youwereach some goal or definitively fail to achieve it

The agent, as we will learn later in the notebook, will try to maximize its reward over each episode

For example, we may get a reward of $-100$

This is better than a reward of $-1,000,000$

Maybe $-100$ reward corresponds to successfully solving the environment

But in the end, it's just a number

Don't associate $-100$ with negative connotation and $+100$ with positive connotation

So just remember this, the reward is not a prize, the reward is a number which is to be maximized

We can think of it as kind of the opposite of a loss function, whereas we want to minimize the loss in a supervised or unsupervised learning problem, In reinforcement learning we want to maximize reward

<h3>Policies</h3>

Now that we know how to represent states and actions in code, it's time to talk about policies

instructor's comment: When I first started learning about reinforcement learning, I found policies to be kind of an odd concept, well, actually, I found all of reinforcement learning to be kind of odd and foreign, but you'll get used to it

The idea of a policy makes sense at a high level, but it becomes somewhat ambiguous when we start thinking about how to represent it mathematically or in code

The policy is what the agent uses to determine what action to perform given a state

It's important to keep in mind that the policy yield's an action using only the current state

It doesn't use any combination of the current state and previous states, and it doesn't use any information about rewards

Technically, as we mentioned earlier, the state may be made up of multiple observations and that may also include the reward, although that is unconventional

But strictly speaking, the policy will yield an action using only the current state

---

<h3>Policy as a Function or Mapping</h3>

The simplest way to think about a policy is that it's a dictionary mapping or a function that returns an action given a state 

Here is such a function

```python
def get_action(s):
    a = policy[s]
    return a
```

As we can see, the only input is a state $s$ and it returns an action $a$ where the state $s$ was the key to the dictionary, and the action $a$ was the value

The real question is how can we represent this mathematically?

---

<h3>An example policy</h3>

This is why it's useful to talk about how we encode states and actions first

So imagine again we are in GridWorld

<img src='extras/60.7.PNG' width='200'>

note : the above image also shows our policy

Our agent now has a dictionary representing what action to perform given the state as we can see here (so thats how we may write our policy encode)

```python
actions = [up,down,left,right]
policy = {
    (0,0) : right,
    (0,2) : right,
    (1,0) : up,
    (1,2) : up,
    (2,0) : up,
    (2,1) : right,
    (2,2) : up,
    (2,3) : left
}
```


So, for example, if we're to the left of the goal state, then the appropriate action is to move right

And just to be clear, the left of the goal state is $(0,2)$

If we're at the initial stage, then the appropriate action is to move up, and just to be clear that is the state (2,0)

Now, from the initial state, moving right is just as valid since we can still reach the goal from that point

We can see that although we've encoded the states here explicitly as tuples, we could gain more efficiency by encoding them as integers corresponding to the tuples and using those integers to index an array

As we may recall from our computer science studies, indexing arrays is faster than indexing dictionaries

<h3>Policy as a Function or Mapping</h3>

Thinking about policies as dictionary mappings is somewhat limited

There are two reasons for this

First is that this won't work if we have an infinite state space

We would need an infinite sized dictionary, which is not possible to have

Second is that it doesn't allow our agent to explore its environment

Think of training our agent like teaching a baby 

At first a baby knows nothing

It must try new things in order to figure out how the world works and build up its intuition

A reinforcement learning agent is the same way

If it has a fixed policy and it only does the same thing all the time, then it can't gain new experiences

Thus, it makes sense for policies to be stochastic

Stochastic is just a fancy word for random

In other words, a more general way to represent policies is to represent them as probabilities

---

<h3>Policy as Probability</h3>

Representing policies as probabilities actually solves both of the problems we posed above

Let's see how

First, it deals with this problem of randomness

The common way of dealing with this in reinforcement learning to allow the agent to explore, is to give it a small chance of performing a random action

So here is a python function that can accomplish this

```python
def get_action(s):
    if random < epsilom:
        a = sample from action space
    else:
        a = fixed_policy[s]
    return a
```

We first generate a random number 

If this number is less than some small number epsilon, let's say 0.1, then it will choose an action at random from the action space

Otherwise, we will grab an action from our fixed policy dictionary mapping

This method is called Epsilon Greedy, and we'll learn later in this notebook why it's useful and what the relevance of exploration is

---

So what about continuous state spaces?

In fact, seeing policies as probabilistic easily lends itself to continuous or infinite state spaces

Imagine out state is a vector $s$

Now imagine we have some policy parameters $W$ 

The shape of $W$ is the dimensionality of the state space by the size of the action space , $D \times \vert A \vert$

For now, we'll presume that the action space is still categorical

So what do we do when we want to output probabilities from a set of categories?

In fact, this is just like classification, we can use the softmax function

So now our policy is 

$$\large \pi(a\vert s) = \text{softmax}(W^Ts)$$

As we can see, this allows us to introduce a little more notation

It's common in reinforcement learning to denote the policy with the symbol $\pi$, not to be confused with the number $\pi$ 

For any given state, we can calculate a probability distribution over the actions space $\pi(a\vert s)$ 

Then to decide which action to perform in the environment, we can simply sample from this distribution

This allows us to explore if necessary, but we can still treat this policy deterministically if we want it simply by using the $\arg \max$

Also note that it's not necessary to use a linear model, as we are doing here

We can't, in fact, use any function approximater, such as a neural network

<h3>Is only the state needed?</h3>

Now, at this point, we may be wondering, how can an intelligent agent possibly know what to do using only the current state?

This is the problem we described before

Just by looking at a still image of the road, how can we know what the correct action is?

If we're more inclined to think inside a supervised learning paradigm, we may insist that there should be a target here so that the agent learns to associate this date with this action

<h3>The Markov Assumption</h3>

The main assumption we make in reinforcement learning is the Markov assumption

This is something we often discuss in terms of Markov models and sequence modeling, Bbt let's review it here anyway

The Markov assumption goes like this

Suppose we want to predict whether tomorrow will be rainy, sunny or cloudy

Perhaps our idea may be to base this on whether it was raining, sunny or cloudy in the past seven days

Well, the Markov assumption is that tomorrow's weather doesn't depend on all of the past seven days, only the immediate previous day

<h3>The Markov Assumption</h3>

Here's another example of the Markov assumption

Suppose we want to predict the next word of a sentence

We are told that the previous word in the sentence is lazy

The Markov assumption is that the next word only depends on the previous word, therefore, if the Markov assumption is true, then we should be able to predict the next word in our sentence

---

Of course, things are not so easy

We might think, because we are taking a course by the lazy programmer that the next word in the sentence is programmer

But in fact, that's not what the instructor had in mind :)

The Markov assumption has actually been quite successful so far

In general, the Markov assumption states that the probability of the state at time $t$ depends only on the state at time $t-1$ and not on any state that came before that

$$\large p(s_t \vert s_{t-1},s_{t-2},\cdots,s_0) = p(s_t \vert s_{t-1})$$

Now, by itself, the Markov assumption is weak

The words are merely observations, but the state is made up of a sequence of observations 

In this way the Markov assumption is not as bad as we might initially think

<h3>Markov Decision Process (MDP)</h3>

So why do we need to know about the Markov assumption?

This is because reinforcement learning problems are commonly described as a Markov decision process or MDP

Previously, we discussed the Markov assumption in terms of a state only

But as we know, reinforcement learning problems involve other objects as well, namely actions and rewards

So the way we describe an MDP is to use the state transition probability

It's the probability of arriving in the state at time $t+1$ and getting the reward at time $t+1$, given the state of time $t$ and taking the action at time $t$

$$\large p(S_{t+1},R_{t+1} \vert S_t,A_t)$$

Another simpler way of writing this without time indices is just to write 

$$\large p(s^\prime,r \vert s,a)$$

note that because the reward $r$ has no prime symbol, the prime symbol does not indicate time $t+1$

We get the reward at time $t+1$ for arriving in state $s^\prime$, but we do not put a $'$ symbol on $r$

---

<h3>State-Transition</h3>

So we just saw the most general way of writing down the state transition probability, but often we can make it less general

For example, if we are solving a maze, then most likely we are going to make the reward deterministic

In other words, there's no need to represent it as a probability distribution

In this case, we can use the notation $p(s^\prime \vert s,a)$

And the reward can be a symbol all by itself, usually denoted as $R(s,a,s')$

This encodes the idea that we were in state $s$, we performed action $a$ and we arrived in the next state $s'$

We can even just say $R(s)$ or $R(s')$ in the case where the reward depends only on the state where we arrive at, which is actually quite common

$\large \text{State transition probability} : p(s' \vert s,a)$

$\large \text{Reward function} : R(s,a,s') \text{ or simply } R(s), R(s')$

---

An important point to consider is, what is the usefulness of the state transition probability?

We can imagine that if we are playing some game like breakout on Atari, it is very unlikely we will ever be able to calculate these probabilities, given that the state space would be in a feasible to enumerate

We want to think of the MDP and the state transition probability as stepping-stones

They are simply conceptual tools which we will use to further advance our knowledge and take us to a point where we can actually come up with a practical algorithm for reinforcement learning

In other words, while we're not going to be using state transition probabilities directly in Q-learning, they do help us build on what we've done so far so that we can actually arrive at Q-learning in a logical manner

<h3>The Language of Probability</h3>

<h3>Maximising Rewards</h3>

So far, we said that the goal of the agent is to maximize the reward it gets

But as we recall, the reward may be structured differently in different games

For example, in tic-tac-toe, we may receive a $+1$ for winning or a $-1$ for losing

On the other hand, if we're solving a maze, we might get a $-1$ at every step

So what does it mean to maximize the reward?

Does it mean to maximize the reward at the next step?

Or does it mean to maximize the total reward over an entire episode?

---

<h3>Future Rewards</h3>

Here's the answer 

To state it more accurately, the goal of the agent is to maximize the sum of future rewards

Why is that?

Well, it can't maximize the rewards it got already, those are in the past, they cannot be changed

Furthermore, we don't want to only maximize the reward on the next step

What if we are solving a maze and we get a $-1$ for any step we take?

In that case, the agent isn't incentivized to do anything useful because no matter what it does, the immediate reward is still $-1$

Thus, the agent's true goal is to maximize the sum of future rewards until the episode is over

In this way, the agent is planning its future steps as well

It must have some concept of where it will end up because that's the only way it will know how to maximize rewards in the future

---

<h3>Preparing for a math exam</h3>

To take a real world example, consider again the idea of preparing for a math exam 

For a math exam, we don't receive any reward until we've completed the exam

The reward signal is our grade on the exam

But imagine all the actions it will take to actually maximize that reward

We'll have to study, we'll have to do homework, we have to forego socializing with your friends (not necessarily)

In fact, all of those actions do not sound very rewarding at all

And thus, if our only incentive is immediate gratification, in other words, the rewards we'll receive immediately, then you will not do well on our math exam

Instead, we must make use of long term planning

Sure, we may not want to study today, it may be very annoying and we will miss our favorite TV show, but because we are planning long term, we're not thinking only about today, we're thinking about the results of our math exam

The desire to maximize the total future reward is necessary for long term planning

---

<h3>The Return</h3>

We call the sum of future rewards the return 

We describe the return mathematically using the symbol $G$

Because it depends on future rewards only, it is time dependent, so we indexed it with a $t$

We can say the return at time $t$ is the sum of rewards at time $t+1$ up to the terminal state at time $T$

$$\large G(t) = \sum^{T-t}_{\tau=1} R(t+\tau) = R(t+1) + R(t+2) + \ldots + R(T)$$

---

<h3>Infinite Horizon MDP</h3>

Now, we might be wondering what happens if we have an infinite horizon MDP, a game that never ends

In this case, our return might be infinity

Therefore, we introduce a concept known as discounting

Discounting is used for infinitely long tasks, but it's also used for episodic tasks as well

We introduce a discount factor called $\gamma$ 

Each future reward is weighted by Gamma to some power

$$\large G(t) = \sum^{T-t}_{\tau=1} \gamma^{\tau-1}R(t+\tau) = R(t+1) + \gamma R(t+2) + \ldots + \gamma^{T-t-1}R(T)$$

$\gamma$ is usually a number close to $1$ like $0.9,0.99,0.999,\ldots$

It's a hyper parameter, so we'll have to choose its value based on the performance of our agent

The idea is the further we go into the future, the harder it is to predict

Therefore, we care a little more about getting rewards now than we do later

Intuitively, this works just like money

We would rather receive one hundred dollars today than receive one hundred dollars ten years from now

In 10 years, one hundred dollars will be worth much less than it is today due to interest

---

<h3>Recursiveness</h3>

One important feature of the return which we will make use of throughout the rest of this notebook, is that it can be defined recursively

$$\large G(t) = R(t+1) + \gamma G(t+1)$$

In other words, in terms of itself

This may not seem like much more than a simple math substitution now, but we'll see how it will become very useful later on

<h3>Expressing the problem as an equation</h3>

Let's begin by discussing expected values

By the way, if we've never taken a probability course, then we've probably never heard of expected values, in which case if we want to understand this, we want to take a course in probability 

To think about it, simply, the expected value is just the mean

So, for example, suppose we've measured the heights of 1000 students and we want to model this as a Gaussian distribution

We find that the average height is 70 inches with a standard deviation of four inches

In this case, the mean of the distribution is 70 and thus the expected value is 70

---

<h3>More about expected values</h3>

Let's look at a coin as the next example

A coin has a binary outcome

If we flip a coin, we may get one or we may get zero

Suppose the probability of both one and zero is 50%

In this case, what's the expected value?

Well, since it's the mean, that means the expected value is $0.5$

Sometimes this terminology confuses people

They ask, how can the value we expect to get be $0.5$ if a result can only be zero or one?

It's important to realize the expected value, despite its name, is not the value we expect to get

So if we flip a coin, we don't actually expect to see the value $0.5$

---

<h3>More about expected value</h3>

The expected value actually has a precise meaning for both continuous and discrete distributions

Although for most of this notebook we will assume that we are working with discrete distributions

The idea is that the expected value is a weighted sum of all the possible values of a random variable where the weights are the probabilities of those values

$$\large \text{Discrete : } E(X) = \sum^{+ \infty}_{k = - \infty} p(x=k)k$$

$$\large \text{Continuous : } E(X) = \int\limits^{+ \infty}_{k=-\infty} x p(x) dx$$

---

<h1>Coin example</h1>

As a simple example, we can return to our coin example 

If we have a biased coin, so that the probability of heads is $0.6$, then the expected value becomes $E(X) = 0.6 \times 1 + 0.4 \times 0 = 0.6$ 

note : heads = 1, tails = 0

Because the probability of one is now a little higher than the probability of zero, the expected value also becomes a little higher

---

<h3>Why expected values?</h3>

All right, so what's the significance of expected values?

Well, let's remember that our rewards, and hence also our return, which is the sum of rewards, are random variables 

At a high level, what this means is that if we play a game 100 times using the same policy in the same environment, we will get different rewards

This is because both the environment dynamics and our policy are probabilistic, and thus it makes sense to not think about a single return by itself, but the expected value of the return 

We would like to maximize the sum of future rewards, but because the results is probabilistic, what we really want to do is maximize the expected sum of future rewards

---

<h3>Value function</h3>

The expected sum of future awards has a special name and reinforcement learning

We call it the value function

$$\large V(s) = E(G_t \vert S_t = s) = E(R_{t+1} + \gamma R_{t+2} + \ldots \vert S_t = s)$$

We always mention that this is a very unfortunate name, because the word value is such a generic word

Nonetheless, this is what it is called

Because the return is the sum of future awards after we have arrived in some state, so too is the value function

As we can see, it's not just a plain expected value, but rather it's conditioned on the fact that we arrived in the state $s$ at time $t$

Thus we denote this as $V(s)$ the value of the state $s$

As a side note, we sometimes refer to the return as simply the reward, even though the return should be the sum of rewards

So we always have to pay attention to the context

---

<h3>Recursiveness</h3>

As we recall, we said that one of the most important things about the return is that it can be defined recursively 

$$\large G(t) = R(t+1) + \gamma G(t+1)$$

note : recall $G(t)$ is the return at time $t$

And because of this, we can also define the value function recursively 

$$\large V(s) = E(R_{t+1} + \gamma V(s') \vert S_t = s)$$

---

<h3>Why do this?</h3>

Now, we must be wondering, what is the purpose of all this?

Are we just defining equations for the sake of it?

Of course, the answer is no

Just keep in mind the high level picture

We are setting up a problem from which we can find a solution 

In order to find a solution, we must have a problem

And thus, our job right now is to continue building up the problem so that it is well-defined

The next step is to remember that the expected value is really just the weighted sum where the weights are the probabilities

So if we write it out in full, this is what we get

$$\large V(s) = \sum_a \sum_{s'} \sum_{r} \pi(a \vert s) p(s',r \vert s,a) \{r + \gamma V(s')\}$$

This is the sum over all possible actions $a$ and the sum over all possible next states $s'$ and the sum over all possible rewards $r$ 

Inside the sum, we have the probability of performing action $a$ given that we are in state $s$ multiplied by the probability of landing in the next state $s'$ and receiving the reward $r$, and then this is multiplied by the reward $r + \gamma V(s')$

In other words, it's $r+ \gamma V(s')$ multiplied by the probability of actually getting the reward $r$ and arriving at the state $s'$

In other words, the expected value of the return

This is called the Bellman equation

It is the centerpiece of all the subsequent solutions to reinforcement learning that we will discuss

---

<h3>What are its implications?</h3>

It's worth thinking about the Bellman equation a little more 

Notice that in order to come up with this equation, we only use the rules of mathematics, the rules of probability

It seems like just a regular old expected value

However, the implications are deep

It's interesting to consider that the two probabilities here, $\pi(a \vert s) and p(s',r \vert s,a)$, come from totally different physical processes

$\pi$ is our policy and it represents our agent, so we can think of that as like an animal

The state transition probability represents our environment, so you can think of that as the world

So it's interesting that while we can freely manipulate this equation mathematically, these two objects are actually completely distinct physically

---

<h3>Is there an end in sight?</h3>

At this point, we must be extremely tired :|

All we are doing is making our formulas progressively more complicated with no end in sight

But in fact, this is exactly where we want to be 

Although this may seem complicated, we will demonstrate that, in fact, what we have arrived at is pretty simple

---

<h3>Problem and Solution</h3>

We can now define the first problem in reinforcement learning from which we can find a solution

Remember that in a reinforcement learning problem, we can have multiple policies

Some may be good, but some may be bad

How can we tell what is a good policy and what is a bad policy?

Well, how about the value function?

In other words, if we can find the value function for a given policy, that tells us how good that policy is

We call $V_\pi(s)$, the value function given the policy $\pi$

We call the problem of finding the value function given a policy, the prediction problem

---

<h3>How is this simple?</h3>

As promised, we said that we've arrived at something pretty simple

Let's look at the Bellman equation again

$$\large V(s) = \sum_a \sum_{s'} \sum_{r} \pi(a \vert s) p(s',r \vert s,a) \{r + \gamma V(s')\}$$

These probabilities are a little deceptive

They have complicated symbols, but they are not complicated

Remember that the policy $\pi$ is just a function which we implement in our own code

We know this probability

Therefore, it's not a problem

How about the state transition probability?

Let's assume this is known as well

This is entirely possible

Take, for example, the grid world environment

Now, what happens if we know these two probabilities?

---

<h3>Systems pf linear equations</h3>

Surprisingly, this just becomes a system of linear equations, which we know how to solve from high school algebra

Suppose we have three states 

Then, if we simplify the summation from the Bellman equation, we would arrive at something like this

$$\large c_{11} V(s_1) + c_{12} V(s_2) + c_{13}V(s_3) = b_1$$

$$\large c_{21} V(s_1) + c_{22} V(s_2) + c_{23}V(s_3) = b_2$$

$$\large c_{31} V(s_1) + c_{32} V(s_2) + c_{33}V(s_3) = b_3$$

So here the $b$s and the $c$s are just constants which are derived from the probabilities which we have assumed that we know

From this, we can call a simple function like `np.linalg.solve`, and this would tell us $V(s_1),V(s_2)$ and $V(s_3)$

In other words, if we know both our policy and the environment dynamics, we can solve for the value function using only linear algebra

<h3>What does it mean to 'learn'?</h3>

In reinforcement learning, there are two main types of tasks

Previously, we discussed what we call the prediction problem 

That is given a policy $\pi$, find the associated value function $V(s)$

The second type of task is called the control problem

This means to find the optimal policy $\pi$, which leads to the maximum $V(s)$

In other words, the control problem is maximising the sum of future rewards

We've already known since the beginning of the notebook that this is our goal

Now we can accurately define what this means and thus find a solution

---

<h3>Action-Value</h3>

First, we need a little bit more math and a few more symbols to accurately describe what we are doing

We already know about the value function given a state $V(s)$

To be more precise, we refer to this as the state value function

There is another quantity known as the action value function, where the value depends both on the state $s$ and the action $a$

We denote with the symbol $Q$

$$\large Q(s,a) = E(r+ \gamma V(s') \vert S_t = s,A_t = a)$$

As we can see, it has almost the same definition of $V(s)$, except that it's also conditioned on the action

And so in the bellman equation, since $a$ is given, we do not sum over the policy distribution

$$\large Q(s,a) = \sum_{s'}\sum_r p(s',r \vert s,a) \{r+ \gamma V(s')\}$$

---

<h3>Space Complexity</h3>

One interesting question to consider from a computer science perspective is how much space is required to store the value functions?

Let's consider the scenario where both states and actions are discrete and that we have a finite number of them

So let's say we have $S$ and $A$ actions

In this case, we can store the state value function in an array of size $S$

But for the action value function, we'll have to store it in a two dimensional array of size $S \times A$

And thus the storage required for $Q$ is quadratic, whereas the storage required for $V$ is only linear

---

<h3>The oprimal policy</h3>

OK, so why do we need this concept of the action value?

Well, this is going to help us find the optimal policy

Remember, some policies may be good and some policies may be bad

The optimal policy is the best policy, the one that maximizes the value

First, let's just think about how to compare two different policies

We can say that policy one is better than policy two, if $V(s)$ given $\pi_1$ is greater than $V(s)$ given $\pi_s$ for all the states as in the state space

$$\large \pi_1 \ge \pi_2 \ \ \ \text{ if } V_{\pi_1}(s) \ge V_{\pi_2}(s) \ \ \ \forall_s \in S$$

So this helps us describe the relative goodness of different policies

---

<h3>The best policy and the best value function</h3>

From here, we can define the best policy and the best value function 

The best value function is the value function for which there is no greater value function

It's the max over all possible policies of $V(s)$ given $\pi$

So we'll denote it with the symbol $V^*$

$$\large V^*(s) = \max_\pi V_\pi(s), \ \ \forall_s \in S$$

Similarly, the best policy will be the $\arg \max$ overall policies $\pi$

$$\large \pi^* = \arg \max_\pi V_\pi(s), \ \ \forall_s \in S$$

So we call that $\pi^*$

$$\pi^* = \arg \max_\pi V_\pi(s), \ \ \forall_s \in S$$

---

<h3>Relationship Between Q and V</h3>

So far, we haven't needed to invoke the action value function, but let's see how it's related

First, the optimal action value is defined similarly to the optimal state value, we'll call it $Q^*$

$$\large Q^*(s,a) = \max_\pi Q_\pi(s,a), \ \ \forall_s \in S,a \in A$$

This must hold over all states and all actions

Furthermore, the optimal state value is equal to the max over all actions from the optimal action value

$$\large V^*(s) = \max_a Q^*(s,a)$$

So from this, we can define the relationship between $Q$ and $V$

---

<h3>What about the action-value?</h3>

The real value, no pun intended, of the action value function is this

Suppose we are playing some game and we would like to know what's the best action to perform right now?

Well, we have a dictionary telling us exactly what to do

All we have to do is 

$$\large a^* = \arg \max_a Q^*(s,a)$$

In other words, the best action to perform becomes a simple dictionary look-Up

---

<h3>The optimal policy is not Unique</h3>

As a side note, notice how the optimal policy is not unique

It could be that multiple different policies lead to the same best value function

In this case, it suffices to find just one of them

On the other hand, the optimal value function is unique because if there are two different value
functions, then logically one of them must be greater than the other one

---

<h3>Solving the Bellman Equation</h3>

Previously, we looked at some simple and naive solutions to reinforcement learning

Let's recap what they were

First, we recall that there are two types of problems, the prediction problem and the control problem

The prediction problem means, given a policy, find $V(s)$

The control problem means find the best policy, which yields the maximum $V(s)$

<h3>Are these approaches realistic?</h3>

Let's consider now why both of these approaches are somewhat unrealistic

First, let's consider the prediction problem

Previously, the solution required us to know both the policy distribution and the environment dynamics

But the reality is, if we imagine, say, any Atari game, we won't know the environment dynamics

All we can really do, practically speaking, is play the game thousands of times

But usually the state space is so large that it would be infeasible to measure the state transition probabilities

And so unless we're playing with a toy environment, we would not be told these probabilities either

---

<h3>Are these probabilities realistic?</h3>

Now, let's consider the control problem

Is it really possible to enumerate all possible policies?

Consider if we have $S$ possible states and $A$ possible actions 

In this case, the total number of possible policies is $A^S$

In other words, this grows exponentially and hence it is not feasible to enumerate all possible policies for most practical problems

Furthermore, this would not even work in the case where the states space or action space is infinitely large

---

<h3>The Solution</h3>

All right, so now that we've realized there is a problem, what is the solution?

The trick to this is to remember the relationship between the expected value and the mean

The problem with the expected value is that it requires us to know the probability distribution of the random variable in question

But importantly, there is a way for us to estimate the mean or equivalently the expected value

This is called the sample mean

This is the basis for many scientific experiments

For example, if we want to test a drug, we don't know the true values, but we can do an experiment on a set number of people and calculate the average values

The sample mean is simply the sum of all the samples we collect divided by the number of samples

$$\large E(X) \approx \frac{1}{N} \sum^N_{i=1} x_i$$

The idea is that as $N$ approaches infinity, this estimate will become more and more accurate

---

<h3>Application to RL</h3>

OK, so what does this have to do with reinforcement learning?

Well, remember that the value function is simply the expected return


Therefore, using the sampling approach, what we can do is sample a set of returns for each state in the state space and take the average 

That will give us an estimate of the value of each state

$$V(s) = E(G_t \vert S_t = s) \aaprox \frac{1}{N} \sim^N_{i=1} g_{i,s}$$

We'll notice that we've abused notation a little bit here by using different indices for the return $G$

When we say $G_t$, we mean a generic return $G$, the return for the state at time $t$

But when we index $g$ using $i,s$, what we mean is this is a sample of the return, it's the $i$th sample return from the state $s$

---

<h3>Wgere do samples come from?</h3>

Now, one obvious, but maybe not so obvious point, where do these samples come from?

We may recall that when we're in numpy and we want a sample from, say, the standard normal, we can simply call the function `np.random.randn()`

But what does it mean to sample a return?

Well, remember that when we play an episode, even if we use the exact same policy and the exact same environment, the result will be different

This is because both the policy and the environment dynamics are probabilistic

So just playing the game itself results in collecting samples

---

<h3>Pseudocode</h3>

Using this concept, let's now attempt to write some pseudocode to describe this algorithm

By the way, this approach is called the Monte Carlo approach, since this is a form of Monte Carlo sampling

First, we're going to describe the prediction problem or in other words, given a policy, find the value function

Let's just think about this at a high level first

Suppose we play one episode of a game

This consists of entering a series of states and corresponding rewards so we can call them $s_1$ up to $s_T$ and $r_1$ up to $r_T$ 

From this, how do we calculate the return of each state?

---

<h3>Calculating Returns</h3>

Well, it's helpful to actually go backwards, for example, $g_t$ is zero 

Importantly, note that the return is only the sum of future rewards

Since a terminal state is the end of an episode, that means there are no future states, no future
states means no future rewards

Therefore, the return of a terminal state is always zero and hence the value of a terminal state is also always zero

In any case, let's continue

How do we calculate the return of the previous timestep?

Well, by definition, $g_{T-1} = r_T$

OK, the next one, how about $g_{T-2}$?

$g_(T-2) = r_{T-1} + \gamma r_T$

But importantly, remember that the return is recursive

So this is also equal $g_(T-2) = r_{T-1} + \gamma r_T = r_{T-1} + \gamma g_{T-1}$

We can repeat this pattern

So $g_{T-3} = r_{T-2} + \gamma g_{T-2}$

And of course this follows for any value of $t$, so we can say $g_t = r_{t+1} + \gamma g_{t+1}$

So we can keep repeating the same pattern until we calculate the return for each state

---

<h3>Calculating Returns</h3>

Practically speaking, here's how we would do it in code 

```python
states, rewards = play_episode_using_policy

returns = []
g = 0
returns.append(g)
for r in reversed(rewards):
    g = r + gamma * g
    returns.append(g)
```

First, we would play an episode using our policy and we get back a series of states and rewards

Next, we initialize a list of returns to an empty list and we initialize the return to zero 

Then and this is the important part, we loop through the rewards in reverse 

Inside the loop we just use the recursive definition of $g$, `g = r + gamma *g`

Then we append $g$ to our list of returns

---

<h3>Pseudocode for Prediction</h3>

All right, so how do we put the above algorithm into pseudocode?

Well, that's simple

```python
policy = ... # given as input
sample_returns = {} # state -> list of returns
for i in range(num_episodes):
    states, rewards = play_one_episode(policy)
    returns = ... # calculate as previously discussed
    for s, r in zip(states,returns):
        sample_returns[s].append(r)
```

Everything we just did, now we just do it in a loop, do the thing we just did a several hundred or several thousand times then for each state, take the average return

First we assume we are given some policy

Then we instantiate an empty dictionary to store our sample returns

The key of this dictionary will be the state and the value will be a list of returns encountered throughout each episode

Then we enter a loop which goes on for a predetermined number of episodes

Inside the Loop, we first play one episode using the given policy

This returns us a list of states and corresponding rewards

From that, we can calculate the return for each state as previously described

Next, we loop through each state and return in corresponding order 

For each return we encounter, we append that to the list of returns for that state

This loop is how we collect our samples

---

```python
# calculate the average return
V = {}
for s, g_list in sample_returns.items():
    V[s] = np.mean(g_list)
```

Finally, when the first loop is complete, we are done collecting our samples

Now we can calculate the average return, which is by definition the value function

So we loop through each of the items in our sample returns dictionary 

For each iteration, we get the state `s` and a list of sample returns `g_list` 

Inside the loop `V(s)` is simply assigned the sample mean of `g_list`

---

<h3>Details</h3>

It's important to realize that there are some complications with the algorithm as described 

First, because we're only sampling, how do we ensure that we actually encounter every possible state in the number of episodes we played?

In fact, we cannot 

Although we may surmise that because we didn't encounter a particular state, we don't need to know its value because our policy does not allow us to go there

We could simply ignore the value of those states in any subsequent function we plug the value function into

But there is a better solution 

To understand it, we're going to move on to the second problem, the problem of control or finding the best policy

<h1>Math</h1>

<h3>Prediction vs. Control</h3>

As a general pattern, note that for the prediction problem, usually we are interested in finding $V(s)$ which is what we just did 

For the control problem, we are usually interested in working with the action value $Q(s,a)$

The reason for this is, if we recall, the $Q$ function makes it easy to look up what action to take given a state $s$

We always take the $\arg \max$ over all possible actions

$$\large a^* = \arg \max_a Q^*(s,a)$$

We call this the greedy action

In this way we maximize what we believe to be the sum of future rewards

So just keep this general pattern in mind 

For the prediction problem, we work with $V(s)$ 

For the control problem, we work with $Q(s,a)$

---

<h3>Policy Iteration and Policy Improvement</h3>

In order to understand how to apply the Monte Carlo approach to the control problem, we first have to understand the principle of policy iteration and policy improvement

Let's consider two basic facts

Number one, given a policy, we can use Monte Carlo to evaluate the value function, whether that be the state value or the action value

We just saw that in the previous section

Number two, given an action value function, we can always choose based on this what we believe to be the best action given the current state, this is just the $\arg \max$ over all possible actions $a$

Well, it turns out that these two facts are interdependent

Given a policy, we can find its corresponding action value, and from that, we can take the $\arg \max$ to find a possibly better policy

But what if this policy is different from the original policy? 

Then we can find the value function for this new policy

And from that, we can take the $\arg \max$ again to find yet another new and possibly better, but at least as good policy

From there, we can find the value function again

So we see, this is just a loop where we go back and forth finding the value function and given a policy and improving the policy, given that value function

<img src='extras/60.10.PNG' width='200'>

We can see that we have named these two steps, the act of finding the value function is called the evaluation step and the act of finding the best policy given that value function is called the improvement step

And just to be clear, the reason why this is a loop is because they both change each other

So by doing step one, we change the policy, and by doing step two, we change the value

---

It's been proven, although we won't discuss it here, that performing these steps leads to an monotonic improvement in the policy

Therefore, if we just keep doing this process, eventually we will end up at the optimal policy

---

<h3>Monte Carlo Approach</h3>

So how can we apply this to Monte Carlo?

Here's a rough outline

```python
Q = random, policy = random
```

This is called the policy improvement step

<h3>Improving our solution</h3>

At this point, our solution works, but it's not an ideal solution

Let's think about why

First, if we recall, $V(s)$ only stores $S$ values, but $Q(s,a)$ stores $S \times A$ values

As we know, with Monte Carlo sampling, the more samples we collect, the more accurate our answer becomes

When we use $Q(s,a)$, we have many more values to estimate and therefore we have to collect many more samples in order to get an accurate estimate

<h3>Problem #2: Nested Loops</h3>

But there is a second problem

Each policy evaluation step is a Monte Carlo estimation, which means that we must calculate a large number of samples


What's the effect of this?

Well, let's say our policy iteration loop needs to run a 1000 times to find the optimal policy

Now, let's say our inner loop needs to run 1000 times in order to get an accurate estimate of $Q(s,a)$

How many episodes do we have to play?

The answer is 1000 $\times$ 1000, which is one million

As we can see, our data usage is not efficient, and the number of times we need to play the game grows quite fast

<h3>Value Iteration</h3>

A better solution is to use what's called value iteration

At first, this approach may seem kind of wacky, but in fact it's been shown to work


The idea is this for the policy evaluation step, instead of playing multiple episodes in order to get a good estimate of the value of our policy, we are only going to play one episode 

After playing this one episode, we'll get a series of states actions and rewards from which we can calculate the corresponding returns

From that, we can loop through each state, action and return and use the latest sample of the return to update $Q(s,a)$

Now we'll notice that we've just put some pseudocode here, `update Q(s,a)`, but we haven't mention exactly how we're going to do that

We'll be discussing this shortly

The important thing to note right now is that we'll only keep a single running copy of $Q(s,a)$

The next and final step is the same as before

We update the policy by taking the $\arg \max$ over $Q$ for a given state overall actions

```python

for i in range(num_episodes):

    states, actions, rewards = play_one_episode(policy)
    returns = ... # calculate as previously discussed
    
    for s, a, g in zip(states, actions, returns):
        Q(s,a) = update Q(s,a) with latest return g
    for s in Q.states(): # policy improvement step
        policy[s] = argmax{ Q(s, :) }
```


So what's going on with this line here `Q(s,a) = update Q(s,a) with latest return g`

How do we update you of $Q(s,a)$, given the old estimate of $Q(s,a)$

For this, we need to take a short digression back to sample means and expected values yet again

As much as we're tired of repeating by now, this is the expression for the sample mean 

We sum up all the samples and we divide by the total number of samples

The question we have to ask is, is this an efficient calculation?

---

<h3>Efficient Sample Means</h3>

The answer is no

Summing up and values is $O(n)$

The more values we store, the more time it takes to calculate

This is not good

So the question becomes, is there a way to reduce the time it takes to calculate the sample mean every time we collect a new sample 

To see how this is done, let's express the sample mean in terms of the previous sample mean and the latest sample

$$\large \bar X_N = \frac{1}{N} \sum^N_{i=1}x_i = \frac{1}{N} \left(\sum^{N-1}_{i=1} x_i +x_N\right) = \frac{1}{N} \left(\frac{N-1}{N-1} \sum^{N-1}_{i=1} x_i + x_N\right)$$

$$\large \bar X_N = \frac{1}{N} \left((N-1) \bar X_{N-1} + x_N\right)$$

$$\large \bar X_N = \frac{N-1}{N} \bar X_{N-1} + \frac{1}{N}x_N$$

note : $\bar X_N$ means the sample mean using the first $N$ samples, hence $\bar X_{N-1}$ means the sample mean using the first $N-1$ samples

---

<h3>Rearrange</h3>

The trick is we can turn the above expression into something that looks like, and in fact is gradient descent

We can write it so it looks a little more familiar

$$\large \bar X_N = \bar X_{N-1} + \frac{1}{N} \left(x_N - \bar X_{n-1}\right)$$

In this scenario, the target is the latest sample we've collected, while the old estimate is our prediction, 1/N is the learning rate, so this learning rate is one that decays over time, by using such a learning rate, we get back exactly the sample mean

note : $\frac{1}{N}$ : learning rate, $x_N$ : target, $\bar X_{N-1}$ : prediction

Remember, all we've done so far is basic algebra

---

<h3>Rewrite</h3>

Here's what it would look like if we translated it into our Monte Carlo reinforcement learning code to update $Q$

$$\large Q(s,a) = Q(s,a) + \frac{1}{N} (g - Q(s,a))$$

Just note that the equals sign here means assignment, we're not saying that both sides are equal

The left side is the new estimate and the right side holds the old estimate

We just haven't bothered to subscript them

However, we are not done yet because we are going to make yet another modification to this

---

<h3>Wait!</h3>

Remember that in our latest value iteration loop, we are updating the same $Q$ dictionary using different policies

```python
Q = random, policy = random
for i in range(num_episodes):
    states, actions, returns = ... # get sample
    update Q(s,a) and policy
```

This policy is being updated on each step

And therefore, the samples which we are using to calculate $Q(s,a)$ are not coming from the same
distribution

In this case, we don't want to use exactly the sample mean 

Intuitively, the oldest samples come from the oldest policies

They don't really matter that much

The newest samples come from the newest policies, and they matter more

---

<h3>Constant learning rate</h3>

The key is to use a constant learning rate instead of one that decays over time

$$\large Q(s,a) = Q(s,a) + \eta(g-Q(s,a))$$

Doing so gives us what is called the exponentially decaying average

The idea is, when we take the conventional sample, mean each sample is weighted equally

As we said, we don't want that because old values come from old policies, and so they come from a different distribution than the one we are now interested in

On the other hand, the exponentially decaying average weights each sample at an exponentially decaying fashion

The most recent sample matters most, and the weights for each sample decay exponentially as we go backwards in the order they were collected

---

<h3>Pseudocode</h3>

So now we can express our value iteration code again, but this time we can fill in this little missing detail about how to update $Q(s,a)$

```python
Q = random, policy = random
for i in range(num_episodes):
    # replace policy evaluation with one episode only
    states, actions, rewards = play_one_episode(policy)
    returns = ... # calculate as previously discussed
    
    for s, a, g in zip(states, actions, returns):
        Q(s,a) = Q(s,a) + learning_rate * (g-Q(s,a))
        
    for s in Q.states(): # policy improvement step
        policy[s] = argmax{ Q(s, :) }
```

So this is the exact same pseudocode as before, except we've replaced the middle block of code with the actual update we can use the update $Q$

Unfortunately, at this point, there is a subtle but important detail we have yet to discuss.

So we are not quite done specifying our algorithm, but we've laid out most of the groundwork

<h3>Policy as a Distribution</h3>

Earlier in the notebook, we talked about the value of treating the policy as a probability distribution

First, it allows us to describe the entire MDP as a set of two probabilities, one representing the environment dynamics and one representing the agent

This allows us to reason about the MDP mathematically and find a solution to both the prediction and control problems, as we just did

The other important reason, which we mentioned briefly earlier, is that in order to find the best action, we must first know the results of performing different actions

In other words, we must have collected those samples

<h3>Problem!</h3>

Here is the problem 

When we take the $\arg \max$ over all actions, $Q(s,a)$, the $Q$ value here is not changing that much from one episode to the next

Let's say we have three actions

So we want to choose from, $Q(s,a_1),Q(s,a_2),Q(s,a_3)$

Well perhaps we initialise $Q(s,a_1),Q(s,a_2)= 0 $ while we initialised $Q(s,a_3)=1$

Also assume that all the rewards are positive

During the process of playing the game, maybe $Q(s,a_3)$ gets updated to a value of two

We will now always choose action three because that gives us the best $Q$-value out of the three possible actions

The problem is we don't know the true values of $Q(s,a_1),Q(s,a_2)$ because we never collected data from performing those actions

How do we know the values of the other actions if we have never actually performed those actions?

In fact, we don't know those values at all

We just happened to initialize them to zero 

Because of that there is no way for us to confidently choose the right action

<h3>The Explore-Exploit Dilemma</h3>

This is a more general problem known as the explore exploit dilemma

Suppose we go to a casino and we're playing a simplified version of slot machines

We have several slot machines to choose from, and you don't know which one we should play

However, we do know that not all these slot machines yield the same reward

As we said, these slot machines are simplified, so there are only two possible outcomes, we can either win or you lose

If we win, you get one point (no real money, gambling is HARAM :| )

And if we lose, we get zero points

Obviously, we would like to play the slot machine in which our chances of winning are highest

---

OK, so how can you figure out which slot machine will give us the best chance of winning?

Well, we cant ask the manager of the casino because he wants to make points :),  he will not share that information with us

So what do we do?

Our only method of figuring out which slot machine is best is to collect some data

So let's say we play each slot machine 1000 times

The win rate of the slot machine is simply equal to the number of times we won divided by the total number of times we played

---

But what's the problem with what we just did?

Let's say we have to choose between five slot machines

If we played a slot machine 1000 times, that means we've played 5000 games

Now, lets remind ourselves that playing a slot machine at a casino is not free (costs points)

In general, collecting data is not free

It usually requires time, resources or both

Let's say each time we play a slot machine, you spend 0.25 points

We need to win at least one out of every four times in order to break even

Imagine that for some of these slot machines, we win zero times

Now it becomes clear that playing those slot machines 1000 times would be a huge waste of both time and money

---

Here is another question to consider

Why should we collect 1000 samples per slot machine?

Why not 100?

Why not 10?

As we know, the more samples we collect, the better and more accurate our estimate becomes

Unfortunately, this also equates to more time and money spent

---

That's why we call this a dilemma

They explore-exploit dilemma means we are trying to balance both exploration and exploitation

We want to explore so that we can collect more data to find out which slot machine is truly the best

On the other hand, we want to exploit because we want to play what we believe is the best slot machine in order to win more often and to make more money

These two forces always oppose each other

<h3>Epsilon-Greedy</h3>

The typical answer to this problem in reinforcement learning is called epsilon greedy

This allows us to be greedy and exploit what we think will yield the highest future sum of rewards

But we have a small probability, epsilon, of choosing a random value so that we can explore and collect samples for our $Q$-table to make our estimates of $Q$ more accurate

And remember, they do need to be somewhat accurate in order for us to confidently choose the best action

---

<h3>Epsilon-Greedy in Code</h3>

Here's how he will do this in code

```python

```

We will generate a random number between zero and one 

If this random number is less than epsilon, we will choose an action at random

Otherwise, we will perform the action specified by the greedy policy, which is greedy with respect to $Q$ given the state $s$

```python
if random() < epsilon:
    a = action_space.sample()
else:
    a = argmax(Q[s,:])
```

<h3>Q-Learning</h3>

Let's recap what we've done so far because it's been quite a long process to even get to Q-learning

First, we defined all the relevant terms, such as agent, environment, state, action, reward and so
forth

Having these definitions allowed us to structure our problem mathematically

Specifically, we can model reinforcement learning problems as MDPs, Markov Decision Processes

Next, we considered how we would solve an MDPs, meaning two things

First, we solve the problem of prediction, finding the value function and given a policy

Second, we solve the problem of control, finding the best policy in a given environment

We learned that if we know all the probabilities in an MDP, this is quite easy

But when we don't know the probabilities, we can use sampling methods which we call Monte Carlo

---

<h3>Monte Carlo Limitations</h3>

There is one major problem with Monte Carlo 

It is that in order to calculate returns, we must wait until an episode is over

This is because the return is defined as the sum of all future rewards

We can't know the sum of future rewards until we have collected them, and therefore we must reach a terminal state before calculating the sample returns

Why is this a problem?

Well, imagine the scenario where we have very long episodes or the scenario where there is no terminal state

In the latter case, Monte Carlo is not an option 

In the former, it's still not ideal because that means our agent has to spend a very long time performing suboptimally, even though it has already collected a lot of data from which it may improve

---

<h3>Temporal Difference Methods</h3>

The answer to this is temporal different methods

If we recall, we said earlier that one of the most important features of the return is that it can be defined recursively 

The return at time $t$ can be expressed in terms of the return at time $t+1$

$$\large G(t) = R(t+1) + \gamma G(t+1)$$

We saw how this helped us both create and solve the bellmna equation

Now it's going to help us once again

One simple way to think about it is this

Monte Carlo methods were an approximation to an expected value problem

Temporal difference methods are simply an approximation to Monte Carlo

So in other words, they are an approximation of an approximation

---

<h3>Monte Carlo Update Trick</h3>

Recall this one neat trick with Monte Carlo methods 

When we're updating $Q$ or $V$, which is the sample mean of all the samples we've collected, we don't have to express the update as a sample mean

$$\large Q(s,a) = Q(s,a) + \eta(g - Q(s,a))$$

That would mean we have to keep all the samples around, which can take up lots of memory and take lots of time to calculate

Instead, we can express the current sample mean in terms of the past sample mean

We noted that this looks a lot like gradient descent

---

<h3>The Gradient Descent Perspective</h3>

Well, why not test out this theory

Let's set our squared error $J$ to be the squared error between our true target $G$ and our value for the state $s$, $V(s)$

$$\large J = (g - V(s))^2$$

---

<h3>The Gradient</h3>

Now, let's say we want to update $V(s)$ using the latest sample $g$, which we've just collected 

In order to perform the update, we're going to use gradient descent 

$$\large V(s) \leftarrow V(s) - \eta \nabla_{V(s)}J$$

Well, what is the gradient?

$$\large \nabla_{V(s)}J = -2(g-V(s))$$

If we plug this gradient into our gradient descent update, and ignore the $2$, since it can be absorbed into the learning rate, we get back the exact opposite equation we would use for the exponentially decaying average

$$\large V(s) \leftarrow + \eta(g-V(s))$$

---

<h3>Gradient Ascent vs. Gradient Descent</h3>

As a side note, we want to mention that there is ultimately no difference whether we call what we are doing gradient ascent and or gradient descent

$$\large V(s) \leftarrow V(s) + \eta(g - V(s))$$ 

$$\large V(s) \leftarrow  V(s) - \eta(V(s)-g)$$

Because there is a $+$ sign here, we might think of it as gradient descent

The $+$ sign is natural to use if we derive this expression from the sample mean update

But if we derive this expression from the squared error, you will get a $-$ sign

And we might think of that as gradient descent

Ultimately, however, this is just basic algebra and we should confirm to ourselves that both expressions are equivalent

---

<h3>Lets Combine These Ideas</h3>

So what's so significant about this?

Well, now we're going to put together these two ideas

Idea number one is that updating the value function, using the exponentially decaying average is just like gradient descent 

Idea number two is that the return can be defined recursively

---

<h3>Estimate the return</h3>

So how about this?

We're going to continue using gradient descent, but instead of using the full return, we are going to estimate the return

We collect the next reward $r$, but instead of waiting to collect any future awards, we simply guess that they will be close to $V(s')$, the value of the state where we landed

So instead of using $g = r + \gamma r' + \gamma^2 r'' + \ldots$, we recognize that $g = r+ \gamma g'$, where $g'$ is the next return

But $g'$ has the expected value $V(s')$

So we instead just say,$g = r+ \gamma g' \approx r + \gamma V(s')$ 

In this way, we only have to wait one step before updating our model

We no longer have to wait until the end of the episode

We call $r + \gamma V(s')$ a bootstrapped estimate of the return

It allows us to update $V(s)$ immediately after obtaining the next reward $r$, instead of having to wait until we've collected all the future rewards

This method is called the temporal difference method

---

<h3>Pseudocode (Prediction Problem)</h3>

Here's some pseudocode so we can conceptualize how this will work 

```python
# given: env, policy
V = random
for i in range(num_episodes):
    s = env.reset()
    done = false
    while not done:
        a = policy[s]
        s_prime, r, done = env.step(a)
        
        # the big update
        V(s) = V(s) + learning_rate * (r + gamma * V(s_prime) - V(s))
        
        # important: update the current state
        s = s_prime
    # by now, V(s) has converged (can also check ourselves)
        
```

As a side note, this should also give us some insight into what goes on inside our play episode function, which we haven't really discussed yet

In this pseudocode, we don't have any need to abstract away the play episode function because playing the episode is part of this loop

We have things to do on each step of the episode, and therefore we cannot encapsulate it or delegate it to some other function

To start assuming we are given some environment and some policy

We initialize the value function to be random

Then we enter a loop, which plays a predetermined number of episodes

Alternatively, we can also run the loop until we find that $V(s)$ converges, or, in other words, settles on some value and doesn't deviate from it

Inside the loop, we begin playing an episode

The first thing we do is call `env.reset()`, which resets the environment and puts us back into the initial state and returns that initial state, we'll call it $s$

Next, we initialize a `done` boolean flag to `false`

This boolean flag will get set to `true` when we've completed an episode

Next, we enter a while loop that completes when `done` becomes true 

Inside the loop we grab the action from our policy

We then perform the action in the environment by calling `env.step()`

This returns three things

The next state `s_prime`, the reward `r` and the next `done` flag

Note that we've designed this pseudocode to be similar to the OpenAI Gym API, which has become somewhat standard over the past few years

It's easy to understand and it will help us in the future if we ever do start using OpenAI Gym

The reverse is also true

If we've had experience with OpenAI Gym in the past, this should make things easier to understand

Next, we do the big update, the temporal difference update we've been discussing throughout this section 

Finally, and this is important not to forget, we must update the variable `s` for the next iteration of the loop

What is currently the next state `s_prime` will
become the current state `s` in the next iteration

Lastly, when the loop is complete, `V(s)` has converged

---

<h3>Semi-Gradient</h3>

There is one odd thing about temporal difference learning

Look carefully at this so-called gradient descent update

```python
# the big update
V(s) = V(s) + learning_rate * (r + gamma * V(s_prime) - V(s))
```

The target is `r + gamma * V(s_prime)`

The prediction is `V(s)`

If we relate this back to supervised learning, we notice something strange 

In supervised learning, we are given the target as part of the set

But here we are doing something surprising

We are predicting the target itself

Part of the target is given, that's the reward $r$

But the other part, `V(s_prime)` is actually a model prediction

Thus, it's more correct to say that what we are doing is not quite gradient descent

It's called a semi gradient instead

But this is just a name, it is the principle that's important

The principle is we don't know the true target, we are estimating it

And this makes it very different from supervised learning

---

<h3>Q-learning</h3>

All right, so what we looked at so far, was the prediction problem

Now it's finally time for the big reveal

We are going to solve the control problem using the famous Q-learning algorithm

At this point, we spent so long building up the prerequisites for Q-learning that we should hardly be surprised that what we see

Nonetheless, let's have a look

---

<h3>Q-Learning</h3>

As before, because $Q$-Learning is a control algorithm, we are interested in updating $Q$ rather than updating $V$ 

At a high level, we are mostly interested in the innermost part of the loop

That is where we choose an action and take a step in the environment and where we update the $Q$-table

So these are the two pieces we are going to focus on here 

```python
def choose_action(Q,s):
    if random() < epsilon:
        return random action
    else:
        return argmax{ Q(s, :) }
    
Inside the main loop:
    # choose action and execute
    a = choose_action(Q,s)
    s_prime, r, done = env.step(a)
```

When we choose an action, we are again going to use an epsilon greedy approach

So with small probability epsilon, we will choose a random action

Otherwise, we'll take the $\arg \max$ over $Q$ given a state $s$

Once we choose our action, we then take that action in the environment

---

When we update $Q$, we do something very subtle but important

That is when we calculate the target, we ignore whatever action we are going to take next

Instead, we assume that we will take the greedy action and take the max over $Q$ given the state $s'$

```python
# update Q
y = r + gamma * max_a_prime { Q(s_prime,a_prime) } # target
Q(s,a) = Q(s,a) + learning_rate * (y-Q(s,a))
```

This has two advantages

First, it means that in order to update $Q$, we don't have to wait until we obtain the next action $a'$ in the next iteration of the loop

Second, it makes $Q$ learning what is called an off-policy algorithm

This means that we can freely explore by our algorithm will update the $Q$ table as if we had acted greedily

---

<h3>Pseudocode</h3>

So here's what $Q$-learning looks like when we put it all together

```python
env = ... # given: some enviroment object
Q = random
for i in range(num_episodes):
    s = env.reset()
    done = false
    while not done:
        a = choose_action(Q,s)
        s_prime, r, done = env.step(a)
        
        # update Q
        y = r + gamma * max_a_prime { Q(s_prime, a_prime) }
        Q(s,a) = Q(s,a) + learning_rate * (y - Q(s,a))
        
        # important : update the current state
        s = s_prime
```

Luckily, it looks quite similar to the prediction problem pseudocode

First, we are given some environment object

Then we initialize $Q$ with random values

Then we enter a loop that goes for a predetermined number of episodes

Optionally we can keep going until $Q$ or the policy converges

Next, it's the same as what we had before 

We reset the environment and start back at the initial state

We initialize the `done` flag to `false`

Then we enter a loop for the episode, quitting only when we are done

Next, we use epsilon-greedy to choose an action

Then we take a step in the environment

We get back the state `s_prime`, the reward `r` and the `done` flag

Next, we create the target for the $Q$ update, taking the max over $Q$ given the state `s_prime`

Next we update `Q(s,a)` using the target we just calculated

Lastly, we update the current state `s` to be the next state `s_prime`

Finally, when we exit the loop, we have found the optimal policy

---

<h3>Summary</h3>

Let's summarize what we just did, since that was pretty ling 

First, we started by noting that Monte Carlo will not work for infinitely long episodes and that it's problematic in that we have to wait until an episode is over in order to do any learning

Instead, we make use of the fact that the return can be defined recursively

This allows us to approximate the return using only the next reward and the value at the next state

We also learned that reinforcement learning ultimately starts to look like supervised learning, where our target is not a real target, but rather an estimation made by our own model

We are essentially doing gradient descent on a $Q$-table 

Using this approach, it allows us to update our $Q$-table on every step because we only have to wait until we receive a single reward to make an update

We call such an approach online learning because the agent learns while it collects data

<h3>Random Search</h3>

Since we've just been introduced to the OpenAI Gym interface, we're not going to code anything too complex right away 

In next code section, we are going to add just a little bit of code to the code we just wrote to do a random search in the parameter space for a linear model

In other words we are going to take our state `s` dotted with a vector of weights `W`, and if this is greater than zero will do one action and if it's less than zero will do the other action

```
state.dot(params) > 0 --> do action 1
state.dot(params) < 0 --> do action 0
```

---

<h3>Pseudocode</h3>

Let's do a quick overview of the algorithm in pseudocode

```
for number of times we want to adjust the weights:
    new_weights = random
    for number of episodes we want to play ro decide whether to update the weights:
        Play episode
    if avg episode length > best so far:
        weights = new_weights
Play a final set of episodes to see how good my best weights again

```

What this does is we're first going to choose the number of times we want to attempt to adjust the weights

For that number of times, we're going to generate new weights randomly

Then we're going to play a number of episodes with these new weights

We need to try more than once just in case an episode is really long just by chance

From this we'll calculate the average episode length 

If the average episode length for these new weights is better than our best average so far, we'll keep these waits for later

Once we've done this a number of times, we've supposedly chosen some good weights that do well over a number of episodes so hopefully not just by chance

Then we play a final set of episodes to see how good our best weights perform

<h1>code</h1>

````python
import gym
import numpy as np
````

````python
class RSL: # Random Search Linear Model 
    def train(self,env,epochs=100,episodes=100):
        best_W = None
        best_avg = 0
        D = env.observation_space.shape[0]
        # save averages for plotting
        averages = []
        for epoch in range(epochs):
            if (epoch+1)%10 == 0:
                print('Epoch: ',epoch+1,' / ',epochs)
            # weights are uniformly distributed
            # between -1 and 1
            # we can also use randn as usual
            W = np.random.random(D)*2-1
            total_steps = 0
            for episode in range(episodes):
                s = env.reset()
                done = False
                while not done:
                    dot = s@W
                    if dot > 0:
                        a = 1
                    else:
                        a = 0
                    s, reward, done, _ = env.step(a)
                    total_steps += 1
            avg_steps = total_steps/episodes
            averages.append(avg_steps)
            if avg_steps > best_avg:
                best_W = W
                best_avg = avg_steps
        plt.plot(averages)
        plt.title('Average number of step for each weight configuration')
        plt.show()
        print("Average number of steps for best weights: ",best_avg)
        
````

````python
env = gym.make('CartPole-v0')
model = RSL()
````

````python
model.train(env)
````

````python
import gym
import numpy as np
import matplotlib.pyplot as plt
````

````python
env = gym.make('CartPole-v0')
````

````python
# next lets build our feature transformer
class FT: # Feature Transformer
    def __init__(self,n_bins=10):
        # we have one bin that spans values out of range
        # left are n_bins - 1
        
        # to make these ranges more accurate
        # we need to plot histograms of the state visited
        # whilist training the model
        cart_position_bins = np.linspace(-2.4,2.4,n_bins-1)
        cart_velocity_bins = np.linspace(-2,2,n_bins-1)
        pole_angle_bins = np.linspace(-0.4,0.4,n_bins-1)
        pole_velocity_bins = np.linspace(-3.5,3.5,n_bins-1)
        self.bins = [cart_position_bins,cart_velocity_bins,pole_angle_bins,pole_velocity_bins]
        self.n_bins = n_bins
    
    def to_num(self,s):
        base = self.n_bins
        val = 0
        for n,digit in enumerate(reversed(s)):
            val += digit*base**n
        return val
    def transform(self,s):
        s = [np.digitize([d],bins=d_bin)[0] for d,d_bin in zip(s,self.bins)]
        # next we convert this to a number of base n_bin sysyem
        # so if s_new = [0 0 0 0 ], then thats state 0
        # if s_new = [0 0 0 1], then thats state 1
        s = self.to_num(s)
        return s
````

````python
class QTB: # Q-learning + Table + Bins
    def __init__(self,env,n_bins=10):
        self.env = env
        self.transformer = FT(n_bins=10)
        
        num_states = n_bins ** env.observation_space.shape[0]
        num_actions = env.action_space.n
        self.Q = np.random.uniform(low=-1,high=1,size=(num_states,num_actions)) # Q table
     
    def get(self,s,get=None):
        x = self.transformer.transform(s)
        Q = self.Q[x]
        best_action = np.argmax(Q)
        max_Q = np.max(Q)        
        if get == 'value':
            return max_Q
        if get == 'action':
            return best_action
        
        return max_Q,best_action
        
    def epsilon_greedy(self,best_action,eps=0.1):
        if np.random.random() < eps:
            return np.random.choice(self.env.action_space.n)
        else:
            return best_action
    
    def sgd(self,s,a,G):
        x = self.transformer.transform(s)
        # so x is just a single number that indexes the state
        self.Q[x,a] += self.alpha*(G - self.Q[x,a])
                
    def play_episode(self,training = True):
        env = self.env
        s = env.reset()
        done = False
        episode_reward = 0
        while not done:
            a = self.get(s,get='action')
            if training == True:
                a = self.epsilon_greedy(a,self.eps)
            s_prime, r, done, _ = env.step(a)
            episode_reward += r
            
            if done and episode_reward < 200: # so we lost
                r = -300
            
            if training == True:
                G = r + self.gamma * self.get(s_prime,get='value')
                self.sgd(s,a,G)
            
            s = s_prime
        return episode_reward
      
    # the return of each episode are going to vary a lot
    # so it will be hard to see
    # we're interested in averaging the returns over 100 episodes
    # because the OpenAI documnetation says our agent is judged by how much it does over 100 episdoes
    def plot_running_avg(self,rewards):
        rewards = np.array(rewards)
        T = len(rewards)
        running_avg = np.zeros(T)
        for t in range(T):
            running_avg[t] = rewards[max(0,t-100):t+1].mean()
        plt.plot(running_avg)
        plt.title('running average')
        plt.show()
        
    def train(self,episodes = 10000,gamma = 0.90,alpha=1e-2,eps=0.1):
        rewards = []
        self.gamma = gamma
        self.alpha = alpha
        self.eps = eps
        for episode in range(episodes):
            reward = self.play_episode()
            rewards.append(reward)
            if (episode+1)%100 == 0:
                print('Episode: ',episode+1,'/',episodes,' reward = ',reward)
        plt.plot(rewards)
        plt.title('rewards per epoch')
        plt.show()
        self.plot_running_avg(rewards)
    
    def test(self,episodes=100):
        rewards = []
        for i in range(episodes):
            r = self.play_episode(training=False)
            rewards.append(r)
        print('Average test reward: ',np.mean(rewards))
````

````python
model = QTB(env)
````

````python
model.train()
````

````python
# as we can see we get a higher average than that of training
# this is due to turining of epsilon lambda
model.test()
````

<h3>RBF Networks</h3>

RBF stands for radial basis function

They allow us to proceed to the next step of using function approximation

There are two ways to think of an RBF network

<img src='extras/60.15.PNG' width='400'>

The first way to think of an RBF network is that it's really a linear model where we've done feature extraction first and the features happen to be RBF kernels

We'll be discussing what an RBF kernel is shortly

The second way to think of an RBF network is that it's a one hidden layer neural network with radial basss functions as the activation function

---

Keep in mind that when we first talked about neural networks we actually looked at them from the reverse perspective

We first looked at the neural network as a big nonlinear function approximater, but then we learned that one way to think of the hitting units is that they were automatically learning feature representations of the input

---

<h3>Radial Basis Function</h3>

So what is a Radial Baasis Function? 

$$\large \varphi(x) = \exp \left(- \frac{\Vert x - c \Vert^2}{\sigma^2} \right)$$

It's basically just a non-normalized the Gaussian bell curve centered at some center point which we'll call $c$ 

$c$ is a vector that lives in the same vector space as the input vectors

Notice how the function depends only on the distance between $x$ and $c$ 

So it doesn't matter if $x$ is to the left of $c$ or to the right of $c$ only the distance matters

That's where the term radial comes from

We can see that the maximum value was $1$ when $x$ is equal to $c$ 

It then approaches zero as $x$ goes away from $c$ in any direction

---

<h3>How do we choose $c$</h3>

One good question is how do we choose $c$

Another good question is how many $c$s should we choose

So the number of $c$s we choose is basically the number of hidden units there will be in the hidden layer

Each unit corresponds to a different RBF with a different center

These centers are sometimes called exemplars

There are a few different ways to choose exemplars

---

<h3>Support Vector Machines</h3>

In support vector machines or SVMs, the number of exemplars is equal to the number of training points, and in fact they are just the training points themselves

This is one of the reasons why SVMs have fallen out of favor

They don't scale to the data sizes that we have today 

Because each training sample becomes an exemplar, training and SVM is order $O(n^2)$ and prediction is $O(n)$, where $n$ is the training sample size

This is an important piece of deep learning history because SVMs were once thought to be superior to neural networks

---

<h3>Another method</h3>

Another method is just to sample a few points from the state space

This allows us to limit the number of exemplars we use

Remember that `env.observation_space.sample()` allows us to sample from the state space of the environment

At this point how many exemplars we choose is just like how many hidden units we should choose in a neural network

It's a hyper parameter that must be tuned

---

<h3>Implementation</h3>

Luckily, although we could, we won't write any code to do the RBF kernel transformation since these capabilities are already in `scikit-learn`

Also our implementation would be unnecessarily slow.

The `scikit-learn` implementation is called `RBF-sampler` and it implements a Monte Carlo algorithm that allows us to do the computation much faster

As usual this has a standard `scikit-learn` interface

```python
from sklearn.kernel_approximation import RBFSampler

sampler = RBFSampler()
sampler.fit(raw_data)
features = sampler.transform(raw_data)
```

We create an instance, call the `fit` function and then from that point onward we can call `transform`

---

<h3>The rest we have done before</h3>

So now that we know what an RBF is and how to do the feature transformation the rest should be easy

We pass the features into a linear regression model and use gradient descent to update the parameters of the linear regression model

Notice how unlike a feedforward neural network the features don't change as we learn 

The exemplars we chose at the beginning will remain the same forever

Although this might seem restrictive it actually works better than a plain feedforward neural network where we try to run stochastic back propagation through the whole network

---

<h3>Old perspective vs. New perspective</h3>

OK so that was the perspective where instead of using polynomials as the feature extractor, we use our RBF kernels as the feature and then put a linear model on top 

The other perspective is that this is actually a $1$ hidden layer neural network

$$\large \text{Feedforward net : } z_i = \tanh(W_i^Tx) = f(\text{cosine dist}(W_i,x))$$

$$\text{RBF net : } z_i = \exp \left( - \frac{\Vert x-c_i \Vert ^2}{\sigma^2}\right) = f(\text{squared dist}(c_i,x))$$

Remember that a one hit and layer neural network is just a nonlinear transformation at the hidden layer and then a linear model for the final layer

Well this is exactly what we're doing here

And even closer connection can be made if we recognize that a dot product is actually the cosine distance as well

In both cases we first calculate some kind of distance between the parameter and the input and then we apply a non linear function on top of that

This gives us the value at one hitting unit

It's an interesting connection but that's as far as it goes

---

<h3>Implementation Details</h3>

Now let's go through some practical implementation details

We may have noticed that the RBF kernel has a scale parameter which is like the variance of a Gaussian

$$\varphi(x) = \exp \left( - \frac{\Vert x - c \Vert^2}{\sigma^2}\right)$$

Since we don't actually know what scale is good or perhaps multiple scales are good, `scikit-learn` has tools that allow us to use multiple RBF kernels with different scales at the same time

This is called the feature Union, and it essentially allows us to use multiple feature transformers of any kind, not just RBFs, and then concatenate the individual feature vectors together to make one big set of feature vectors

```python
from sklearn.pipeline import FeatureUnion
```

We'll also be standardising the data before applying the RBF kernel 

`scikit-learn` has a class for this too, it's called `StandardScalar`

```python
from sklearn.preprocessing import StandardScaler
```

---

Another opportunity to make use of `scikit-learn` is the linear regression component with gradient descent 

`scikit-learn` learn already has a class for this called as `SGDRegressor`

So we are going to make use of that as well

Unlike typical `scikit-learn` models, instead of calling the `fit` function will be using the `partial_fit` function which allows us to do one step of gradient descent only, which is what we want

```python
from sklearn.linear_model import SGDRegressor

# functions

partial_fit(X,Y) # one step of gradient descent
predict(X)
```

---

At the same time `SGDRegressor` behaves a little strangely compared to what you would expect from a linear model

First if we don't call the `partial_fit` function at least once then it won't work because it hasn't initialised any parameters

But at the same time the first thing we need to do is make a prediction from it, because to do $Q$-learning we need to take the max over all actions 

In order to get around this problem we'll call `partial_fit` with some dummy values specifically with the start state as input and zero as the target

```python
input_ = transform(env.reset())
target = 0
model.partial_fit(input_,target)
```

---

The second thing that's weird about `SGDRegressor` is that by setting zero as the target, this is going to make all the predictions zero for a while

This is weird because if we were actually using a linear set of weights to make the prediction, then the prediction wouldn't always be zero

However this quirk is actually useful because for our next task which is the mountain car environment, at every time step we get a reward of $-1$

That means a $Q$ prediction of 0 is higher than any return we can get, and hence it allows us to use the optimistic initial values method of exploration, which we looked at in a previous notebook, instead of epsilon greedy

We're given the option of trying both in the code

---

<h3>Prove it to yourself</h3>

Here's some code to try if we want to prove this to ourselves

````python
from sklearn.linear_model import SGDRegressor
model = SGDRegressor()
model.partial_fit([[0,0]],[0])

p = model.predict([[0,0]])
print(p) # makes sense, we get zeros

p = model.predict([[0,1]]) 
print(p) # hmmmmm, still zeros

p = model.predict([[99,99]])
print(p) # still zeros

# in fact the weights a have been initialised to 0s
print(model.coef_,model.intercept_)
````

<h3>One model per action</h3>

The next implementation detail we need to discuss is a technique used by Depp Q-learning as well 

Specifically, instead of trying to model $Q(s,a)$, with $x$ as the input vector, that is some feature transformation on the state $s$ and action $a$, we'll instead just make $x$ the transformation of the state $s$ 

Since the actions are discrete, we'll instead create a different linear model for each action

<img src = 'extras/60.16.PNG' width = '200'>

So in the case of mountain car where we have three actions left right and do nothing, we'll have three linear models all giving us the $Q$ for a separate action 

Another way to think about this is that it's a neural network with three output nodes

This is exactly the approach we'll be taking when we do Deep Q-learning

<h3>Mountain Car</h3>

And lastly, we mentioned earlier that we are going to be looking at the mountain environment next

If we don't know what mountain car is you can visit the wiki page and the documentation page

It basically works like a swing

We're in a car trying to get to the top of the mountain, but our car can't generate enough force to go straight up the mountain

We therefore need to swing back and forth to gain enough momentum so that we can be propelled to the top

If we look at the Wiki (see <a href='https://github.com/openai/gym/wiki/MountainCar-v0'>here</a>), we'll see that mountain cars states only have two variables, position and velocity

<img src='extras/60.17.PNG' width='300'>

Unlike cartpole, the velocity cannot be unrealistically large, so that makes this problem a little easier

<h3>Cosst-To-Go function</h3>

One of the functions we'll be plotting is called the cost to go function which is a strange name but it's just the negative of the value function

This is what they call it in Sutton and Barto

Because the state space is only two variables, this allows us to plot the cost to go function as a three dimensional plot which we'll be doing in the code

<h1>code</h1>

````python
import gym
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.kernel_approximation import RBFSampler
from sklearn.linear_model import SGDRegressor
from matplotlib.cm import coolwarm # used in cost-to-go
````

````python
env = gym.make('MountainCar-v0')
````

````python
# next lets build our feature transformer
# last time this was with bins
# this time we will be using RBF kernels
# this will be the union of n RBF networks
# each with a different scale

class FT: # Feature Transformer
    def __init__(self,env,n_samples = 10000,D=500,scales=[5,2,1,0.5]):
        # gather samples
        samples = []
        for _ in range(n_samples):
            samples.append(env.observation_space.sample())
        samples = np.array(samples)
        self.scaler = StandardScaler()
        samples = self.scaler.fit_transform(samples)
        
        samplers = []
        
        if type(D) != list:
            D = [D]*len(scales)
        
        for i in range(len(scales)):
            sampler = RBFSampler(gamma=scales[i],n_components=D[i])
            name = 'rbf' + str(i)
            samplers.append((name,sampler))

        # create our feature union
        self.FU = FeatureUnion(samplers)
        self.FU.fit(samples)
        
            
    def transform(self,x):
        x = self.scaler.transform([x])
        x = self.FU.transform(x)
        return x
````

````python
# next we define our SGDRegressor
class LinearModel:
    def __init__(self,start_x,learning_rate):
        # linear model
        lm = SGDRegressor(learning_rate=learning_rate) 
        # as we discussed, fit on zeors
        lm.partial_fit(start_x,[0])
        # weights will be all 0s, suitable for UCB
        self.lm = lm
        
    def sgd(self,x,y):
        return self.lm.partial_fit(x,[y])
        
    def predict(self,x):
        return self.lm.predict(x)
````

````python
class QALM: # Q learning + Approximation + Linear Model
    def __init__(self,env,gamma=0.99,learning_rate='constant'):
        self.gamma = gamma
        self.featuriser = FT(env)
        # recall we'll have a seperate model for each action
        A = env.action_space.n
        # linear models
        start_x = self.featuriser.transform(env.reset())
        self.LMs = [LinearModel(start_x,learning_rate) for a in range(A)]
        self.env = env
    
    def epsilon_greedy(self,best_action,eps=0.1):
        if np.random.random() < eps:
            return np.random.choice(self.env.action_space.n)
        else:
            return best_action
        
    def get(self,s,get=None):
        x = self.featuriser.transform(s)
        Q = [lm.predict(x)[0] for lm in self.LMs]
        best_action = np.argmax(Q)
        max_Q = np.max(Q)        
        if get == 'value':
            return max_Q
        if get == 'action':
            return best_action
        
        return max_Q,best_action
    
    def play_episode(self,training = True,render = False):
        env = self.env
        s = env.reset()
        done = False
        episode_reward = 0
        while not done:
            if render:
                env.render()
            a = self.get(s,get='action')
            if training == True:
                # as we stated the initial weights of SGD regressor are 0s
                # these are optimistic values 
                # compared to the rewards that Mountain Car returns (-1)
                # and hence we can skip using epsilon greedy 
                # (feel free to comment the line or pass epsilon = 0)
                # still, using it may lead to better results
                a = self.epsilon_greedy(a,self.eps)
                # pass
                
            s_prime, r, done, _ = env.step(a)
            episode_reward += r
            if training == True:
                G = r + self.gamma * self.get(s_prime,get='value')
                self.LMs[a].sgd(self.featuriser.transform(s),G)
            
            s = s_prime
        if render:
            env.close()
        return episode_reward
      
    # the return of each episode are going to vary a lot
    # so it will be hard to see
    # we're interested in averaging the returns over 100 episodes
    # because the OpenAI documnetation says our agent is judged by how much it does over 100 episdoes
    def plot_running_avg(self,rewards):
        rewards = np.array(rewards)
        T = len(rewards)
        running_avg = np.zeros(T)
        for t in range(T):
            running_avg[t] = rewards[max(0,t-100):t+1].mean()
        plt.plot(running_avg)
        plt.title('running average')
        plt.show()
        
    def plot_cost_to_go(self, num_tiles=20):
        env = self.env
        x = np.linspace(env.observation_space.low[0], env.observation_space.high[0], num=num_tiles)
        y = np.linspace(env.observation_space.low[1], env.observation_space.high[1], num=num_tiles)
        X, Y = np.meshgrid(x, y)

        Z = np.zeros((num_tiles,num_tiles))
        for i in range(num_tiles):
            for j in range(num_tiles):
                s = (X[i,j],Y[i,j])
                x = self.featuriser.transform(s)
                Q_s =  [lm.predict(x)[0] for lm in self.LMs]
                Z[i][j] = -np.max(Q_s) # - V(s)

        fig = plt.figure(figsize=(10, 5))
        ax = fig.add_subplot(111, projection='3d')
        surf = ax.plot_surface(X, Y, Z,
            rstride=1, cstride=1, cmap=coolwarm, vmin=-1.0, vmax=1.0)
        ax.set_xlabel('Position')
        ax.set_ylabel('Velocity')
        ax.set_zlabel('Cost-To-Go == -V(s)')
        ax.set_title("Cost-To-Go Function")
        fig.colorbar(surf)
        plt.show()
    
    def train(self,episodes = 500):
        rewards = []
        for episode in range(episodes):
            # decaying epsilon
            self.eps = 0.1*(0.97**episode)
            reward = self.play_episode()
            rewards.append(reward)
            if (episode+1)%10 == 0:
                print('Episode: ',episode+1,'/',episodes,' reward = ',reward)
        plt.plot(rewards)
        plt.title('rewards per epoch')
        plt.show()
        self.plot_running_avg(rewards)
        self.plot_cost_to_go()
    
    def test(self,episodes=100):
        rewards = []
        for i in range(episodes):
            r = self.play_episode(training=False)
            rewards.append(r)
        print('Average test reward: ',np.mean(rewards))
    
````

````python
model = QALM(env)
````

````python
model.train()
````

````python
# our model seems not to improve at the beggining
# but in fact it is
# recall that the max num of steps we can take is 200
# at first episode the model may take 4000 steps to finish
# at the second episode it may take 3000 episode to finish
# that a great imporvmement, but in both cases we wont get to see it
# since the enviroment returns done = True after only 200 moves
# so the model is improving
# but we can only see the improvement once it gets below 200 moves
````

````python
model.test()
````

````python
# see agent playing an episode
model.play_episode(training=False,render=True)
````

<h3>Modifications for CartPole</h3>

In this section we're going to modify our of network code to work for CartPole instead of Mountain Car


This section is all about what those modifications are

---

<h3>SGDRegressor</h3>

The first modification we're going to make is we're going to build our own `SGDRegressor`

This is going to give us some practice with building a gradient descent model in case we don't remember

In order to make sure it has the same API as `scikit-learn`'s `SGDRegressor`, we need to implement a `partial_fit` function and `predict` function

Note that doing this means we won't be using the optimistic initial values method because the predictions will be determined by the weight


---

<h3>RBF exemplars</h3>

The next modification is important

Recall that in the mountain car code we sampled from the observations based to get exemplars for our RBF kernels 

This is OK for mountain car because the max and min values for each state variable are small 

For CartPole this is not the case

Recall that the velocities can approach infinity.

Unfortunately the OpenAI Gym sample function just samples uniformly from the state space which is not representative of the states we would find yourself in in a real episode

That means the samples would make poor exemplars

Instead what we need to do is guess what the plausible ranges are or if we wanted to get really fancy we could play a few thousand episodes and sample from those values

Because the state space is different, we'll also want different scales for our RBF kernels too

<h1>code</h1>

````python
import gym
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.kernel_approximation import RBFSampler
from sklearn.linear_model import SGDRegressor
````

````python
env = gym.make('CartPole-v0')
````

````python
class FT: # Feature Transformer
    def __init__(self,env,n_samples = 20000,n_components=1000,scales=[0.05,1.0,0.5,0.1]):
        # this time we sample from a unifrom distribution
        # around the points we think to be good representetives of what we'll see during training
        D = env.observation_space.shape[0]
        samples = np.random.random((n_samples,D))*2-2
        self.scaler = StandardScaler()
        samples = self.scaler.fit_transform(samples)
        
        samplers = []
        
        if type(n_components) != list:
            n_components = [n_components]*len(scales)
        
        for i in range(len(scales)):
            sampler = RBFSampler(gamma=scales[i],n_components=n_components[i])
            name = 'rbf' + str(i)
            samplers.append((name,sampler))

        # create our feature union
        self.FU = FeatureUnion(samplers)
        self.FU.fit(samples)
        self.D =  sum(n_components)
        
            
    def transform(self,x):
        x = self.scaler.transform([x])
        x = self.FU.transform(x)
        return x
````

````python
# next we define our SGDRegressor
class LinearModel:
    def __init__(self,D,learning_rate=1e-3):
        # recall one model per action
        # so its just D
        self.W = np.random.randn(D)/np.sqrt(D)
        self.lr = learning_rate
        
    def sgd(self,x,t):
        y = x@self.W
        delta = t-y
        self.W += self.lr*(delta@x)
        
    def predict(self,x):
        return x@self.W
````

````python
class QALM: # Q learning + Approximation + Linear Model
    def __init__(self,env,gamma=0.99,learning_rate=1e-1):
        self.gamma = gamma
        self.featuriser = FT(env)
        # recall we'll have a seperate model for each action
        A = env.action_space.n
        # linear models
        D = self.featuriser.D
        self.LMs = [LinearModel(D,learning_rate) for a in range(A)]
        self.env = env
    
    def epsilon_greedy(self,best_action,eps=0.1):
        if np.random.random() < eps:
            return np.random.choice(self.env.action_space.n)
        else:
            return best_action
        
    def get(self,s,get=None):
        x = self.featuriser.transform(s)
        Q = [lm.predict(x)[0] for lm in self.LMs]
        best_action = np.argmax(Q)
        max_Q = np.max(Q)        
        if get == 'value':
            return max_Q
        if get == 'action':
            return best_action
        
        return max_Q,best_action
    
    def play_episode(self,training = True,render = False):
        env = self.env
        s = env.reset()
        done = False
        episode_reward = 0
        while not done:
            if render:
                env.render()
            a = self.get(s,get='action')
            if training == True:
                # as we stated the initial weights of SGD regressor are 0s
                # these are optimistic values 
                # compared to the rewards that Mountain Car returns (-1)
                # and hence we can skip using epsilon greedy 
                # (feel free to comment the line or pass epsilon = 0)
                # still, using it may lead to better results
                a = self.epsilon_greedy(a,self.eps)
                # pass
                
            s_prime, r, done, _ = env.step(a)
            episode_reward += r
            if training == True:
                G = r + self.gamma * self.get(s_prime,get='value')
                self.LMs[a].sgd(self.featuriser.transform(s),G)
            
            s = s_prime
        if render:
            env.close()
        return episode_reward
      
    # the return of each episode are going to vary a lot
    # so it will be hard to see
    # we're interested in averaging the returns over 100 episodes
    # because the OpenAI documnetation says our agent is judged by how much it does over 100 episdoes
    def plot_running_avg(self,rewards):
        rewards = np.array(rewards)
        T = len(rewards)
        running_avg = np.zeros(T)
        for t in range(T):
            running_avg[t] = rewards[max(0,t-100):t+1].mean()
        plt.plot(running_avg)
        plt.title('running average')
        plt.show()
            
    def train(self,episodes = 500):
        rewards = []
        for episode in range(episodes):
            # decaying epsilon
            self.eps = 1.0/(np.sqrt(episode+1))
            reward = self.play_episode()
            rewards.append(reward)
            if (episode+1)%10 == 0:
                print('Episode: ',episode+1,'/',episodes,' reward = ',reward)
        plt.plot(rewards)
        plt.title('rewards per epoch')
        plt.show()
        self.plot_running_avg(rewards)
    
    def test(self,episodes=100):
        rewards = []
        for i in range(episodes):
            r = self.play_episode(training=False)
            rewards.append(r)
        print('Average test reward: ',np.mean(rewards))   
````

````python
# finally we are going to do a warmup in tensorflow
# just the same code but we modify the SGDRegressor

import gym
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.kernel_approximation import RBFSampler
from sklearn.linear_model import SGDRegressor
import tensorflow as tf
from tensorflow.keras.models import Model
````

````python
env = gym.make('CartPole-v0')
````

````python
class FT: # Feature Transformer
    def __init__(self,env,n_samples = 20000,n_components=1000,scales=[0.05,1.0,0.5,0.1]):
        # this time we sample from a unifrom distribution
        # around the points we think to be good representetives of what we'll see during training
        D = env.observation_space.shape[0]
        samples = np.random.random((n_samples,D))*2-2
        self.scaler = StandardScaler()
        samples = self.scaler.fit_transform(samples)
        
        samplers = []
        
        if type(n_components) != list:
            n_components = [n_components]*len(scales)
        
        for i in range(len(scales)):
            sampler = RBFSampler(gamma=scales[i],n_components=n_components[i])
            name = 'rbf' + str(i)
            samplers.append((name,sampler))

        # create our feature union
        self.FU = FeatureUnion(samplers)
        self.FU.fit(samples)
        self.D =  sum(n_components)
        
            
    def transform(self,x):
        x = self.scaler.transform([x])
        x = self.FU.transform(x)
        return x
````

````python
# next we define our SGDRegressor
class LinearModel:
    def __init__(self,D,learning_rate=1e-3):
        # recall one model per action
        # so its just D
        W = np.random.randn(D,1)/np.sqrt(D)
        self.W = tf.Variable(W)
        self.lr = learning_rate
        self.optimiser = tf.keras.optimizers.SGD(learning_rate=learning_rate)

        
    def sgd(self,x,t):
        with tf.GradientTape() as tape:
            y = tf.matmul(x,self.W)
            delta = t-y
            loss = delta@delta
        grads = tape.gradient(loss, [self.W])
        self.optimiser.apply_gradients(zip(grads, [self.W]))   
        
    def predict(self,x):
        return x@self.W.numpy()
````

````python
class QALM: # Q learning + Approximation + Linear Model
    def __init__(self,env,gamma=0.99,learning_rate=1e-1):
        self.gamma = gamma
        self.featuriser = FT(env)
        # recall we'll have a seperate model for each action
        A = env.action_space.n
        # linear models
        D = self.featuriser.D
        self.LMs = [LinearModel(D,learning_rate) for a in range(A)]
        self.env = env
    
    def epsilon_greedy(self,best_action,eps=0.1):
        if np.random.random() < eps:
            return np.random.choice(self.env.action_space.n)
        else:
            return best_action
        
    def get(self,s,get=None):
        x = self.featuriser.transform(s)
        Q = [np.squeeze(lm.predict(x)) for lm in self.LMs]
        best_action = np.argmax(Q)
        max_Q = np.max(Q)        
        if get == 'value':
            return max_Q
        if get == 'action':
            return best_action
        
        return max_Q,best_action
    
    def play_episode(self,training = True,render = False):
        env = self.env
        s = env.reset()
        done = False
        episode_reward = 0
        while not done:
            if render:
                env.render()
            a = self.get(s,get='action')
            if training == True:
                # as we stated the initial weights of SGD regressor are 0s
                # these are optimistic values 
                # compared to the rewards that Mountain Car returns (-1)
                # and hence we can skip using epsilon greedy 
                # (feel free to comment the line or pass epsilon = 0)
                # still, using it may lead to better results
                a = self.epsilon_greedy(a,self.eps)
                # pass
                
            s_prime, r, done, _ = env.step(a)
            episode_reward += r
            if training == True:
                G = r + self.gamma * self.get(s_prime,get='value')
                self.LMs[a].sgd(self.featuriser.transform(s),G)
            
            s = s_prime
        if render:
            env.close()
        return episode_reward
      
    # the return of each episode are going to vary a lot
    # so it will be hard to see
    # we're interested in averaging the returns over 100 episodes
    # because the OpenAI documnetation says our agent is judged by how much it does over 100 episdoes
    def plot_running_avg(self,rewards):
        rewards = np.array(rewards)
        T = len(rewards)
        running_avg = np.zeros(T)
        for t in range(T):
            running_avg[t] = rewards[max(0,t-100):t+1].mean()
        plt.plot(running_avg)
        plt.title('running average')
        plt.show()
            
    def train(self,episodes = 300):
        rewards = []
        for episode in range(episodes):
            # decaying epsilon
            self.eps = 1.0/(np.sqrt(episode+1))
            reward = self.play_episode()
            rewards.append(reward)
            if (episode+1)%10 == 0:
                print('Episode: ',episode+1,'/',episodes,' reward = ',reward)
        plt.plot(rewards)
        plt.title('rewards per epoch')
        plt.show()
        self.plot_running_avg(rewards)
    
    def test(self,episodes=100):
        rewards = []
        for i in range(episodes):
            r = self.play_episode(training=False)
            rewards.append(r)
        print('Average test reward: ',np.mean(rewards))   
````

````python
model = QALM(env)
````

````python
model.train()
````

<h3>Catastrophic Forgetting</h3>

One downside to this is the catastrophic forgetting effect

This idea has gotten lots of attention recently because researchers have found ways to successfully implement transfer learning

In other words they could train an AI in one game and use parts of the neural network to improve training on another game

They then showed that the neural network was still able to play the old game so the neural network didn't simply just learn the new game and forget the old game

The reason this research is noteworthy is because this is not how neural networks work by default

In a more typical case we actually expect the neural network to forget how to play the first game 

---

But this forgetting effect doesn't only apply across different tasks 

As we've seen, when we do stochastic and batch gradient descent the cost function can jump around quite a bit

We would say the effect seems to be more pronounced on highly non linear regression problems, which is exactly what we're working with now

Ideally we'd like the data in our cost function to represent the true distribution of the data

When we train on all training data simultaneously, this is actually only an approximation of the true distribution

It's like doing a clinical trial for a drug 

We might have a sample from thousands of people then we hope that they are representative of the population as a whole

---

So when you use batch or stochastic gradient descent this approximation becomes even worse

What makes it even worse in this notebook than a typical stochastic gradient descent, is that there is a large bias in the ordering of the training samples

In particular we're always proceeding from the start state to the end state

There is no randomization which is recommended if we're doing stochastic or batch gradient descent

So what happens is we'll train the neural network to do well on a cost that's with respect to the most recent samples but then we've forgotten about the earlier samples

---

Drop out regularisation may be able to help since it means some of the weights that connect some of the nodes will randomly not be effected during any particular training iteration

We can also try other types of regularisation and different architectures as well
