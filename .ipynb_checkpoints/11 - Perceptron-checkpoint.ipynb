<h1>Math</h1>

Now we will talk about perceptrons , al inear classifier which has an important role in the history of ML (but we will skip a hsitory lesson) , and is seen as the precursor to neural network , as famous as it is it has its limitations , for ex it ant solve the xor problem 

Perceptron is naturally a binary classifier

instead if targets = {0,1} , we will use targets = {1,-1} (we will see why)

prediction is simple , just like any other linear classifier calculate $w^Tx+b$

if $w^Tx+b = 0$ , we fall directly on the sepeerating line/hyperplane

if $w^Tx+b > 0$ , we predict +1

if $w^Tx+b = 0$ , we predict -1

so prediction = $sign(w^Tx+b)$

the training procedure is the interesting part

classification rate should go up overall as we step through each epoch

Psuedocode:

```python
w = random vector (b=0)
for epoch in range(max_epochs):
    get all currently misclassified examples
    if no misclassified examples:
        break
    x,y = randomly select one misclassified example
    w = w + eta*y*x , eta = 1.0 , 0.1 , 0.001 , etc. typically 
```

But how does w = w + $\eta$yx help put w in right direction ?

recall that a line can be defined by its normal vector (a line perpendicular to it)

for a line of equation : ax +by + c , the normal  n = (a,b) 

<img src='extras/11.1.PNG'></img>

so in our case if the line is we refer to the normal vector as w

Bias term usually tells us wehere we intersect the vertical axis but we can ignore it since it can be always absorbed into w by adding a column of 1s to X

initial model : $y = w_0 + w_1x_1 + w_2x_2$

new model = $y = w_0x_0 + w_1x_1 + w_2x_2$ , $x_0 = 1$

therefore any model where we dont consider bias explicitly can be assumed to contain a bias term anyway

back to our original , we now know that w is normal to the seperating line ax +by + c = 0 , using this and w = w + $\eta$yx can we understand the perceptron ?


<img src='extras/11.2.PNG'></img>

The black lines are vector w and the seperator line $w^T x$ perpendicular to it

The red line represents the vector of a misclassified point x

since x is above the line we classify it as +1 , Mathematically , since x is above the line , the angle between w and x is < 90 $^\circ$ then $w^T x$ is > 0 , again the we classify  x as +1 

since x is misclassified , then the true label of x = -1

this means we perform $w = w + \eta y x$ , say $\eta = 1$ then we set $w = w + yx$ , $w = w + (-1)x$ , $w = w-x$

here y is the true label (the target)

the green lines show (w-x) and the new seperator line $(w-x)^T x$ , updating w causes the seperator line to shift such that it classifies x correctly , or at least it gets closer to x so it classifies x correctly next time we update , the amount of shift would also depend on $\eta$

Now lets consider the opposite scenario

<img src='extras/11.3.PNG'></img>

again the black lines are vector w and the seperator line $w^T x$ perpendicular to it

The red line represents the vector of a misclassified point x

since x is below the line we classify it as -1 , Mathematically , since x is below the line , the angle between w and x is > 90 $^\circ$ then $w^T x$ is < 0 ,  so we classify  x as +1 

since x is misclassified , then the true label of x = +1

this means we perform $w = w + \eta y x$ , say $\eta = 1$ then we set $w = w + yx$ , $w = w + (+1)x$ , $w = w+x$

here y is the true label (the target)

the green lines show (w+x) and the new seperator line $(w+x)^T x$ , updating w causes the seperator line to shift such that it classifies x correctly , or at least it gets closer to x so it classifies x correctly next time we update , the amount of shift would also depend on $\eta$

````python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
````

````python
def get_data(): # this will create linearly seperable data that we can plot
    # these are the true weights the perceptron should find
    w = np.array([-0.5,0.5]) 
    b = 0.1
    # these are random weights
    X = np.random.random((300,2))*2 - 1
    # these are labels
    Y = np.sign(X@w + b)
    return X,Y

# we will use the KNN on the MNIST dataset
# https://www.kaggle.com/oddrationale/mnist-in-csv

def get_mnist(limit=None): 
    df = pd.read_csv('datasets/MNIST/train.csv')
    data = df.to_numpy()
    np.random.shuffle(data)
    X = data[:,1:]/255.0
    Y = data[:,0]
    # data is too large , so training would take too much time
    # of course to little data is not good and leads to overfitting
    if limit is not None:
        X,Y = X[:limit] , Y[:limit]
    return X,Y

def get_xor():
    # we will make the XOR grid like before
    N = 200
    X = np.random.random((N, 2))*2 - 1
    # generate labels
    T = np.zeros(N)
    T[(X[:,0] < 0) & (X[:,1] > 0)] = 1
    T[(X[:,0] > 0) & (X[:,1] < 0)] = 1
    return X,T
    
    
# now again for the donut problem 
def get_donut():
    N = 200
    D = 2
    R_inner = 5
    R_outer = 10

    # in a polar system , this is the distance from origin
    # we sample from a uniform distribution , set mean to 0
    R1 = np.random.randn((N//2)) + R_inner
    # and this is the angle
    # 2pi * (number from 0-1) so we get thetas between 0-2*pi
    theta = 2*np.pi*np.random.random(N//2)
    # in a cartesian system x component is R cos theta , y component is R sin theta
    X_inner = np.concatenate([[R1*np.cos(theta)],[R1*np.sin(theta)]]).T

    R2 = np.random.randn((N//2)) + R_outer
    theta = 2*np.pi*np.random.random(N//2)
    X_outer = np.concatenate([[R2*np.cos(theta)],[R2*np.sin(theta)]]).T

    X = np.concatenate([X_inner , X_outer])
    T = np.array([0]*(N//2) + [1]*(N//2))
    
    return X,T
````

````python
class Perceptron:
    def fit( self,X,Y,lr = 1.0 , epochs = 1000 ):
        N,D = X.shape
        ones = np.atleast_2d(np.ones(N)).T
        X = np.hstack((ones,X))
        w = np.random.randn(D+1)
        
        for i in range(epochs):
            preds = np.sign(X@w)
            wrong = preds != Y
            # get wrong point
            if (preds==Y).mean() == 1: # all correct
                break
            x = X[wrong][0]
            y = Y[wrong][0]
            w = w + lr*y*x
        
        self.w = w
        print(w)
        
    def predict(self,X):
        N,D = X.shape
        ones = np.atleast_2d(np.ones(N)).T
        X = np.hstack((ones,X)) 
        return np.sign(X@self.w)
    
    def score(self,X,Y):
        preds = self.predict(X)
        return np.mean(preds==Y)
        
            
````

````python
# first lets try our linearly seperable data
X,Y = get_data()
plt.scatter(X[:,0],X[:,1],c=Y,s=100,alpha=0.5)
plt.show()
````

````python
Ntrain = len(Y)//2
Xtrain,Ytrain = X[:Ntrain] , Y[:Ntrain]
Xtest,Ytest = X[Ntrain:] , Y[Ntrain:]
````

````python
model = Perceptron()

t0 = datetime.now()
model.fit(Xtrain,Ytrain)
print('Training time ',(datetime.now()-t0))

t0 = datetime.now()
print('Train accuracy ',model.score(Xtrain,Ytrain))
print('Time to compute train accuracy ',(datetime.now()-t0) , " Train size: ", len(Ytrain))

t0 = datetime.now()
print('Test accuracy ',model.score(Xtest,Ytest))
print('Time to compute test accuracy ',(datetime.now()-t0) , " Test size: ", len(Ytest))
````

````python
# now for mnist
X,Y = get_mnist()
# we are doing just binary classification
idx = np.logical_or(Y == 0 , Y == 1)
X = X[idx]
Y = Y[idx]

Y[Y == 0 ] = -1 # remember labels are {-1,1} not {0,1}

Ntrain = len(Y)//2

Xtrain,Ytrain = X[:Ntrain] , Y[:Ntrain]
Xtest,Ytest = X[Ntrain:] , Y[Ntrain:]
````

````python
model = Perceptron()

t0 = datetime.now()
model.fit(Xtrain,Ytrain)
print('Training time ',(datetime.now()-t0))

t0 = datetime.now()
print('Train accuracy ',model.score(Xtrain,Ytrain))
print('Time to compute train accuracy ',(datetime.now()-t0) , " Train size: ", len(Ytrain))

t0 = datetime.now()
print('Test accuracy ',model.score(Xtest,Ytest))
print('Time to compute test accuracy ',(datetime.now()-t0) , " Test size: ", len(Ytest))
````

````python
# now lets show that the perceptron fails to solve the xor problem
# this is to be expected since the perceptron is a linear classifier
# now for mnist
X,Y = get_xor()
# we are doing just binary classification
idx = np.logical_or(Y == 0 , Y == 1)
X = X[idx]
Y = Y[idx]

Y[Y == 0 ] = -1 # remember labels are {-1,1} not {0,1}

Ntrain = len(Y)//2

Xtrain,Ytrain = X[:Ntrain] , Y[:Ntrain]
Xtest,Ytest = X[Ntrain:] , Y[Ntrain:]
````

````python
model = Perceptron()

t0 = datetime.now()
model.fit(Xtrain,Ytrain)
print('Training time ',(datetime.now()-t0))

t0 = datetime.now()
print('Train accuracy ',model.score(Xtrain,Ytrain))
print('Time to compute train accuracy ',(datetime.now()-t0) , " Train size: ", len(Ytrain))

t0 = datetime.now()
print('Test accuracy ',model.score(Xtest,Ytest))
print('Time to compute test accuracy ',(datetime.now()-t0) , " Test size: ", len(Ytest))
````

<h1>Math</h1>

we want now to know what loss function is optimised by the Perceptron

In ML , we hava a prerference to formulate loss functions then optimise them , but sometimes it can be ad-hoc , we may invent an algorithm that solves a problem and only discover later that it optimises a certain loss function

The loss function of the perceptron is defined as follows :

$$L(y,\hat y) = - \sum^{N}_{i=1} y_i(w^T x_i) 1(y_i \neq \hat y_i)$$

the 1 function , is defined as follows :

1(true) returns 1

1(false) returns 0

if $y_i = \hat y_i$ , the 1 function return 0 and the entire term is cancelled , so correctly classified points do not contribute to the loss function

conversely $y_i \neq\hat y_i$ , the 1 function return 1 and contributes to loss function , so only misclassified points contribute to loss function


For misclassifiion , the predicted label $w^T x$ and the true label $y$ must be of opposite signs , therefore L >= 0

Also the further the $x_i$ from the hyper plane the larger the loss 

this is to be expected , since if $x_i$ was very close to hyperplane but is still wrong

the angle is then almost $90^\circ$ , and therefore $w^Tx_i = |w| |x_i| cos(89.99)$ which is almost 0

next imagine $x_i$ being almost parallel to w , the angle becomes almost $90^\circ$

therefore $w^Tx_i$ is almost $w^Tx_i = |w| |x_i| cos(89.99)$ which is the mac value

$x_i$ being very wrong causes a bigger loss

now if we take the derivative , we find that :

$$\frac {\partial L}{\partial w} = - \sum^{N}_{i=1} y_i x_i 1(y_i \neq \hat y_i)$$

seeing this ,we can clearly say the the algorithm we implemented before was stochastic (one point at a time) gradient descent
