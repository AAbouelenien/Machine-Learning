$$p(y|x) = softmax \left( V^Tf \left( W^Tx \right)\right)$$

but the weights we had were random so the output prediction are also random , and hence not accurate

so we want to change the weights so that the predictions are accurate

recall that the training data consists of (input,target) pairs and we want to make our prediction y as close to targets t as possible

as we did in logistic regression , we will create a cost function and try to minimise it (using gradient descent) , though in the case of neural network this is harder since equations are more complex

in neural networks gradient descent has a special name Backpropagation

Backpropagation is recursive in nature and allows us to find gradient in a systematic way 

This recursiveness will also allow us to do backpropagation for neural networks of arbitary depth without much added complexity


To train a neural network we will have to go through 2 steps:

1) define loss function

2) minimise loss function wrt weights

<h1>Math</h1>

now we want to discuss the cost we want to minimise in our cost function 

lets begin by the simplest error possible the squared error (from linear regression) :

$$SE = \sum^N_{n=1} {\left( {target}_n - {prediction}_n\right)}^2 = \sum^N_{n=1}{\left( t_n - y_n \right)}^2$$

if our prediction goes further away from target , error becomes larger , so it makes sense to want to minimise this error to train our model

notice how in order to get the error we need first to get the predictions , so we need to feedforward first

Minimising the squared error is the same as maximising the log-likelihood when the error is Gaussian-distributed (discussed thorougly in linear regression)

lets consider what happens if each target $t_n$ is a gaussian-distributed random variable with mean y_n so we say:

$$t_n \sim N(y_n , \sigma^2)$$

given this info we can set up the likelihood function as follows :

$$Likelihood = L = \prod^N_{n=1} \frac{1}{\sqrt{2\pi{\sigma}^2}} \exp \left( - \frac{1}{2\sigma^2} {\left( t_n - y_n \right)}^2 \right)$$

this might seem like a typical MLE (maximum likelihood estimation) problem , but its actually more advanced since $y_n$ it self is not meant to be optimised

we want to maximise likelihood wrt model weights w , we can say that $y_n$ is a function of the weights and that we want the weights that maximise the likelihood , so we write

$$y_n = f(x_n;w)$$
$$w = arg \max_w L$$

next , as usual , we maximise the log-likelihood , once again this makes math easier and is valid since log is a monotonically increasing function

$$ logL = \sum^N_{n=1} \log \left[ \frac{1}{\sqrt{2\pi{\sigma}^2}} \exp \left( - \frac{1}{2{\sigma}^2} {(t_n - y_n)}^2\right)\right]$$

doing so we get :

$$ logL = \sum^N_{n=1} \left[ \log  \frac{1}{\sqrt{2\pi{\sigma}^2}} + \left( - \frac{1}{2{\sigma}^2} {(t_n - y_n)}^2 \right) \right]$$

this can be rewritten as :

$$logL = C_1 - C_2 \sum^N_{i=1} {(t_n - y_n)}^2$$

so our loss function is the -ve log likelihood ,  recall that mimising the loss is the same as minimising the -ve log likelihood which is the same as maximising the loglikelihood

but the constants dont actually matter , the may yeild a different value for logL but that is irrelevant , all we care about are the values of w , and that they minimise loss no matter what the scale is 


So we proven that minimising squared error is the same as maximising likelihood

Now we discuss a slightly more complicated scenario , where we use the binary corss-entropy as a cost function

$$Loss = - \sum^N_{n=1} \left[ t_n \log y_n + (1-t_n) \log (1-y_n)\right]$$

the log-likelihood equivalent of binary cross-entropy corresponds to a Bernoulli-distributed random variable with probability success as prediction , so we write :

$$t_n \sim Bernoulli(y_n)$$

$$L = \prod^N_{n=1} {y_n}^{t_n} {(1-y_n)}^{(1-t_n)}$$

its then easy to see that our loss function , as in the case of the squared error , is the -ve log likelihood 

finally , what about multiclass classification , instead of Bernoulli distribution where we have only 2 choices (example coin toss H or T) , we use categorical distribution (ex: die roll , result : 1-6)

how doe we model this ?

assume we role few times and we get something like {1,1,6,3,2,5,4,3} , now we need an indicator matrix t , where $t[n,k] = 1$ if we rolled k on the $n^{th}$ roll , otherwise 0

since for each role we can only get 1 value , this becomes similar to a one-hot encode matrix

in this case there are no 'model predictions' , we just want to get propability of rolling 1-6 , namely $w_1,w_2,...,w_6$

$$L = \prod^N_{n=1} \prod^6_{k=1} w_k^{\  t_{nk}}$$

its good to remember that the likelihood is the product of PDFs or PMFs for each of the N datapoints

now it is easy to see how we can apply this to a neural network 

$$L = \prod^N_{n=1} \prod^K_{k=1} y_{nk}^{\ \ \ \ \ t_{nk}}$$


and so our output prediction $y_{nk}$ tells us our predictions for each of the targets $t_{n,k}$ , so if $y_{nk}$ is large its more probable that $t_{nk} = 1$ , or if $t_{nk} = 1$ then we want $y_{nk}$ to be close to 1 in order to maximise our likelihood

this is the same as before replacing 6 by more generic K , and probability of each category is $y_{nk}$ rather than fixed $w_{nk}$

so in other words $t_{nk}$ follows the categorical distribution given probabilities $y_{nk} $

REMEMBER what we actually want is the weights , we dont care about y , since we already have t , what we care about are the weights that can cause y to be as closest as possible to t

as usual our los is the -ve log likelihood :

$$Categorical \ Cross \ Entropy \ Loss = - \sum^N_{n=1} \sum^K_{k=1} t_{nk} \log y_{nk}$$

now does this work ? 

if y(n,k) is more wring we want loss to be larger , and if it is more right , we want loss to be smaller

lets assume we have one sample only so no 'n' index: 

Exactly right : -1*log(1) = 0

50% probability on correct target: -1*log(0.5) = 0.693

25% probability on correct target: -1*log(0.25) = 1.386

0% probability on correct target: -1*log(0) = $\infty$

it works !

we finished the first step in training which is defining the loss

next we want to minimise the loss wrt to the weights , this involves gradient descent which needs us to find gradients

what we will need to apply now is the 'Law of total derivatives'

but what is that ?

suppose we are tracking the position of some partical in 3-D space at regular time steps t

we have position = x(t) , y(t) , z(t)

suppose we have some function of position f(x,y,z) (eg : potential energey) , and we want to find the derivative of f wrt t ?

we know that f is a function of x,y,z which are all in turn function of time t so we know that the chain rule gets applied somehow

but the answer is somehow non-intuitive :

$$\frac{\partial f}{\partial t} = \frac{\partial f}{\partial x}\frac{\partial x}{\partial t} + \frac{\partial f}{\partial y}\frac{\partial y}{\partial t} + \frac{\partial f}{\partial z}\frac{\partial z}{\partial t}$$

we can use the same rule for any number of variables parameterised by t :

$$x_k(t) \ for \ k = 1...K$$

$$\frac{\partial f}{\partial t} = \sum^K_{k=1} \frac{\partial f}{\partial x_k} \frac{\partial x_k}{\partial t}$$

How does this apply to neural networks :

We have a function J

J is a function of a(n,k) for all n's and k's

a(n,k) is parameterised by W(d,m)

then :

$$\alpha_{nm} = W_{:,m}^{ \ \ \ T} x_n + bm$$

$$\frac{\partial \alpha_{nm}}{\partial W_{dm}} = x_{nd}$$

lastly : 

$$z_{nm} = \sigma(\alpha_{nm})$$

this differs depending in activation function $\sigma$

for sigmoid : 

$$\sigma(x) = \frac{1}{1+\exp(-x)}$$

$$\frac{\partial z_{nm}}{\partial \alpha_{nm}} = z_{nm}(1-z_{nm})$$

for tanh : 

$$\sigma(x) = \tanh(x)$$

$$\frac{\partial z_{nm}}{\partial \alpha_{nm}} = 1 - z_{nm}^2$$

for relu: 

$$\sigma(x) = relu(x)$$

$$\frac{\partial z_{nm}}{\partial \alpha_{nm}} = u(z_{nm})$$ 

where u() is the step function

Now lets plu gin all derivatives to arrive at the final answer :

$$\frac{\partial J}{\partial W_{dm}} = \sum^{K}_{k=1} \sum^N_{n=1} \left( t_{nk} - y_{nk} \right) V_{mk} z_{nm}(1-z_{nm})x_{nd}$$

$$\frac{\partial J}{\partial b_{m}} = \sum^{K}_{k=1} \sum^N_{n=1} \left( t_{nk} - y_{nk} \right) V_{mk} z_{nm}(1-z_{nm})$$

for the bias term $x_{nd}$ is just replaced by 1

Now lets try to vectorise the gradient for our implementation

first  $z_{nm}(1-z_{nm})$ has only element-wuse multiplications , for convenience just call it z' for short

$$Z' = Z \odot (1-Z) $$

it helps to consider the shape of each object we multiply :

<img src = 'extras/18.5.PNG'><img>

None of the inner shapes match !!! , so this is not a straightforward matrix multiplication
 
ok so lets see what we can do 

first let transpose V , so we get

$${(T-Y)}_{NxK} V^T_{KxM} \rightarrow NxM$$

(sice we sum over k , its logical to have k as the inner dimension)

well this matches the shape of Z' which is also NxM , so we do another element wise multiply

$${\left[ (T-Y)V^T\right]}_{NxM} \odot Z'_{NxM} $$

lastly x is NxD and the previous is NxM , we now the final shape is DxM , and we know we need to sum over N which what matrix multiplication does (so if N is the innder dimension we will some over it) so we transpose x and do a matrix multiply with what we found previously

finally we get :

sigmoid :

$$\large \nabla_wJ = X^T \left[ \left((T-Y)V^T \right) \odot Z \odot (1-Z) \right]$$

or more generally for any activation function :

$$\large \nabla_wJ = X^T \left[ \left((T-Y)V^T \right) \odot Z' \right]$$

````python
# now we will try to write a code for training a neural network
# nothing fance for now just one hidden layer
# then we will see how to modify code for arbitary number of layers
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# these are the same from 17-feedforward
Nclass = 500

# 3 gaussian clouds
X1 = np.random.randn(Nclass,2) + np.array([0,-2])
X2 = np.random.randn(Nclass,2) + np.array([2,2])
X3 = np.random.randn(Nclass,2) + np.array([-2,2])

X = np.concatenate((X1,X2,X3))
# labels
Y = np.array([0]*Nclass+[1]*Nclass+[2]*Nclass)

plt.scatter(X[:,0],X[:,1],c=Y,s=100,alpha=0.5)
plt.show()
````

````python
def softmax(output):
    output = np.exp(output)
    output = output / output.sum(axis=1,keepdims=True)
    return output

def sigmoid(a):
    return 1/(1+np.exp(-a))
````

````python
D = 2
M = 3
K = 3

# we want targets to be NxK matrix , so we one hot encode the targets
# we need to one hot encode labels
def one_hot_encode(Y):
    N = len(Y)
    D = Y.max()+1
    encoded = np.zeros((N,D))
    for i in range(N):
        encoded[i,Y[i]] = 1
    return encoded

T = one_hot_encode(Y)

# next we define weights and bias terms
W = np.random.randn(D,M)
b = np.random.randn(M)

V = np.random.randn(M,K)
c = np.random.randn(K)


def forward(X,W,b,act_fn):
    out = X@W+b
    act = act_fn(out)
    return act

    
epochs = 1000

lr = 0.01 

for i in range(epochs):
    # feedforward
    Z = forward(X,W,b,sigmoid)
    y = forward(Z,V,c,softmax)
    
    #calculate gradients
    delta = T-y
    grad_V = Z.T@(delta)
    grad_c = delta.sum(axis=0)
    
    Z_prime = Z*(1-Z)
    grad_b = ((delta@V.T)*Z_prime)
    grad_W = X.T@grad_b
    grad_b = grad_b.sum(axis=0)
        
    # now we perform gradient ascent
    b += lr*grad_b
    c += lr*grad_c
    W += lr*grad_W
    V += lr*grad_V
````

````python
# now lets get predictions and caluclate accuracy
Z = forward(X,W,b,sigmoid)
y = forward(Z,V,c,softmax)
preds = np.argmax(y,axis=1)
print("accuracy: ",np.mean(preds==Y))
````

````python
# lets re plot using predictions
plt.scatter(X[:,0],X[:,1],c=Y,s=100,alpha=0.5)
plt.title("Labels")
plt.show()

plt.title("predictions")
plt.scatter(X[:,0],X[:,1],c=preds,s=100,alpha=0.5)
plt.show()
````

we can think x = $z^{(0)}$ and y = $z^{(3)}$

also lets consider the sizes :

the general pattern is $size(W^{(t)}) = M^{(t-1)} x M^{(t)}$

$$size(W^{(3)}) = M^{(2)} \times K , size(b^{(3)}) = K$$

Also think of $D = M^{(0)} , K = M^{(3)}$

As usual we now want to find the gradient of our objective , since these are composite functions , it makes sense to work backwards hence the name backpropagation 

so first we find gradient wrt $W^{(3)}$ and $b^{(3)}$ then wrt $W^{(2)}$ and $b^{(2)}$ which will require more term due to chain rule s then wrt $W^{(1)}$ and $b^{(1)}$ which will require even more terms

First gradients in last layer

$$\frac{\partial J}{\partial W^{(3)}_{\ \ \ m^{(2)}k}} = \sum^N_{n=1} \sum^K_{k'=1} \frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{W^{(3)}_{m^{(2)}k}}}$$

$$\frac{\partial J}{\partial b^{(3)}_{\ k}} = \sum^N_{n=1} \sum^K_{k'=1} \frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{b^{(3)}_{\ k}}}$$

These are the same as before only different symbols :

$$\frac{\partial J}{\partial W^{(3)}_{\ \ \ m^{(2)}k}} = \sum^N_{n=1} \left( t_{nk} - y_{nk}\right) z^{(2)}_{\ nm^{(2)}}$$

$$\frac{\partial J}{\partial b^{(3)}_{\ k}} = \sum^N_{n=1} \left( t_{nk} - y_{nk}\right) $$

Next , Gradients in the second layer :

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ m^{(1)}m^{(2)}}} = \sum^{K}_{k=1}\sum^N_{n=1} \sum^K_{k'=1} \frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{z^{(2)}_{nm^{(2)}}}}\frac{\partial{z^{(2)}_{nm^{(2)}}}}{\partial{a^{(2)}_{nm^{(2)}}}}\frac{\partial{a^{(2)}_{nm^{(2)}}}}{\partial{W^{(2)}_{\ \ m^{(1)}m^{(2)}}}}$$

$$\frac{\partial J}{\partial b^{(2)}_{\ \ \ m^{(2)}}} = \sum^{K}_{k=1}\sum^N_{n=1} \sum^K_{k'=1} \frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{z^{(2)}_{nm^{(2)}}}}\frac{\partial{z^{(2)}_{nm^{(2)}}}}{\partial{a^{(2)}_{nm^{(2)}}}}\frac{\partial{a^{(2)}_{nm^{(2)}}}}{\partial{b^{(2)}_{\ \ m^{(2)}}}}$$

again : we already derived these

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ m^{(1)} m^{(2)}}} = \sum^K_{n=1} \sum^N_{n=1}  \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)}_{\ nm^{(2)}} \left( 1 - z^{(2)}_{nm^{(2)}} \right) z^{(1)}_{\ nm^{(1)}}$$

$$\frac{\partial J}{\partial b^{(2)}_{\ \   m^{(2)}}} = \sum^K_{n=1} \sum^N_{n=1}  \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)}_{\ nm^{(2)}} \left( 1 - z^{(2)}_{nm^{(2)}} \right) $$

finally we derive gradients for the first layer , we need to remember the law of total derivatives, that is , in short , "Sum over any index that doesnot appear on LHS"

$$\frac{\partial J}{\partial W^{(1)}_{\ \ \ dm^{(1)}}} = \sum^{M^{(2)}}_{m^{(2)}=1} \sum^{K}_{k=1}\sum^N_{n=1} \sum^K_{k'=1} \frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{z^{(2)}_{nm^{(2)}}}}\frac{\partial{z^{(2)}_{nm^{(2)}}}}{\partial{a^{(2)}_{nm^{(2)}}}}\frac{\partial{a^{(2)}_{nm^{(2)}}}}{\partial{z^{(1)}_{\ \ nm^{(1)}}}}\frac{\partial{z^{(1)}_{\ \ nm^{(1)}}}}{\partial{a^{(1)}_{\ \ nm^{(1)}}}}\frac{\partial{a^{(1)}_{\ \ nm^{(1)}}}}{\partial{W^{(1)}_{\ \ dm^{(1)}}}}$$

$$\frac{\partial J}{\partial b^{(1)}_{\ \  m^{(1)}}} = \sum^{M^{(2)}}_{m^{(2)}=1} \sum^{K}_{k=1}\sum^N_{n=1} \sum^K_{k'=1} \frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{z^{(2)}_{nm^{(2)}}}}\frac{\partial{z^{(2)}_{nm^{(2)}}}}{\partial{a^{(2)}_{nm^{(2)}}}}\frac{\partial{a^{(2)}_{nm^{(2)}}}}{\partial{z^{(1)}_{\ \ nm^{(1)}}}}\frac{\partial{z^{(1)}_{\ \ nm^{(1)}}}}{\partial{a^{(1)}_{\ \ nm^{(1)}}}}\frac{\partial{a^{(1)}_{\ \ nm^{(1)}}}}{\partial{b^{(1)}_{\ \ m^{(1)}}}}$$

we have already solved all these derivatives before , all follow the same form but differnet subscripts , for instance , $W^{(1)T}x$ and $W^{(2)T}z^{(1)}$ have same form , for compactness , we replace the derivative of activation function by z' , so for ex : $z^{(2)'}_{\ nm^{(2)}} = z^{(2)}_{\ nm^{(2)}} \left( 1 - z^{(2)}_{nm^{(2)}} \right)$ in case of sigmoid

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ d m^{(1)}}} = \sum^K_{n=1} \sum^N_{n=1}  \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} W^{(2)}_{\ \ m^{(1)}m^{(2)}} z^{(1)'}_{\ nm^{(1)}} x_{nd}$$

$$\frac{\partial J}{\partial b^{(2)}_{\ \ \ m^{(1)}}} = \sum^K_{n=1} \sum^N_{n=1}  \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} W^{(2)}_{\ \ m^{(1)}m^{(2)}} z^{(1)'}_{\ nm^{(1)}} $$

now lets consider the pattern we have seen so far

first layer had 7 derivatives

second layer had 5 derivatives

Output layer had 3 derivatives

$$a^{(l)} = W^{(l)T}z^{(l-1)} + b^{(l)}$$

$$z^{(l)} = \sigma \left( a^{(l)} \right)$$


what we want to notiice that , the further we go back , the more commom terms we encounter , this means that the calculation is recursive and we can use the results of previous calculations without recalculating them

$$\frac{\partial J}{\partial W^{(3)}_{\ \ \ m^{(2)}k}} = \sum^N_{n=1} \sum^K_{k'=1} \color{green}{\boxed{ \color{black}{\frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}}}}\frac{\partial{a^{(3)}_{nk}}}{\partial{W^{(3)}_{m^{(2)}k}}}$$


$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ m^{(1)}m^{(2)}}} = \sum^{K}_{k=1}\sum^N_{n=1} \sum^K_{k'=1} \color{red}{\boxed{\color{green}{\boxed{ \color{black}{\frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}}}}\color{black}{\frac{\partial{a^{(3)}_{nk}}}{\partial{z^{(2)}_{nm^{(2)}}}}\frac{\partial{z^{(2)}_{nm^{(2)}}}}{\partial{a^{(2)}_{nm^{(2)}}}}}}}\frac{\partial{a^{(2)}_{nm^{(2)}}}}{\partial{W^{(2)}_{\ \ m^{(1)}m^{(2)}}}}$$


$$\frac{\partial J}{\partial W^{(1)}_{\ \ \ dm^{(1)}}} = \sum^{M^{(2)}}_{m^{(2)}=1} \sum^{K}_{k=1}\sum^N_{n=1} \sum^K_{k'=1} \color{red}{\boxed{\color{green}{\boxed{ \color{black}{\frac{\partial{J_{nk'}}}{\partial{y_{nk'}}}\frac{\partial{y_{nk'}}}{\partial{a^{(3)}_{nk}}}}}}\color{black}{\frac{\partial{a^{(3)}_{nk}}}{\partial{z^{(2)}_{nm^{(2)}}}}\frac{\partial{z^{(2)}_{nm^{(2)}}}}{\partial{a^{(2)}_{nm^{(2)}}}}}}}\frac{\partial{a^{(2)}_{nm^{(2)}}}}{\partial{z^{(1)}_{\ \ nm^{(1)}}}}\frac{\partial{z^{(1)}_{\ \ nm^{(1)}}}}{\partial{a^{(1)}_{\ \ nm^{(1)}}}}\frac{\partial{a^{(1)}_{\ \ nm^{(1)}}}}{\partial{W^{(1)}_{\ \ dm^{(1)}}}}$$

As we can see the first part of the gradeint for $W^{(3)}$ appears in the gradient for $W^{2}$ , an eecen bigger part of gradient of $W^{(2)}$ appears in gradient of $W^{(3)}$

Now lets look at the vectorised updates for layers 1 and 2

$$\nabla_{W^{(output)}}J = Z^T(T-Y)$$

$$\nabla_{W^{(input)}}J = X^T\left[ \left((T-Y)V^{T}\right) \odot Z' \right] $$

A general trend can be noticed , for a general w at layer l  , its gradient is the input matrix to that layer multiplied by some other term call it delta

$$\nabla_{W^{(l)}}J = Z^{(l-1)^T}\delta^{(l)}$$

recognising this pattern lets reason about what the shape of everything has to be

we know that 

$$shape(W^{(l)}) = M^{(l-1)} \times M^{(l)} $$

we also know that

$$shape(Z^{(l-1)}) = N \times M^{(l-1)} $$

so to satisfy ryles of matrix multiplication :

$$shape(\delta^{(l)}) = N \times M^{(l)} $$

now lets look back at gradients of Ws and try to 'guess' what $\delta$ might be

$$\frac{\partial J}{\partial W^{(3)}_{\ \ \ m^{(2)}k}} = \sum^N_{n=1} \left( t_{nk} - y_{nk}\right) z^{(2)}_{\ nm^{(2)}}$$

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ m^{(1)} m^{(2)}}} = \sum^K_{n=1} \sum^N_{n=1}  \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} z^{(1)}_{\ nm^{(1)}}$$

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ d m^{(1)}}} = \sum^K_{n=1} \sum^N_{n=1}  \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} W^{(2)}_{\ \ m^{(1)}m^{(2)}} z^{(1)'}_{\ nm^{(1)}} x_{nd}$$

In each of these equations , the last term is the input to that layer , so we can identify $\delta$ a t each layer by taking all the terms up to (but not including) last term

$$\frac{\partial J}{\partial W^{(3)}_{\ \ \ m^{(2)}k}} = \sum^N_{n=1} \color{green}{\boxed{ \color{black}{\left( t_{nk} - y_{nk}\right) }}} z^{(2)}_{\ nm^{(2)}}$$

$$\delta^{(3)}_{nk} = t_{nk} - y_{nk}$$

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ m^{(1)} m^{(2)}}} = \sum^K_{n=1} \sum^N_{n=1} \color{green}{\boxed{ \color{black}{ \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} }}}z^{(1)}_{\ nm^{(1)}}$$

$$\delta^{(2)}_{nm^{(2)}} = \sum^{K}_{k=1} \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}}$$

notice that $z^{(1)}$ does not depend on k , so we can include the summation over k in expression for $\delta^{(2)}$

$$\frac{\partial J}{\partial W^{(2)}_{\ \ \ d m^{(1)}}} = \sum^K_{n=1} \sum^N_{n=1}  \color{green}{\boxed{\color{black}{\left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} W^{(2)}_{\ \ m^{(1)}m^{(2)}} z^{(1)'}_{\ nm^{(1)}}}}} x_{nd}$$

$$\delta^{(1)}_{nm^{(1)}} = \sum^{M^{(2)}}_{m^{(2)}=1} \sum^{K}_{k=1} \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} W^{(2)}_{\ \ m^{(1)}m^{(2)}} z^{(1)'}_{\ nm^{(1)}} $$

since $x$ is neither indexed by $M^{(2)}$ nor K , so again we can include the summations over K and $M^{(2)}$ 

Another way of thinking about this is that we follow our rule of summing up over all indicies that dont show up outisde summation

now if we consider $\delta$s ,we can clearly see the pattern:

$$\delta^{(3)}_{nk} = t_{nk} - y_{nk}$$

$$\delta^{(2)}_{nm^{(2)}} = \sum^{K}_{k=1} \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}}$$


$$\delta^{(1)}_{nm^{(1)}} = \sum^{M^{(2)}}_{m^{(2)}=1} \sum^{K}_{k=1} \left( t_{nk} - y_{nk}\right) W^{(3)}_{\ \ m^{(2)}k} z^{(2)'}_{\ nm^{(2)}} W^{(2)}_{\ \ m^{(1)}m^{(2)}} z^{(1)'}_{\ nm^{(1)}} $$

each $\delta^{(l)}$ contains the expression for $\delta^{(l+1)}$ , using this idea we can express each delta in terms of the delta we found priveously

of course the only $\delta$ we can npt do this with is  $\delta^{(3)}$ , which is at the head of the neural network and depends on no others , if we think of this in terms of recursion , then this is the base case

$$\delta^{(3)}_{nk} = t_{nk} = y_{nk}$$

$$\delta^{(2)}_{nm^{(2)}} = \sum^K_{k=1} \delta^{(3)}_{nk} W^{(3)}_{m^{(2)}k} z^{(2)'}_{nm^{(2)}}$$

$$\delta^{(1)}_{nm^{(1)}} = \sum^{M^{(2)}}_{m=1} \delta^{(2)}_{nm^{(2)}} W^{(2)}_{m^{(1)}m^{(2)}} z^{(1)'}_{nm^{(1)}}$$

so every delta (except for the very last one) follows the recursive pattern :

$$\delta^{L}_{nk} = t_{nk} - y_{nk}$$

$$\delta^{l}_{nm^{(l)}} = \sum^{M^{(l+1)}}_{m^{(l+1)}=1} \delta^{(l+1)}_{nm^{(l+1)}} W^{(l+1)}_{m^{(l)}m^{(l+1)}} z^{(l)'}_{nm^{(l)}} , \ for \ l = 1 ... L-1$$


The vectorised for for $\delta$ is:

$$\delta^{(L)} = T-Y$$

$$\delta^{(l)} = \left( \delta^{(l+1)} W^{(l+1)T} \right) \odot Z^{(l)'} , \ for \  l =1 ... L-1$$

Finally the generic weight updates are :

$$\nabla_{W^{(l)}}J = Z^{(l-1)T} \delta^{(l)}$$

$$\nabla_{b^{(l)}}J = sum \left( \delta^{(l)} , axis = 0 \right) = \delta^{(l)T}1_N$$

````python
# now lets split data into train and test
X,Y = shuffle(X,Y)
Ntrain = 400
Xtrain,Ytrain = X[:Ntrain],Y[:Ntrain]
Xtest,Ytest = X[Ntrain:],Y[Ntrain:]
````

````python
# we need to one hot encode labels
def one_hot_encode(Y):
    N = len(Y)
    D = Y.max()+1
    encoded = np.zeros((N,D))
    for i in range(N):
        encoded[i,Y[i]] = 1
    return encoded

Ttrain = one_hot_encode(Ytrain)
Ttest = one_hot_encode(Ytest)
````

````python
# now for the neural network , one hidden layer

def sigmoid(a):
    return 1/(1+np.exp(-a))


# next we define weights and bias terms
M = 5
W = np.random.randn(D,M)
b = np.random.randn(M)

V = np.random.randn(M,K)
c = np.random.randn(K)


def forward(X,W,b,act_fn):
    out = X@W+b
    act = act_fn(out)
    return act

    
epochs = 10000

lr = 0.001

for i in range(epochs):
    # feedforward
    Z = forward(Xtrain,W,b,sigmoid)
    y = forward(Z,V,c,softmax)
    
    #calculate gradients
    delta = Ttrain-y
    grad_V = Z.T@(delta)
    grad_c = delta.sum(axis=0)
    # we used tanh
    Z_prime = Z*(1-Z)
    grad_b = ((delta@V.T)*Z_prime)
    grad_W = Xtrain.T@grad_b
    grad_b = grad_b.sum(axis=0)
        
    # now we perform gradient ascent
    b += lr*grad_b
    c += lr*grad_c
    W += lr*grad_W
    V += lr*grad_V
````

````python
# now lets get predictions and caluclate accuracy
Z = forward(Xtest,W,b,sigmoid)
y = forward(Z,V,c,softmax)
preds = np.argmax(y,axis=1)
print("accuracy: ",np.mean(preds==Ytest))
````

<h1>Math</h1>

we end this with some few notes on training

suppose that instead of doing multi-class classification we do binary classification

now we have one output node and sigmoid instead of softmax , now what will the gradient look like (assuming 1-hidden-layer) ?

Assume Input to hidden weights are called W , and hidden to input weights are called v


First we need to rememebr that we are using the binary cross-entropy function

$$J = - \sum^{N}_{n=1} \log(y_n) + (1-t_n) \log (1-y_n)$$

next lets write the equation of the prediction :

$$y = \sigma(v^T tanh(W^Tx)) = \sigma(v^Tz) , z = tanh(W^Tx)$$

notice here howa $v$ is a vector not a matrix , and that we use tanh as our activation function for the hidden layer (of course we could have used any other)

now we want to calculate the derivatives :

$$\frac{\partial J}{\partial W_{ij}} , \frac{\partial J}{\partial v_j}$$

first we calculate hidden to output weights derivatives :

we use chain rule , all the derivatives have been derived before 

chain rule:

$$\frac{\partial J}{\partial v_j} = \sum^N_{n=1}\color{green}{\boxed{\frac{\partial J_n}{\partial y_n}}} \color{red}{\boxed{ \frac{\partial y_n}{\partial a_n }}}\color{blue}{\boxed{ \frac{\partial a_n}{\partial v_{j}}}}$$

derivatives :

$$\frac{\partial J}{\partial v_j} =  \sum^N_{n=1} \color{green}{\boxed{-\frac{t_n}{t_n} \frac{1-t_n}{1-y_n} }} \color{red}{\boxed{y_n (1-y_n)}} \color{blue}{ \boxed{z_{j,n}}}$$

we can then simplify this to get

$$\frac{\partial J}{\partial v_j} = \sum^N_{n=1} \left[ y_n - t_n \right] z_{j,n}$$

Next we we want to get the input to hidden weights derivatives :


chain rule:

$$\frac{\partial J}{\partial W_{ij}} = \sum^N_{n=1}\color{green}{\boxed{\frac{\partial J_n}{\partial y_n}}} \color{red}{\boxed{ \frac{\partial y_n}{\partial a_n }}}\color{blue}{\boxed{ \frac{\partial a_n}{\partial z_n}}}\color{purple}{\boxed{ \frac{\partial z_n}{\partial \alpha_n}}}\color{gray}{\boxed{ \frac{\partial \alpha_n}{\partial W_{ij}}}}$$

derivatives :

$$\frac{\partial J}{\partial W_{ij}} =  \sum^N_{n=1} \color{green}{\boxed{-\frac{t_n}{t_n} \frac{1-t_n}{1-y_n} }} \color{red}{\boxed{y_n (1-y_n)}} \color{blue}{ \boxed{v_j}} \color{purple}{\boxed{1 (1-z_{j,n}^2)}} \color{gray}{\boxed{x_{i,n}}}$$

we can then simplify this to get

$$\frac{\partial J}{\partial W_{ij}} = \sum^N_{n=1} (y_n - t_n) v_j (1 - z_{j,n}^2) x_{i,n}$$

Next , what if we want to do regression instead of classification ?

We keep all the nonlinear layers , but the head becomes linear regression instead of logisitc regression

and so we use the squared error

$$J = \sum_n {(t_n - y_n)}^2$$

$$z = tanh(W^Tx)$$

$$y = v^Tz = v^ttanh(W^Tx)$$


then for our derivatives :

$$\frac{\partial J}{\partial v_j} = \sum_n 2(t_n - y_n)(-1)z_{j,n} = \sum_n 2(y_n -t_n) z_{j,n}$$

Here is the same but we dont have to deal with the sigmoid , notice that if we drop the constant 2 , which can be absorbed into the learning rate , we get the same update as that of logistic regression

next , the input to hidden weights

$$\frac{\partial J}{\partial W_{ij}} = \sum_n 2(y_n - t_n)v_j (1-z_{j,n}^2)x_{i,n}$$

again we get the same update !!!

This doesnot mean both are the same , rememebr that ys and ts take on different values and represent different things in each case

````python
# we want to see how ANNs perform on XOR and Donut problems
# we will modify the equations since now we are doing binary classification
def sigmoid(a):
    return 1/(1+np.exp(-a))

def forward(X,W,b,act_fn):
    out = X@W+b

    act = act_fn(out)
    return act


def train(X,T,lr=0.01,epochs=1000,M=5):
    N,D = X.shape
    # we need to one hot encode labels

    # next we define weights and bias terms
    W = np.random.randn(D,M)
    b = np.random.randn(M)
    V = np.random.randn(M)
    

    for i in range(epochs):
        # feedforward
        Z = forward(X,W,b,sigmoid)
        Y = forward(Z,V,np.zeros((N)),sigmoid)
        #calculate gradients
        delta = T-Y
        grad_V = Z.T@(delta)
        Z_prime = Z*(1-Z)
        grad_b = ((np.atleast_2d(delta).T@np.atleast_2d(V))*Z_prime)
        grad_W = X.T@grad_b
        grad_b = grad_b.sum(axis=0)

        # now we perform gradient ascent
        b += lr*grad_b
        W += lr*grad_W
        V += lr*grad_V
    # now lets get predictions and caluclate accuracy
    Z = forward(X,W,b,sigmoid)
    Y = forward(Z,V,np.zeros((N)),sigmoid)
    preds = np.round(Y)
    print("final classification rate: ",np.mean(preds==T))
    
````

````python
def get_xor():
    # we will make the XOR grid like before
    N = 200
    X = np.random.random((N, 2))*2 - 1
    # generate labels
    T = np.zeros(N)
    T[(X[:,0] < 0) & (X[:,1] > 0)] = 1
    T[(X[:,0] > 0) & (X[:,1] < 0)] = 1
    return X,T
    
    
# now again for the donut problem 
def get_donut():
    N = 200
    D = 2
    R_inner = 5
    R_outer = 10

    # in a polar system , this is the distance from origin
    # we sample from a uniform distribution , set mean to 0
    R1 = np.random.randn((N//2)) + R_inner
    # and this is the angle
    # 2pi * (number from 0-1) so we get thetas between 0-2*pi
    theta = 2*np.pi*np.random.random(N//2)
    # in a cartesian system x component is R cos theta , y component is R sin theta
    X_inner = np.concatenate([[R1*np.cos(theta)],[R1*np.sin(theta)]]).T

    R2 = np.random.randn((N//2)) + R_outer
    theta = 2*np.pi*np.random.random(N//2)
    X_outer = np.concatenate([[R2*np.cos(theta)],[R2*np.sin(theta)]]).T

    X = np.concatenate([X_inner , X_outer])
    T = np.array([0]*(N//2) + [1]*(N//2))
    
    return X,T
````

````python
X,Y = get_donut()
````

````python
train(X,Y,lr=0.00005,epochs=100000,M=8)
````

````python
X,Y = get_xor()
````

````python
train(X,Y,lr=0.00005,epochs=100000,M=8)
````

<h1>Math</h1>

now we want to discuss some practical considerations for Activation functions (we will also include derivatives just as a recap)

first Sigmoid :

$$\sigma(x) = \frac{1}{1+e^{-x}}$$

$$\sigma'(x) = \sigma(x)(1-\sigma(x))$$

<img src = 'extras/17.3.PNG' width = 300></img>

problem : most of it is flat ! (slope~=0)

if gradient is 0  ,then nothing will be learned (or learn very slowly if near 0)

Another problem : we like mean = 0 , var = 1 , if the data is in a small range , this helps us with the previous problem , where if the inputs get too large , derivative goes to 0

but what happens when input to sigmoid = 0 , $\sigma(0) = 0.5$

so the sigmoid keeps offsetting the data , even if the input is standarised to have 0 mean , after it goes through sigmoid , it is going to be at 0.5


next tanh

$$y = tahn(x) = \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$$

$$\frac{dy}{dx} = 1 - y^2$$


<img src = 'extras/17.4.PNG' width = 300></img>

solves the problem we just talked about , $tanh(0) = 0$ , however the ends of the tanh are still flat , so slope still vanished at edges

tanh is typically better than sigmoid , still depends on use (though in some scenarios we require sigmoid )

ReLU

$$ y = ReLU(x) = max(0,x)$$

$$\frac{dy}{dx} = 1 \ if \ x > 0 \ , \  0 \ if \  x< 0 $$

<img src = 'extras/17.5.PNG' width = 300></img>

gradient at 0 is not even defined , but we can take it as 0

Trains a lot more easily compared to tanh/sigmoid and faster

half of it is still flat (slope=0) , but still there is a non-zero half

this problem is called 'dead neurons' , activations drop down to 0 , and since gradient is 0 , weights behind dont get updated again

one solution is just add more hidden units , and hope not too many of them go down to 0

another solution is the Leaky ReLU

<img src='extras/18.7.PNG' width = 300></img>

small slope on left half , solves the dead neuron problem , but interestingly can make neural networks perform worse

doesnot necessarily mean LReLUs are worse ,depends on data

finally we want to talk about practical methods for choosing learning rate and regularisation penalty

firsr learning rate , we know this is a smalle number , if we try 0.1 , we dont need to try 0.09,0.08 etc..

Better to go down on log scale / factors of 10

$ 10^{-1} , 10^{-2} , 10^{-3} , etc ... even \ 10^{-7} $

if learning rate is too high , cost goes to infinity or  NaN (Not a Number) 

unfortunately , neural network will continue to train normal , multiplying NaNs as if they were actual numbers , so stop code manually

if convergance is slow , learning rate is too low , so try increasing by factors of 10

If a problem persists , turn everything off , just use normal gradient descent (no regularisation or anything else)

learning rate that is too high will mess up things , but one that is too low wont

we will discuss later how to train a neural network in batches of training data

when training data varies in size , lr gets sensitive to it , since the cost function is sum of all individual errors

to normalise this we divide by cost N

$$J = \frac{1}{N}\sum_n \sum_k t_{nk} log(y_{nk})$$

with regularisation (L2):

$$J = \frac{1}{N}\sum_n \sum_k t_{nk} log(y_{nk}) + \lambda \sum_l \Vert W^{(l)} \Vert^2$$

To make regularisation penalty independant of number of parameters , divide the sum of squares of all parameters by number of parameters 

so with regularisation our weight update would be :

$$w_{t+1} = w_t - \eta \left( \nabla_wJ - \lambda w_t\right)$$
