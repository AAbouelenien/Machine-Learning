<h1>code</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The function we will try to fit to is y = sin(x)
N = 100
X = np.linspace(0,6*np.pi,N)
Y = np.sin(X)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
def make_poly(X,deg):
    # ex : if deg = 3 , we end up with data having columns : 1 X X^2 X^3
    N = len(X)
    data = np.ones(N)
    for d in range(deg): # add deg from 1 to deg
        data = np.vstack((data,X**(d+1)))
    return np.atleast_2d(data).T
```

```python
def fit(X,Y):
    return np.linalg.solve(X.T @ X , X.T @ Y)
```

```python
def fit_and_display(X,Y,sample,deg):
    N = len(X)
    
    # here we split data
    train_idx = np.random.choice(N,sample) # sample tells us how many samples to take for train , ex : 80
    X_train = X[train_idx]
    Y_train = Y[train_idx]
    
    # lets plot the train data alone
    plt.scatter(X_train,Y_train)
    plt.show()
    
    # here we fit to the train polynomial data
    X_train_poly = make_poly(X_train,deg)
    w = fit(X_train_poly,Y_train)
    
    # now lets display how we fit to the entire data
    X_poly = make_poly(X,deg)
    Y_hat = X_poly @ w
    
    plt.plot(X,Y , label = 'Original function') # this is the real graph
    plt.plot(X,Y_hat , label = 'Approximated/Predicted function') # this is our prediction
    plt.scatter(X_train,Y_train,label = 'Train points') # these are our train points
    plt.title('deg = %d' % deg)
    plt.legend()
    plt.show()
    
    
    
```

```python
for deg in [5,6,7,8,9]:
    fit_and_display(X,Y,10,deg)
    
# some very high values cause curve to get squashed and thus appearing as a line
# but it still is actually a sine wave
# As wee increase the degree , our curve passes through more and more train points
# but it doesnt necessarily fit to the overall shape of the function
# that is overfitting to training data , while miising the general trend of the data
```

```python
def mse(Y,Y_hat): # calculate mean squared error
    e = Y-Y_hat
    return (e@e) / len(e)
```

```python
def plot_train_vs_test_curves(X,Y,sample=20,max_deg=20):
    # Redo everything
    # but this time we will save mse to plot them
    N = len(X)
    train_idx = np.random.choice(N,sample)
    X_train = X[train_idx]
    Y_train = Y[train_idx]
    
    test_idx = [idx for idx in range(N) if idx not in train_idx]
    X_test = X[test_idx]
    Y_test = Y[test_idx]
    
    mse_trains = []
    mse_tests = []

    for deg in range(max_deg+1):
        X_train_poly = make_poly(X_train,deg)
        w = fit(X_train_poly,Y_train)
        Y_hat_train = X_train_poly @ w
        mse_train = mse(Y_train,Y_hat_train)
        
        X_test_poly = make_poly(X_test,deg)
        Y_hat_test = X_test_poly @ w
        mse_test = mse(Y_test,Y_hat_test)
        
        mse_trains.append(mse_train)
        mse_tests.append(mse_test)
        
    plt.plot(mse_trains,label='train mse')
    plt.plot(mse_tests,label='test mse')
    plt.title('Train vs Test mse')
    plt.legend()
    plt.show()
    
    # plot it alone , test mse is too large so train mse gets squashed
    plt.plot(mse_trains,label = 'train mse')
    plt.legend()
    plt.title('Train mse')
    plt.show()
        
        
```

```python
plot_train_vs_test_curves(X, Y)
```

Now lets take a deeper look ar the squared error cost function

$$\sum ^{N} _ {i=1} (y_i - \hat y_i)^2$$

We will show now that linear regression is the maximum likelihood solution to the line of best fit !!!!?????


Now suppose we are measuring the height of all students in our school , assume the heights are gaussian distributed i.e. if we plot a histogram it would be a bell curve

```python
# ex generate randomly distributed numbers
# them more the samples , the closer we get to the bell curve
heights = np.random.randn(10000000)
```

```python
plt.hist(heights)
plt.show()
```

so we collect heights ${x_1,x_2,...,x_N}$

Intuitevly we now that $$ \mu = \frac {1}{N} \sum ^{N} _{i=1} x_i $$

But there is another way to find $\mu$

Suppose we want to find the true mean of the Gaussian from which the data arises

find $\mu$ where  $X$ ~ $N(\mu,\sigma ^2)$

Since all x is assumed to be from the same gaussian distribution , we can write the probability of any single point $x_i$ using the pdf for the gaussian

$$ p(x_i) =  \frac {1} {\sqrt{2 \pi  \sigma^2}} e^{-\frac{1}{2} \frac{(x_i - \mu)^2}{\sigma ^2}}$$


Since we know all the students heights are independant and identically distributed (IID) , we can write the joint like probability of everyones height as the product of their individual probabilities

$$ p(x_1,x_2,....,x_N) = \prod ^{N} _{i=1}  \frac {1} {\sqrt{2 \pi  \sigma^2}} e^{-\frac{1}{2} \frac{(x_i - \mu)^2}{\sigma ^2}} $$

Now we can write this as "likelihood"

$$ P(X|\mu) = p(x_1,x_2,..,x_N) = \prod^{N}_{i=1} \frac{1}{\sqrt{2\pi\sigma^2}}  e^{-\frac{1}{2} \frac{(x_i-\mu)^2}{\sigma^2}}$$

So the likelihood is the Probability of data given X given the parameter of interest $\mu$ (remember its $\mu$ that we want)

So we want to find the $\mu$ that maximises the likelihood , by doing so we seek the best setting of $\mu$ so that the data we measured is likely to have come from this ditribution

Now comes calcules !

$\frac {\partial p} {\partial \mu} = 0 $

but this is hard to solve , instead we can try solving :

$\frac {\partial \log p} {\partial \mu} = 0 $

this works since the $\log$ is a monotonically increasing function , meaning that for any $A>B$ , we know that $\log A > \log B$ 

lets write the expression for the log-likelihood (that just the log of p)

$$ l = \log \prod ^{N} _{i=1} \frac{1}{\sqrt{2\pi\sigma^2}}  e^{-\frac{1}{2} \frac{(x_i-\mu)^2}{\sigma^2}} $$

since we know that the log of prodcut is a sum of logs , we can write this as :

$$ l = \sum^{N}_{i=1} \biggl[-\frac{1}{2} \log(2\pi\sigma^2) - \frac {1}{2} \frac {(x_i - \mu^2)}{\sigma^2} \biggr ] $$

Now take derivative and set to 0

$$ \frac {d l}{d \mu} = \sum^{N}_{i=1} \frac {x_i - \mu}{\sigma^2} = 0$$

Now solce for $mu$ , amazingly :

$$\mu = \frac{1}{N} \sum^{N}_{i=1} x_i$$


Now lets reconsider l :

$$ l = \sum^{N}_{i=1} \biggl[-\frac{1}{2} \log(2\pi\sigma^2) - \frac {1}{2} \frac {(x_i - \mu^2)}{\sigma^2} \biggr ] $$

We want to maximise $l$ wrt to $\mu$ , so $\mu$ is the variable that affects $l$ 

If we do so , we can ignore the term $-\frac{1}{2} \log(2\pi\sigma^2)$ goes away since it is just a constant with respect to mu

Also the $ \frac{1}{2} $ and the $\sigma$ go away since they are just constant and can be ignored

what WE DO HAVE TO KEEP is the "-" sign , otherwise the problem changes from a maximsation problem to a minimisation problem

Thus we say that an equivalent expression for l is:

$$equivalent \ l = - \sum^{N}_{i=1}(x_i - \mu)^2  $$

of course equivalent l and l are not the same , what we are saying that maximising equivalent l is the same as maximising l , equivalent l is just l without all the unnecessary terms

so if we find the value of $\mu$ that maximises equivalent l , then we found the value of $\mu$ that maximises l

So what we are trying to do is maximise :

$$equivalent \ l = - \sum^{N}_{i=1}(x_i - \mu)^2  $$

But this is actually the same as trying to minimise 

$$ \sum^{N}_{i=1}(x_i - \mu)^2  $$

Which is our cost function :)

$$E = \sum^{N}_{i=1}(y_i - \hat y_i)^2  $$

if we have $x_i$ as $y_i$ and $\mu$ as $\hat y_i$

So when we are minimising the squared error for linear regression , we are maximising the likelihood , this is equivalent to writing :

$y$ ~ $N(W^Tx,\sigma^2)$

Which means that y is Gaussian distributed with mean $W^Tx$ and arbitart variance

This is also the same as :

$y = W^Tx + \varepsilon$ , $\varepsilon$ ~ $N(0,\sigma^2)$ 

with $\epsilon$ being some gaussian distributed noise with 0 mean and arbitary variance  

This show that linear regression makes the assumptions that the errors are Gaussian and that the trend is linear




<h1>Math</h1>

Now we discuss a technique knows as l2 regularisation that helps us with overfitting

The idea is that we dont want to have overly large weights , otherwise we find ouselves fitting to outliers to minimise the squared error and missing the general trend of the data

In order to execute this , we modify our cost function to add penalty on large weights

$$J = \sum^{N}_{n=1}(y_n\ - \hat y_n)^2 + \lambda |w|^2$$

$\lambda$ is a constant 
The squared-norm $|w|^2$ is just 

$|w|^2 = w^Tw = w_1^2 + w_2^2 + .... + w_D^2$ 

Now lets consider this probabilistically :

originally what we are trying to do is maximise the likelihood :

$P(data|w)$ (this is equivalnt to  $P(x_i|\mu)$)

Now that we changed the cost function this is no longer the case

we know that the squared error cost is the same as minimising the negative log-likelihood same as maximising the log-likelihood

$$J = \sum^{N}_{n=1}(y_n\ - \hat y_n)^2 + \lambda |w|^2$$

Now since $J$ is the negative log-likelihood , then $-J$ is the log-likelihood and so $exp(-J)$ gives us an expression of the likelihood

We take exp(-J) so we have an idea of what exactly are we maximising 

$$\exp(-J) = \biggl[ \prod^{N}_{n=1} \exp(-(y_n - w^Tx_n)^2)\biggr] \exp(-\lambda w^Tw)$$

We see now that we get two gaussians , the first one os the one from before , the second gaussian is a new one , it is a gaussian where $w$ is the random variable , mean is 0 , and variance is $\frac {1}{\lambda}$

Now :

We know the first Gaussian as the Likelihood , its original form (before removing unecessary elements and arriving at equivalent l ) is:

$$ P(Y|X,w) =  \prod^{N}_{n=1} \frac{1}{\sqrt(2\pi\sigma^2)} \exp(- \frac{1}{2\sigma^2} (y_n - w^Tx)^2)  $$

Once again , this is the same as when we had the likelihood as  $P(X|\mu)$ but now we have our $X$ as $Y$ and $\mu$ as $w^Tx$

The second gaussian is named the Prior , it represents our prior beleifs about w irrespective of the data since it does not depend on $X$ or $Y$

$$P(w) = \sqrt{\frac{\lambda}{2\pi}} \exp{(- \frac{\lambda} {2} w^Tw)}$$

This is the original term , of course when calculating equivalnet l we end up throwing away constant terms which dont have $w$ and so it slightly changes in the cost function

Well this is like Bayes Rule :

$$ J_{OLD} \ \alpha \ \ - lnP(Y|X,w) $$

$$ J_{NEW} \ \ \alpha  \ \ - lnP(Y|X,w) - lnP(w) $$

Bayes Rule : 

$$P(w|Y,X) = \frac{P(Y|X,w)P(w)}{P(Y|X)}$$

or equivalently :

$$P(w|Y,X) \ \ \alpha \ \ P(Y|X,w)P(w)$$


This method is called MAP : maximum a posteriori - it means that we maximise the posterior $P(w|data)$

Now its time for ..... calculus !!!

we can write our new cost function J as :

$$J = (Y-Xw)^T(Y-Xw) + \lambda W^Tw$$

$$J = Y^TY - 2Y^TXw + W^TX^TXw + \lambda W^TW$$

Now lets take the derivative (refer to matrix cookbook) and set it to 0:

$$ \frac{\partial J}{\partial w} = - 2X^TY + 2X^TXw + 2\lambda w  = 0$$

Now we solve for $w$ :

$$ \frac{\partial J}{\partial w} = (\lambda I + X^TX)w = X^TY$$

finally we arrive at the solution 

$$w = (\lambda I + X^TX)^{-1} X^TY$$

Summary :

One way to reduce complexity of model and avoid overfitting to outliers is L2 regularisation / Ridge Regression

We add the square the magnitude of weights x some constant $\lambda$ to cost

This encourages weights to be small since p(w) , the prior m is Gaussuan centered around 0

This is also called MAP estimate of w , because we are maximising the posterior of w given the data

<h1>code</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
X = np.linspace(0,10,N)
# we set Y as a linear function of x + some random gaussian distributed noise
Y = 0.5*X + np.random.randn(N)

# now lets add some outliers 
Y[-1] += 30
Y[-2] += 30

plt.scatter(X,Y)
plt.show()
```

```python
# we already know there is no bias term since we created the data that way
# so we dont bother adding a column of 1 (wouldnt differ anyway)
ridge = 1000
X = np.atleast_2d(X).T
w_ridge = np.linalg.solve(X.T@X + ridge*np.eye(1) , X.T@Y )
w = np.linalg.solve(X.T@X , X.T@Y )
```

```python
Yhat = X@w
Yhat_ridge = X@w_ridge
plt.scatter(X,Y)
plt.plot(X,Yhat,label='no regularisation')
plt.plot(X,Yhat_ridge,label='L2 Regulaisation')
plt.legend()
plt.show()
```

```python
# clearly the MAP solution fits better
```

Dimension 1 : 1 if happy

Dimension 2 : 1 if angry

not happy & not angry == sad

Now that we understand one-hot encoding we proceed

we saw that if a category has K different values , then there will b K columns/dimensions to represent it

we can alternatively save 1 column of space by letting all 0s to represent 1 category value

Ex color = [red,green,blue]

Red = [0,1] , Green = [1,0] , Blue = [0,0]

This is undesirable since effect of blue gets absorbed into bias term , but then why do it anyway ?

Many resources suggest using K-1 encoding , and for a good reason

Remember the solution forl linear regression requires computing $(X^TX)^{-1}$

well $X^TX$ cannot be inverted if we use one-hot-encoding as it becomes a singular matrix , but why ?

Remember that if one column of the matrix is a linear combination of some other columns then the matrix is invertable

well , we added a column of 1s to X for the bias terms , and if we sum all the one-hot encoded columns we get a column of 1s !!!

this is refered to as  the "dummy variable trap"

So how do we solve this ?

1 - Use K-1 encoding , and get the effect of some value absorbed into the bias

2 - remove the column of 1s , and there is no bias term

3 - use L2 regularisation $X^TX$ is singular but $\lambda I + X^T X $ is not , inverting a sinular matrix is equivalent to divison by 0 , adding $\lambda I$ is equivalent to adding a small number in the denominator

4 - Gradient descent !!!




<h1>Math</h1>

Now we introduce Gradient descent

The idea is that we have a function that we want to minimise

$J(w)$ = cost or error

Iteratively update w , in direction of gradient $\frac{\partial j}{\partial w} in small steps$

can maximise things to , just switch signs

so

$w = random \ \ initial \ \ value$

$ while \ \  not \ \ converged: $
$w = w - \eta \nabla_w J(w)$
    
Lets take an example :

$J = w^2$

we already know that min is @ w = 0 , but suppose we dont ;)

$\frac {dj}{dw} = 2w$ , set initial w = 20 , and learning rate $\eta$ = 0.1

Formula : $w = w - 0.1 * \frac{dj}{dw}$

Iteration 1: $w = 20 - 0.1*40 = 16$

Iteration 2: $w = 16 - 0.1*2*16 = 12.8$

Iteration 3: $w = 12.8 - 0.1*2*12.8 = 10.24$

clearly we converge


<h1>code</h1>

```python
# now lets code it !!!
import numpy as np
w = 20

for i in range(30): # do 30 iteration
    w = w - 0.1*2*w
    print('It %d w = %f  ' %(i+1,w))
```

```python
# we converge but not there yet , so we try with more steps
w = 20

for i in range(100): # do 30 iteration
    w = w - 0.1*2*w
print('min w = ',w)
```

<h1>Math<h1>

Now we will see how we can derive the gradient descent algorithm for linear regression

We know what we want to minimise thats the square error

$$J = (Y-Xw)^T (Y-Xw)$$

We also know the gradient 

$$\frac {\partial J} {\partial w} = -2X^T Y + 2 X^TXw = 2X^T (\hat Y - Y)$$

All we need to do now is , instead of setting this to 0 and solving for w , we take small steps in the direction of gradient till w converges to the solution

Note : since 2 (in the expression) is a constant we can just drop it , since it can be absorbed into the learning rate

so here is how our code goes:

$w \ = \ draw \ sample \ from \ \ N(0,1/D)$

$for \ t=1 ... T:$

$ \ \ \ \  w \ = \ w \ - \ learning Rate* X^T(\hat Y-Y)$

How do we choose the Learning rate ?

Well its a hyper paramater

If too big we wont conversion , but keep bouncing back and forth across the 'valley'

Too small and the gradienct descent will be slow

<h1>code</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10 # number of samples
D = 3 # number od fimensions

X = np.zeros((N,D))

X[:,0] = 1 # set bias term

X[:5,1] = 1
X[5:,2] = 1

Y = np.array([0]*5 + [1]*5)
# Now the first colums of X is a linear combination of the next 2 columns
# since column1 = column2 + column3 
print(X)
```

```python
# you may want to try the regular solution now , it fails stating that X.T @ X is a singular matrix
# w = np.linalg.solve(X.T @ X , X.T @ Y)
```

```python
costs = []
epochs = 1000
lr = 0.01
w = np.random.randn(D) / np.sqrt(D)
for i in range(epochs):
    Yhat = X@w
    delta = Yhat - Y
    cost = delta@delta
    costs.append(cost)
    w -= lr*X.T@(delta)
```

```python
plt.plot(costs)
plt.show()
```

```python
Yhat = X@w 
plt.plot(Y, label='true labels')
plt.plot(Yhat , label = 'predicted labels')
plt.legend()
plt.show()
```

```python
# our predictions are very close to the targets
```

<h1>Math<h1>

In general we prefer a skinny matrix D << N than a fat matrix D > N

That is we want the dimensionality of X to be much smaller than the number of sample N

So what we would like to be able to do is select some small number of important features that produce the trend and remove all the features that are just noise

we call this sparsity since the end result would be that most of the weights would be 0 but only few of the weights will be non-zero

L1 regularisation helps us acheive sparsity

Another name for L1 regularisation is LASSO regression, i t takes a very similar form to that of L2 regularisation (RIDGE regression)

$$J_{RIDGE} = \sum^{N}_{n=1}(y_n - \hat y_n) + \lambda {\Vert w \Vert}_2^2$$

$$J_{LASSO} = \sum^{N}_{n=1}(y_n - \hat y_n) + \lambda {\Vert w \Vert}_1$$

In the L2 regularisation we had a gaussian likelihood and a gaussian prior on w , since both where the exp of a negative square

In L1 we no longer have a gaussain prior instead we have 

$$p(w) = \frac {\lambda} {2} exp(-\lambda \vert w \vert)$$

The distribution that has a negative absolute value in exponent is the Laplace distribution

So with L1 regularisation we are putting a laplace distributed prior on the weights and we are solving for the posterior of w on the last prior , now lets try that

$$J = (Y-Xw)^T(Y-Xw) + \lambda \vert w \vert$$

$$J = Y^TY - 2Y^TXw + w^TX^TXw + \lambda \vert w \vert$$

$$\frac {\partial J} {\partial w} = - 2X^TY + 2X^TXw + \lambda sign(w) = 0$$

$$\frac {\partial J} {\partial w} = X^T(\hat Y - Y) + \lambda sign(w) = 0$$

$sign(x) = 1 \ $if$ \ x \ > 0$ 

$sign(x) = -1 \ $if$ \ x \ < 0$ 

$sign(x) = 0 \ $if$ \ x \ = 0$

Since its inverse is not a function , we cant solve for w

BUT we can still use gradient Descent !!!

<h1>code</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

# we will create a situation where Y depends only on few features from X and the rest are noise
# then we will see if L1 regularisation to see if we can get a sparse set of weights that identify that useful dimensions of x

N= 50
D = 50

# unifromly distributed numbers , centered around 0 , from -5 to 5 
X = (np.random.random((N,D))-0.5)*10
# the true weights are have values for only the first 3 features, the rest are noise
true_w = np.array([1,0.5,-0.5]+[0]*(D-3))
# we get Y then add some gaussian distributed noise
Y = X @ true_w + np.random.randn(N)*0.5
```

```python
costs = []
epochs = 500
lr = 0.001
lasso = 10
w = np.random.randn(D) / np.sqrt(D)
for i in range(epochs):
    Yhat = X@w
    delta = Yhat - Y
    cost = delta@delta
    costs.append(cost)
    w -= lr*( X.T@(delta) + lasso*np.sign(w))
```

```python
plt.plot(costs)
plt.show()
```

```python
plt.plot(true_w,label='true weights')
plt.plot(w,label='predicted weights')
plt.show()
```

<h1>Math</h1>

We want to discuss the differences between L1 and L2 regularisations and explain how these differences apeear as a result of the math

L1 encourages a sparse solution , that is few w's non-zero , many-equal 0 )

L2 encourages small weights , that is all w's close to 0 but not exactly 0

Both help prevent overfitting , by not fitting to noise :

L1 accomplishes this by choosing the most important features

L2 accomplishes this by making the assertion that none of the weights are extremely large

Now think of a 1 dimensional weight :

For L2 regularisation we get the -log(Prior) to be quadratic =  $\lambda w^T w$

For L2 regularisation we get the -log(Prior) to be the absolute function =  $\lambda \vert w\vert$

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.linspace(-5,5)
plt.plot(w,w**2,label="-log prior of L2 regularisation")
plt.plot(w,np.abs(w),label="-log prior of L1 regularisation")
plt.legend()
plt.show()
```

Now consider the derivative of these functions , this is the key point since gradeint descent goes in the direction of derivative 

L2 Quadratic : as w approaches 0 , derivative also approaches 0 , so if w is already small , further gradien descent wont change it much

In this way L2 encourages weights to be small

L1 Absolute : derivative is always +/-1 , (0 at w = 0) , so it does not matter where w is , it will fall at a constant rate and when it reaches 0 , it stays there forever

Finally we can include both L1 and L2 regularisations simultaneoulsy , this is given a rather fancy name 'ElasticNet'

$$J_{RIDGE} = J + \lambda_2 {\vert w\vert}^2$$

$$J_{LASSO} = J + \lambda_1 {\vert w\vert}$$

$$J_{ELASTICNET} = J + + \lambda_1 {\vert w\vert} + \lambda_2 {\vert w\vert}^2$$







<h1>Math</h1>

Finally we discuss why we divide by $\sqrt{D}$ when initialsing weights

w = np.random.randn(D) / np.sqrt(D) 

poor weight initialisation can lead to poor cconvergance of loss , sometimes loss can even go to infinity

randn() == sample form N(0,1)

randn()*C == sample from N(0,$C^2$)

To standarise : $z = (x-\mu) / \sigma$

we subtract mean and divide by standard deviation so z has mean 0 and sd 1

we can also do the opposite , take the standarised z , multiply by $\sigma$ then add mean so we arrive at x with mean $\mu$ and sd $\sigma$

inverse transform : $x = z\sigma + \mu$

So dividing by $\sqrt{D}$ is the same as multiplying by $\frac {1} {\sqrt{D}} $ , this is the same as saying that we want our initial weights to have mean 0 and sd $\frac{1}{D}$ , so why do that

looking at things from a high level point of view , we know that one reason that causes the loss to explode (or reach some suboptimal value ) is that weights are too large , and so their variance is too large , and so may be if we decrease the variance we may solve this problem , but we might as well multiply by a constant like 0.01 or so , why exactly $\frac {1}{\sqrt{D}}$

Once again lets consider multiple linear regression , the same argument can be used in the future for a layer of a neural network :

$y = w_1x_1 + w_2x_2 + ... + w_Dx_D$

Now lets consider y , we wish for y to be standarised , that is to have variance of 1

that is because in neural networks the output of one llayer is just the input to the other , and so to prevent problems we want y standarised 

in linear regression we will probably have the targets standarised too , so we want y to be standarised

since all the xs and ws are independant we can express the variance of y as :

$var(y) = var(w_1)(x_1) + var(w_2)(x_2)+...+var(w_D)var(x_D)$

but since we standarised the data , we know the variance of all the xs is actually 1 , so this simplifies too

$var(y) = var(w_1) + var(w_2)+...+var(w)_D$

The variance of all the ws is also the same , since there is no reason to make the variance of the ws different

so this further simplifies too:

$var(y) = Dvar(w)$

so if we want var(y) = 1 , we must have var(w) = $\frac{1}{D}$

That is the same as saying that we want the sd(w) = $\frac{1}{\sqrt{D}}$ which , as shown previously , is aheived by dividing by $\sqrt{D}$


Of course there are other ways to initialise weights

It is not to say that this way gives optimal results , in fact it may cause poor results

It is always important to experiment on the given data
