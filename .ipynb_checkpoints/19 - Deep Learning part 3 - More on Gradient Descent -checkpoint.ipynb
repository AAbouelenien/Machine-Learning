<h1>Math</h1>

we start by considering different ways of doing gradient descent , there are 3 baasic types : Full,Batch,Stochastic dataset

Full :

$J = \sum t(n) log y(n) for \ n = 1...N$

Makes sense since we want to maximise the likelihood over entire training data set

Disadvantage : O(n) , wont work on big data

Stochastic:

$for \  n = 1...N$

$\ \ \ \ W \leftarrow W - \eta \nabla J(n) $ 

Depends on error of one particular sample

Depsend on the fact that all samples are IID (Independant and Identically Distributed) , so in the long run , error improves since all samples are coming from same distribution

While the log-likelihood always improves on every iteration of FGD (full gradient descent) sometimes it can get worse with SGD , but eventually it will improve

we also have to do the update and calculate the gradient for every single sample , multiple times

Batch:

batches = split data into batches

$for \ batch \ in \ batches:$

$\ \ \ W \leftarrow  \eta \nabla J(batch)$

Cost can still get worse , but is still elss erratic than SGD

```python
# now lets try training a neural net using FGD , SGD , BGD
# we will just do logisitc regression to speed things up
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.utils import shuffle
```

```python
# first lets get the data , we will use MNIST dataset
def get_mnist(limit=None,normalise = False , transform = False, D=300): 
    df = pd.read_csv('datasets/MNIST/train.csv')
    data = df.to_numpy()
    np.random.shuffle(data)
    X = data[:,1:]
    Y = data[:,0]
    # data is too large , so training would take too much time
    # of course to little data is not good and leads to overfitting

    Xtrain = X[:-1000]
    Ytrain = Y[:-1000]
    Xtest  = X[-1000:]
    Ytest  = Y[-1000:]
    
    
    if transform:
        # center X
        mu = Xtrain.mean(axis=0)
        Xtrain = Xtrain-mu
        Xtest = Xtest-mu
        pca = PCA()
        Ztrain = pca.fit_transform(Xtrain)
        Ztest = pca.transform(Xtest)
        # take first 300 cols of Z
        Ztrain = Ztrain[:, :300]
        Ztest = Ztest[:, :300]
        # normalize Z
        mu = Ztrain.mean(axis=0)
        std = Ztrain.std(axis=0)
        np.place(std,std==0,1)
        Ztrain = (Ztrain - mu) / std
        Ztest = (Ztest - mu) / std
        return Ztrain,Ztest,Ytrain,Ytest

    
    if normalise:
        mu = Xtrain.mean(axis=0)
        std = Xtrain.std(axis=0)
        np.place(std,std==0,1)
        Xtrain = (Xtrain-mu)/std
        Xtest = (Xtest-mu)/std
        return Xtrain,Xtest,Ytrain,Ytest
    
    return Xtrain,Xtest,Ytrain,Ytest
```

```python
# we need to one hot encode labels

def one_hot_encode(Y):
    N = len(Y)
    D = Y.max()+1
    encoded = np.zeros((N,D))
    for i in range(N):
        encoded[i,Y[i]] = 1
    return encoded

def softmax(A):
    expA = np.exp(A)
    return expA/expA.sum(axis=1,keepdims=True)

def forward(X,W,b,act):
    return act(X@W+b)

def cost(T, Y):
    return -np.sum(T * np.log(Y))

def accuracy(T,Y):
    return np.mean(T==Y)
```

```python
Xtrain,Xtest,Ytrain,Ytest = get_mnist(transform=True)

Ttrain = one_hot_encode(Ytrain)
Ttest = one_hot_encode(Ytest)

N,D = Xtrain.shape
K = len(set(Ytrain).union(Ytest))

epochs = 200
lr = 0.0001
reg = 0.1
```

```python
# full gradient descent (what we normally did up to this point)
W = np.random.randn(D,K)/np.sqrt(D)
b = np.random.randn(K)
costs_FGD = []
for i in range(epochs):
    # feedforward
    y = forward(Xtrain,W,b,softmax)
    W += lr*Xtrain.T.dot(Ttrain-y)
    b += lr*(Ttrain-y).sum(axis=0)
    # get test cost
    y = forward(Xtest,W,b,softmax)
    c = cost(Ttest,y)
    costs_FGD.append(c)
    
plt.plot(costs_FGD)
plt.title('FGD : Test data costs ')
plt.show()
preds = np.argmax(y,axis=1)
print("accuracy: ",accuracy(Ytest,preds))
```

```python
# Stochastic gradient descent 
costs_SGD = []
W = np.random.randn(D,K)/np.sqrt(D)
b = np.random.randn(K)

Ntrain = len(Xtrain)
Ntest = len(Xtest)
for i in range(1): # very slow
    tempX,tempT = shuffle(Xtrain,Ttrain)
    for n in range(min(Ntrain,500)): # veryyyy sloooow
        # feedforward
        point = np.atleast_2d(tempX[n])
        label = tempT[n]
        y = forward(point,W,b,softmax)
        W += lr*point.T.dot(label-y)
        b += lr*(label-y).sum(axis=0)
        # get test cost
        y = forward(Xtest,W,b,softmax)
        c = cost(Ttest,y)
        costs_SGD.append(c)

plt.plot(costs_SGD)
plt.title('SGD : Test data costs')
plt.show()
preds = np.argmax(y,axis=1)
print("accuracy: ",accuracy(Ytest,preds))
```

```python
# Batch gradient descent 
costs_BGD = []
W = np.random.randn(D,K)/np.sqrt(D)
b = np.random.randn(K)

Ntrain = len(Xtrain)
Ntest = len(Xtest)
batch_sz = 500
n_batches = int(Ntrain/batch_sz)

for i in range(50): 
    tempX,tempT = shuffle(Xtrain,Ttrain)
    for i in range(n_batches): 
        # feedforward
        batch = tempX[i*batch_sz : (i+1)*batch_sz]
        batch = np.atleast_2d(batch)
        labels = tempT[i*batch_sz : (i+1)*batch_sz]
        
        y = forward(batch,W,b,softmax)
        W += lr*batch.T.dot(labels-y)
        b += lr*(labels-y).sum(axis=0)
        # get test cost
        y = forward(Xtest,W,b,softmax)
        c = cost(Ttest,y)
        costs_BGD.append(c)

plt.plot(costs_BGD)
plt.title('BGD : Test data costs')
plt.show()
preds = np.argmax(y,axis=1)
print("accuracy: ",accuracy(Ytest,preds))
```

```python
x1 = np.linspace(0,1,len(costs_FGD))
plt.plot(x1,costs_FGD,label='Full Gradeint Descent')

x2 = np.linspace(0,1,len(costs_SGD))
plt.plot(x2,costs_SGD,label='Stochastic Gradeint Descent')

x3 = np.linspace(0,1,len(costs_BGD))
plt.plot(x3,costs_BGD,label='Batch Gradeint Descent')

plt.legend()
plt.show()
```

<h1>Math</h1>

Next , we will discuss one of the most effective improvements over plain gradient descent , Momentum

Imagine pushing a box on ice , the box continues to move without additional force 

<img src = 'extras/19.5.PNG' width = 300></img>

of course ice still has some friction so the box slows down and eventually the it stops 

Now lets try to think about the case with no momentum , this is just like pushing a box on sand , each time we want the box to move we need to apply force

<img src = 'extras/19.6.PNG' width = 300></img>

This is just like plain gradient descent with the gradient being analogous to force , so each time we want to move in direction of gradient , there has to be a gradient

But with momentum we can still move in direction of gradien(force) wven though if gradient = 0 (no force) as in the example of ice

Now lets try to put these ideas into math :

Regular Gradient Descent (No Momentum)

$$\theta_t = \theta_{t-1} - \eta g_t$$

if $g_t$ is 0 , parameter doesnot change (if force is 0 , position doesnot change  , sand example)

Gradient Descent with Momentum:

2 steps :

$$v_t \leftarrow \color{green}{\boxed{\mu v_{t-1}}} - \color{blue}{\boxed{\eta g_t}}$$

$$\theta_t \leftarrow \theta_{t-1} + v_t$$


$\color{green}{\boxed{\mu v_{t-1}}}$ : sliding on  ice , our current velocity is a fraction of our previous velocity , this resembles the effect of continuing to move in the same direction as we were before , of course , like the ice example , there is still friction and we stop eventually , hence $\mu$ , typical values for momentum term $\mu : 0.9,0.95,9.99$

$\color{blue}{\boxed{\eta g_t}}$ : Pushing the box , this is the force currently appliend on us

if we plug in v we get :

$$\theta_t \leftarrow \theta_{t-1} + \mu v_{t-1} - \eta g_t$$

it should be clear now that if we set $\mu$ = 0 (push in sand) , we get the regular gradient descent

$$\theta_t \leftarrow \theta_{t-1} + \require{cancel}\cancel{\mu v_{t-1}} - \eta g_t$$

Using momentum significantly sppeds up training

We also want to discuss one more perspective which will allow us to reason why we use momentum

Imagine the surface of the cost function as a hill , since we want to minimise the cost function we want to reach the lowest point

<img src = 'extras/19.7.PNG' ></img>

Momentum solves the problem of unequal gradients in different directions , so assume we have 2 parameters , vertical and horizontal , the gradient in one direction is very steep and is very shallow in the other 

If we dont have momentum , we rely only on gradient which points more in steep direction than in shallow direction (gradient is the direction of steepest descent) , so we find ourselves zigzaging back and forth acroos valley which is an inefficient way of reaching minimum

BUT , since in the shallow direction we move in the same direction every time , if we use momentum ,  the velocities will accumulate , so we will have a portion of the old velocity added to the new velocity to help us go in that direction , the result is reaching the bottom of hill much faster by taking bigger steps in direction of shallow gradient

<h1>Math</h1>

Next we discuss a modern improvement to momentum called Nesterov Momentum

The Basic idea is that instead of blindly using momentum to keep going in the direction we were already going , we instead peak ahead and make a correction :

<img src='extras/19.8.PNG' width = 500><img>

<font color = 'blue'>The blue vector is regular momentum , the short arrow due to gradient (force) , the long one is due to velocity</font>

<font color = 'brown'> The brown arrow is the jump in the same direction of velocity , notice how it is parralel to the long blue arrow</font>

<font color = 'red'>the red arrow is the correction we make , the gradient we calculate after we take the jump</font>

<font color = 'green'>and the green arrow is simply the sum of the red and brown arrows , that is the accumulated gradient</font>

Lets see this in equations:

Jump : look-ahead value : 

$$w'_{t-1} \leftarrow w_{t-1} + \mu v_{t-1}$$

Next calculate gradient at this point and update V

$$v_t \leftarrow \mu v_{t-1} - \eta \nabla J(w'_{t-1})$$


$$= \mu v_{t-1} - \eta \nabla J(w_{t-1} + \mu v_{t-1})$$

then update w same as standard momentum:

$$w_t \leftarrow w_{t-1} +v_t$$

So this is baasically the same as momentum but instead of using gradient at current parameters , we use gradients of look-ahead parameters

Now lets attempt to express everything in terms of w' only (this will help us in implementation) :

first dconsider the look-ahead values at 2 consecutive steps :

$$w'_t = w_t + \mu v_t$$

$$w'_{t-1} = w_{t-1} + \mu v_{t-1}$$

Then plug these in update of $w_t$

$$w_t = w_{t-1} + v_t$$

so we yeild :

$$w'_{t} - \mu v_t = w'_{t-1} - \mu v_{t-1} + v_t$$

combine like terms (2 terms depend on $v_t$):

$$w'_{t} =  w'_{t-1} - \mu v_{t-1} + (1+\mu)v_t$$

next we want to get rid of $v_{t}$ so that everything depends on things from previous time step (t-1)

we can plug in 

$$v_t = \mu v_{t-1} - \eta \nabla J (w'_{t-1})$$

and we get :

$$w'_{t} =  w'_{t-1} - \mu v_{t-1} + (1+\mu)\left[ \mu v_{t-1} - \eta \nabla J (w'_{t-1}) \right]$$

expand : 

$$w'_{t} =  w'_{t-1} - \mu v_{t-1} + \mu v_{t-1} + \mu^2 v_{t-1} - (1+\mu)\left[\eta \nabla J (w'_{t-1}) \right]$$

then $- \mu v_{t-1} + \mu v_{t-1}$ cancel out


$$w'_{t} =  w'_{t-1} + \mu^2 v_{t-1} - (1+\mu)\left[\eta \nabla J (w'_{t-1}) \right]$$

Now since we arrive at this form , we can compare it to regular momentum

Regular momentum :

$$w_t = w_{t-1} + \mu v_{t-1} - \eta \nabla J(w_{t-1})$$

Nesterov momentum :

$$w'_{t} =  w'_{t-1} + \mu^2 v_{t-1} - (1+\mu)\left[\eta \nabla J (w'_{t-1}) \right]$$


in both equations , we can see the previous w term , the previous velocity term and the gradient term

we can also phrase this in term of $v_t$ instead of $v_{t-1}$

we an use :

$$v_t = \mu v_{t-1} - \eta \nabla J(w'_{t-1})$$

to get :

$$ w'_{t} = w'_{t-1} + \mu v_{t} - \eta \nabla J(w'_{t-1})$$

finally how was this useful ?

Everyting now is phrased in terms of look-ahead w' , we no longer need w at all 

now we can drop the ' symbol:

$$v_t = \mu v_{t-1} - \eta \nabla J(w_{t-1})$$


$$ w_{t} = w_{t-1} + \mu v_{t} - \eta \nabla J(w_{t-1})$$

again here lets write equations for regular and nesterov momentums

Regular momentum :

$$v_t = \mu v_{t-1} - \eta \nabla J(w_{t-1})$$


$$w_t = w_{t-1} + \mu v_{t-1} - \eta \nabla J(w_{t-1})$$


Nesterov momentum :

$$v_t = \mu v_{t-1} - \eta \nabla J(w_{t-1})$$

$$ w_{t} = w_{t-1} + \mu v_{t} - \eta \nabla J(w_{t-1})$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.utils import shuffle
```

```python
# first lets get the data , we will use MNIST dataset
def get_mnist(limit=None,normalise = False , transform = False, D=300): 
    df = pd.read_csv('datasets/MNIST/train.csv')
    data = df.to_numpy()
    np.random.shuffle(data)
    X = data[:,1:]
    Y = data[:,0]
    # data is too large , so training would take too much time
    # of course to little data is not good and leads to overfitting

    Xtrain = X[:-1000]
    Ytrain = Y[:-1000]
    Xtest  = X[-1000:]
    Ytest  = Y[-1000:]
    
    
    if transform:
        # center X
        mu = Xtrain.mean(axis=0)
        Xtrain = Xtrain-mu
        Xtest = Xtest-mu
        pca = PCA()
        Ztrain = pca.fit_transform(Xtrain)
        Ztest = pca.transform(Xtest)
        # take first 300 cols of Z
        Ztrain = Ztrain[:, :300]
        Ztest = Ztest[:, :300]
        # normalize Z
        mu = Ztrain.mean(axis=0)
        std = Ztrain.std(axis=0)
        np.place(std,std==0,1)
        Ztrain = (Ztrain - mu) / std
        Ztest = (Ztest - mu) / std
        return Ztrain,Ztest,Ytrain,Ytest

    
    if normalise:
        mu = Xtrain.mean(axis=0)
        std = Xtrain.std(axis=0)
        np.place(std,std==0,1)
        Xtrain = (Xtrain-mu)/std
        Xtest = (Xtest-mu)/std
        return Xtrain,Xtest,Ytrain,Ytest
    
    return Xtrain,Xtest,Ytrain,Ytest
```

```python
# we need to one hot encode labels

def one_hot_encode(Y):
    N = len(Y)
    D = Y.max()+1
    encoded = np.zeros((N,D))
    for i in range(N):
        encoded[i,Y[i]] = 1
    return encoded

def softmax(A):
    expA = np.exp(A)
    return expA/expA.sum(axis=1,keepdims=True)

def forward(X,W,b,act):
    return act(X@W+b)

def cost(T, Y):
    return -np.sum(T * np.log(Y))

def accuracy(T,Y):
    return np.mean(T==Y)

def sigmoid(a):
    return 1/(1+np.exp(-a))

def relu(a):
    return a*(a>0)
```

```python
Xtrain,Xtest,Ytrain,Ytest = get_mnist(normalise=True)

Ttrain = one_hot_encode(Ytrain)
Ttest = one_hot_encode(Ytest)
```

```python
# remember if you start getting nans , start changing hyperparameters
epochs = 20
lr = 0.0003
reg = 0.01
mu = 0.9
Ntrain = len(Xtrain)
Ntest = len(Xtest)
batch_sz = 500
n_batches = int(Ntrain/batch_sz)

_,D = Xtrain.shape
# intead of transforming data sing PCA (which is a linear transformer)
# we let the neural network do the work it self and extract the features
# so after having and NxD matrix we will have and NxM matrix , same as with PCA after we took most important features
M = 300
K = len(set(Ytrain).union(Ytest))

# accuracy/costs converges quickly
# to compare we will get accuracies and costs after n_batches
print_every = 20
```

```python
W = np.random.randn(D,M)/np.sqrt(D)
b = np.random.randn(M)

V = np.random.randn(M,K)/np.sqrt(M)
c = np.random.randn(K)

original_weights = [W.copy(),b.copy(),V.copy(),c.copy()]
```

```python
# plain Batch gradient descent 

accuracies_BGD = []
costs_BGD = []

for i in range(epochs): 
    #just to have a fair comparison we dont shuffle
    #tempX,tempT = shuffle(Xtrain,Ttrain)
    tempX,tempT = Xtrain,Ttrain
    for j in range(n_batches): 
        # feedforward
        x_batch = tempX[j*batch_sz : (j+1)*batch_sz]
        x_batch = np.atleast_2d(x_batch)
        t_batch = tempT[j*batch_sz : (j+1)*batch_sz]
        
        Z = forward(x_batch,W,b,relu)
        y_batch = forward(Z,V,c,softmax)

        #calculate gradients
        delta = t_batch-y_batch 
        grad_V = Z.T@(delta)
        grad_c = delta.sum(axis=0)

        Z_prime = (Z>0) # for relu
        grad_b = ((delta@V.T)*Z_prime)
        grad_W = x_batch.T@grad_b
        grad_b = grad_b.sum(axis=0)

        # now we perform gradient descent
        b += lr*(grad_b-reg*b)
        c += lr*(grad_c-reg*c)
        W += lr*(grad_W-reg*W)
        V += lr*(grad_V-reg*V)

    
    #get test cost and accuracy
        if j%print_every == 0:
            #get test cost and accuracy
            Z_test = forward(Xtest,W,b,relu)
            y_test = forward(Z_test,V,c,softmax)
            co = cost(Ttest,y_test)
            costs_BGD.append(co)
            preds = np.argmax(y_test,axis=1)
            acc = accuracy(Ytest,preds)
            accuracies_BGD.append(acc)
            print("epoch: "+str(i)+" cost: "+str(co)+" accuracy: "+str(acc))

plt.plot(costs_BGD)
plt.title('Test data costs')
plt.show()
plt.plot(accuracies_BGD)
plt.title('Test data accuracies')
plt.show()
print("final test cost: ",co)
print("final test accuracy: ",acc)
```

```python
# next momentum
# we add mu amd v

v_W = 0
v_V = 0
v_c = 0
v_b = 0


W = original_weights[0].copy()
b = original_weights[1].copy()
V = original_weights[2].copy()
c = original_weights[3].copy()


accuracies_momentum = []
costs_momentum = []

for i in range(epochs): 
    #just to have a fair comparison we dont shuffle
    #tempX,tempT = shuffle(Xtrain,Ttrain)
    tempX,tempT = Xtrain,Ttrain
    for j in range(n_batches): 
        # feedforward
        x_batch = tempX[j*batch_sz : (j+1)*batch_sz]
        x_batch = np.atleast_2d(x_batch)
        t_batch = tempT[j*batch_sz : (j+1)*batch_sz]
        
        Z = forward(x_batch,W,b,relu)
        y_batch = forward(Z,V,c,softmax)

        #calculate gradients
        delta = t_batch-y_batch 
        grad_V = Z.T@(delta)
        grad_c = delta.sum(axis=0)

        Z_prime = (Z>0) # for relu
        grad_b = ((delta@V.T)*Z_prime)
        grad_W = x_batch.T@grad_b
        grad_b = grad_b.sum(axis=0)

        # first we calculate v
        # + or - depends on whether delta in T-Y or Y-T , both are equivalent
        v_W = mu*v_W + lr*(grad_W-reg*W) 
        v_V = mu*v_V + lr*(grad_V-reg*V) 
        v_b = mu*v_b + lr*(grad_b-reg*b) 
        v_c = mu*v_c + lr*(grad_c-reg*c) 
        # now we update weights
        b += v_b
        c += v_c
        W += v_W
        V += v_V

    #get test cost and accuracy
        if j%print_every == 0:
            #get test cost and accuracy
            Z_test = forward(Xtest,W,b,relu)
            y_test = forward(Z_test,V,c,softmax)
            co = cost(Ttest,y_test)
            costs_momentum.append(co)
            preds = np.argmax(y_test,axis=1)
            acc = accuracy(Ytest,preds)
            accuracies_momentum.append(acc)
            print("epoch: "+str(i)+" cost: "+str(co)+" accuracy: "+str(acc))

plt.plot(costs_momentum)
plt.title('Test data costs')
plt.show()
plt.plot(accuracies_momentum)
plt.title('Test data accuracies')
plt.show()
print("final test cost: ",co)
print("final test accuracy: ",acc)
```

Next we will discuss variable and adaptive learning rates

Variable learning is when learning rate is a function of time , e.g. $\eta(t)$ , sometimes also called "Learning rate scheduling"

<h5>Step decay</h5> 

periodically , say every 100 steps , reduce learning rate by a constant factor , so for example divide learning rate by 2

if we plot it it should look something like a staricase

<img src='extras/19.9.PNG' width=300></img>

<h5>exponential decay</h5>

in this method the learning rate follows and exponential curve $\eta(t) = A \ exp(-kt)$

<img src='extras/19.10.PNG' width=500></img>

<h5> 1/t decay </h5>

learning rate decays proportionally to 1\t , $\eta(t) = \frac{A}{(kt+1)}$ , Dropoff is slower than exponential decay , we can also change k to control the how fast/slow the decay is

<img src='extras/19.11.PNG' width=500></img>

All methods have learning rate decrease with time , initially weights are far from optimal , but as we get closer to minimum , learning rate may be too large , we might end up overshoting and in fact the loss may even increase

<img src='extras/19.12.PNG' width=300></img>

Babysitting method:

do a few epochs , see how it goes (use costs plot)

Too slow $\rightarrow$ increase learning rate

But we may hit a plateau temporarily , so we have to be patient and wait (of course this depends on our resources)

<img src='extras/19.13.PNG' ></img>

One other thing in common , more hyperparameters !

In fact choosing between these methods may be considered itself a hyperparameter

Next are Adaptive learning rates , these adapt to the data we have seen so far

<h5>AdaGrad</h5> 

We cant expect the dependance of the cost on each parameter to be the same , in other words , in one directino the gradient may be very steep , but in the other directon the gradient may be very flat

So perhaps it might be beneficial to Adapt the learning rate for each parameter individually based on how much it has changed in the past

so what we do in Adagrad is introdice a variable called cache , so if we have one weight matrix of size 3x4 we also have a cahce matrix of size 3x4 (same for bais vectors)

each parameter of the neural network has its own cache

$$cache = cache + {gradient}^2$$

(the cash is updated with every batch assuming we use BGD)

$$\theta \leftarrow \theta - \eta \frac{\nabla J}{\sqrt{cache + \epsilon}}$$

The idea is that the cahce accuculates the square gradients for every parameter (since we square the cache value is always positive) , so if the parameter had high gradients before , then the cache value becomes very large , and its effective learning rate becomes very small , and so it changes more slowly in the future , on the other hand if a parameter has had small gradients in the past , then its cache becomes small , its effective learning rate becomes large , and it has a greater opportunity to change in the future

we usually add a very small value $\epsilon$ (around $10^{-8}$ or $10^{-10}$) to denominator to avoid dividing by 0 

Once agian we stress aout everything being an element-wise operation , each scalar parameter and its learning rate are updated independantly of the others

<h5>RMSProp</h5> 

This follows observations of how AdaGrad decreases learning rate too aggressively , so the learning rate approaches 0 quickly when in fact there is still much learning left to do

The reason AdaGrad decreases gradients too quickly is because the cache is growing too fast , so in order to make it increase less fast , we decrease it with each update

$$cache = decay*cache+(1-decay)*{gradient}^2$$

we do this by taking a weighted average of the old cache and the new squared gradient , we call this weight the decay rate , and we can see that both weights add up to 1


we will consider this thoroghly later ,for now it is suffecient to remember that we take part of the squared gradient and part of the old cache and add them together

typical values for decay rate are 0.99,0.999,etc

so we say that the cahce now is 'leaky' , since if it had 0 gradient for a large time , it would shrink back to 0 due to the effect of the decay rate

Note : we need to consider the initial value of the cache since it will affect the update

if we let cahce = 0 , and assuming a decay of 0.000 , initial update for cache = 0.001$g^2$

disregarding $\epsilon$ for the time being , our effective learning rate becomes very large (since the denominator is very small only 1/1000 of gradient) , and so our initial update for gradient becomes also very large:

$$\frac{\eta g}{\sqrt{0.001 g^2}}$$

so we need to compensate by making learning rate smaller than usual

one solution is to set the cache to 1 instead , the result is that the first update is almost like doing plain gradient descent

$$\Delta \theta = \eta \frac{g}{0.999(1) + 0.001{g}^2} \approx \eta g$$

that is the effective learning rate = effective_learning_rate/1

```python
lr = 0.001
decay = 0.999
epsilon = 1e-8

W = original_weights[0].copy()
b = original_weights[1].copy()
V = original_weights[2].copy()
c = original_weights[3].copy()

cache_W = np.zeros(W.shape)
cache_b = np.zeros(b.shape)
cache_V = np.zeros(V.shape)
cache_c = np.zeros(c.shape)

accuracies_RMSProp = []
costs_RMSProp = []

for i in range(epochs): 
    #just to have a fair comparison we dont shuffle
    #tempX,tempT = shuffle(Xtrain,Ttrain)
    tempX,tempT = Xtrain,Ttrain
    for j in range(n_batches): 
        # feedforward
        x_batch = tempX[j*batch_sz : (j+1)*batch_sz]
        x_batch = np.atleast_2d(x_batch)
        t_batch = tempT[j*batch_sz : (j+1)*batch_sz]
        
        Z = forward(x_batch,W,b,relu)
        y_batch = forward(Z,V,c,softmax)

        #calculate gradients & add regularisation terms
        delta = t_batch-y_batch 
        grad_V = Z.T@(delta)-reg*V
        grad_c = delta.sum(axis=0)-reg*c

        Z_prime = (Z>0) # for relu
        grad_b = ((delta@V.T)*Z_prime)
        grad_W = x_batch.T@grad_b-reg*W
        grad_b = grad_b.sum(axis=0)-reg*b

        # update cache
        cache_W = decay*cache_W + (1-decay)*grad_W**2
        cache_V = decay*cache_V + (1-decay)*grad_V**2
        cache_b = decay*cache_b + (1-decay)*grad_b**2
        cache_c = decay*cache_c + (1-decay)*grad_c**2
        
        # now we update weights
        # effective learning rate
    
        b += grad_b*(lr/(np.sqrt(epsilon+cache_b)))
        c += grad_c*(lr/(np.sqrt(epsilon+cache_c)))
        W += grad_W*(lr/(np.sqrt(epsilon+cache_W)))
        V += grad_V*(lr/(np.sqrt(epsilon+cache_V)))

    #get test cost and accuracy
        if j%print_every == 0:
            #get test cost and accuracy
            Z_test = forward(Xtest,W,b,relu)
            y_test = forward(Z_test,V,c,softmax)
            co = cost(Ttest,y_test)
            costs_RMSProp.append(co)
            preds = np.argmax(y_test,axis=1)
            acc = accuracy(Ytest,preds)
            accuracies_RMSProp.append(acc)
            print("epoch: "+str(i)+" cost: "+str(co)+" accuracy: "+str(acc))

plt.plot(costs_RMSProp)
plt.title('Test data costs')
plt.show()
plt.plot(accuracies_RMSProp)
plt.title('Test data accuracies')
plt.show()
print("final test cost: ",co)
print("final test accuracy: ",acc)
```

<h1>Math</h1>

Now we will talk about Adam Optimiser , which is often considered as the go-to default

Sometimes it is refered to as RMSProp with momentum , Adam does in a sense add momentum to RMSProp but in a very specific way (which we will see)

tensorflow allows us to perform RMSProp + momentum but that does not yeild Adam

First we need to introduce "Expnentially-Smoothed Averages" , a key piece of theory essential to understanding Adam

lets begin by considering how to calculate then average , if we have N points then we can get the sample mean by:

$$\bar X_N = \frac{1}{N}X_n$$

Now lets sat we have a large amount of data that cant fit into memory all at same time , can we still calculate the mean ?

what we can do is read one data point at a time , then delete each data point after we looked at it , mathematically we begin from :

$$\bar X_N = \frac{1}{N}X_n$$

which could be manipulated to yeild:

$$\frac{1}{N}\left[ (N-1) \bar X_{N-1} + X_N\right]$$

use simpler symbols , here Y is our output , and t is our time step

$$Y(t) = \left(1-\frac{1}{t}\right)Y(t-1) + \frac{1}{t}X(t)$$

now we need to consider the previous equation , think about the $\frac{1}{t}$ in $\frac{1}{t}X(t)$ , it says that as t grows larger , the current sample has less and less effect on the total mean

makes sense as we collect more Xs the total number of datapoints grows , so each new X has less and less influence on current Y

but we also decrease the influence of the previous Y by $(1-\frac{1}{t})$ , so the current Y is part of the old Y + part of the newsest X , this , in the end , balances out to give the exact sample mean of X

for convenience lets call the $\frac{1}{t}$ term $\alpha(t)$

now lets say that we dont want $\alpha$ to equal $\frac{1}{t}$ , lets say we want each datapoint to matter equally at the time we see it , so we set $\alpha$ to a constant

$$Y(t) = \left( 1-\alpha \right) Y(t-1) + \alpha X(t)$$
 
(of course $\alpha$ needs to be $0 < \alpha < 1$ or we may end up negating previous mean)

This gives us what we call "exponential smoothed avaerage" , we can see why its called "exponential when we express it only in terms of X"

$$Y(t) = (1-\alpha)Y(t-1) + \alpha X(t)$$

$$Y(t) = {(1-\alpha)}^t Y(0) + \alpha \sum^{t-1}_{\tau = 0} (1-\alpha)^\tau X (t-\tau)$$

(same as how we solve recurrence in algorithms)

now does this still give us the mean or in other words the expected value of X ?

well if we take the expected value of everyting we see that we arrive at the expected value of X

recall the definition of expected value E(X):

$$E(X) = \sum x_ip(x_i)$$

so we take $Y(0)$ and each $X(t-\tau)$ as datapoints ($x_i$) an we want to prove all their coeffecients ($p(x_i)$) = 1

so we want to prove , for any value of $\tau$: 

$$ (1-\alpha)^\tau + \alpha \sum^{t-1}_{\tau = 0} (1-\alpha) = 1$$

the summation is equivalent to a gemetric series where a = $\alpha$ and $r = (1-\alpha)$: 

$$ \alpha \sum^{t-1}_{\tau = 0} (1-\alpha) = \sum^{t-1}_{\tau=0} \alpha (1-\alpha)^\tau $$

we can use the rule for sum of geometric series :
$$\sum_{k=0}^n ar^k =  \left( \frac{1 - r^{n+1}}{1-r}\right)$$ 

to get

$$ \alpha \sum^{t-1}_{\tau = 0} (1-\alpha) = \alpha \left( \frac{1 - (1-\alpha)^\tau}{1-(1-\alpha)} \right) = \alpha \left( \frac{1 - (1-\alpha)^\tau}{\alpha} \right) = 1 - (1-\alpha)^\tau $$

now lets plug this back into :

$$ (1-\alpha)^\tau + \alpha \sum^{t-1}_{\tau = 0} (1-\alpha) = 1$$

to get :

$$ (1-\alpha)^\tau + 1 - (1-\alpha)^\tau = 1 \ ; \ 1 = 1$$

so this is a valid expectation value for X

lets review again what we said :

$$\bar X_N = \frac{1}{N}X_n$$

$$Y(t) = \left(1-\frac{1}{t}\right)Y(t-1) + \frac{1}{t}X(t)$$

In case of the simple mean all data points are equi-probable and thus all points equally influnce the mean , this is clear in the first formula , in the second formula one might be tempted to beleive that the later the point comes the smaller the influence , but since we also take only a part of the previous mean eventually it balances out , of course this is a valid E(X) 

$$Y(t) = {(1-\alpha)}^t Y(0) + \alpha \sum^{t-1}_{\tau = 0} (1-\alpha)^\tau X (t-\tau)$$

The exponential smoothed average is also another valid E(X) , this time newer points are given higher weights and so they have more influence on the mean

if X is not a fixed distribution , rather it changes with time , then this may be a better way to calculate mean

<h1>Math</h1>

Now lets go back to Adam Optimiser

we already showed the formula of the exponential smoothed average to be

$$Y(t) = \left( 1-\alpha \right) Y(t-1) + \alpha X(t)$$

Now let decay = $\left( 1-\alpha \right)$ , and replace symbol Y by M (M for mean :) ) , now we get :

$$M_T = decay * M_{T-1} + (1-decay)X_T$$

but this looks exactly like the RMSProp cache update !

so now we know that the "cache" of RMSProp i estimating the average of the squared gradient !!!

$$v_t = decay * v_{t-1} + (1-decay)*g^2 \approx mean(g^2)$$

We eventually take the square root of cache in parameter update , hence the name RMSProp (RMS="root mean square")

remember that the mean of something is its expected value , e.g. mean(X) = $E(X)$ , when we take the expected value of the square of a random variable , $E(X^2)$ , we call this the 2nd moment , $E(X^n)$ is the nth moment of X

so $E(g^2)$ is the second moment of the gradient which we calculated using exponential smooth average , since the  second centralised moment of a variable is the definition of variance , we denote $E(g^2)$ by v , hence $v \approx E(g^2)$

the first moment of course is just the mean of X

$$m_t = \mu m_{t-1} + (1-\mu)g_t$$

$$m \approx E(g)$$

Adam makes use of 1st and 2nd moments of g , which explains its name "Adaptive Moment Estimation" , also notice how the update for m (here mean) looks a lot like momentum $m_t = \mu m_{t-1} + g_t$ (here m for velocity , note that there are multiple ways to implement momentum) , hence RMSProp + momentum ,v like RMSProp , and m like momentum

Now we want to see how Adam uses both m and v , but we need one more concept to discuss

let try to quickly apply exponential smooth average on a time-series function like sin

```python
import numpy as np
import matplotlib.pyplot as plt
N = 200
decay = 0.95
x = np.linspace(0,10,N)
# y is a sin wave with some noise
y = 3*np.sin(x) + np.random.randn(N)
plt.plot(x,y)
# now lets try exponential smooth averaging
E_Y = np.zeros(N)
for i in range(len(y)):
    ESA = decay*E_Y[i-1] + (1-decay)*y[i]
    E_Y[i] = ESA


plt.plot(x,y,label = 'function + noise')
plt.plot(x,E_Y,label = 'exponentially smoothed average function')
plt.legend()
plt.show()
```

we can see that this acts like a smoothing funtion , it ignores many of the random spikes but follows the general trend of the signal , but still a problem persists , what is the initial value of E_Y ?

if $Y(0) = 0$ , $Y(1) = 0.99 * Y(0) + 0.01 * X(1) = 0.001 * X(1)$ , so our initial output is bias towards 0 , can we do anything about that ?

we use Bias Correction for this 

first calculate Y as normal then  

$$\hat Y(t) = \frac{Y(t)}{1-{decay}^{\ t}}$$

when t is small we are dividing by a very small number , which makes initial value bigger

lets do an example so things get clearer :

given decay = 0.99

normally :

$$Y(1) = 0.01X(1)$$

corrected :

$$\hat Y(1) = \frac{0.01X(1)}{(1-0.99)}=X(1)$$

normally :

$$Y(2) = 0.0099X(1) + 0.01X(2)$$

corrected:

$$\hat Y(2) = \frac{Y(2)}{0.0199}=0.470X(1) + 0.503 X(2)$$

we can see how as t increases , the bias correction term approaches 1 and so we quickly return to the standard exponential smooth averaging

now back to Adam , surely we will aplly the bias correction to both m and v , now lets write formulas for coorected m and v as well as the formula for update

$$\hat m_t = \frac{m_t}{1-\beta_1^{t}}$$

$$\hat v_t = \frac{v_t}{1-\beta_2^{t}}$$

$$\theta_{t+1} \leftarrow \theta_y - \eta \frac{\hat m_t}{\sqrt{\hat v_t} + \epsilon}$$

note that there are two decay rates for m and v , $\beta_1$ and $\beta_2$

note that the update is similar to RMSProp , where cache is replaced by $\hat v_t$ (both are the same , just cache corrected) , and $\nabla J$ is replaced by $E(g)$ which is sort of applying momentum to gradient (in a specific way)

note :

A May 2017 paper "The Marginal Value of Adaptive Gradient Methods in Machine Learning" , showed that experimentally regulat gradient descent performs better than adaptive techniques !

This may seem surprsining but hyperparameters are meant to be optimised and everything is data dependant , so it may be better to stick to fundementals

No rule of thumb , only rule is "everything is data dependant"

```python
# lets try coding adam
```

```python
lr = 0.001
beta_1 = 0.9
beta_2 = 0.999
epsilon = 1e-8
t = 1

W = original_weights[0].copy()
b = original_weights[1].copy()
V = original_weights[2].copy()
c = original_weights[3].copy()

m_W = np.zeros(W.shape)
m_b = np.zeros(b.shape)
m_V = np.zeros(V.shape)
m_c = np.zeros(c.shape)

v_W = np.zeros(W.shape)
v_b = np.zeros(b.shape)
v_V = np.zeros(V.shape)
v_c = np.zeros(c.shape)


accuracies_Adam = []
costs_Adam = []

for i in range(epochs): 
    #just to have a fair comparison we dont shuffle
    #tempX,tempT = shuffle(Xtrain,Ttrain)
    tempX,tempT = Xtrain,Ttrain
    for j in range(n_batches): 
        # feedforward
        x_batch = tempX[j*batch_sz : (j+1)*batch_sz]
        x_batch = np.atleast_2d(x_batch)
        t_batch = tempT[j*batch_sz : (j+1)*batch_sz]
        
        Z = forward(x_batch,W,b,relu)
        y_batch = forward(Z,V,c,softmax)

        #calculate gradients & add regularisation terms
        delta = t_batch-y_batch 
        grad_V = Z.T@(delta)-reg*V
        grad_c = delta.sum(axis=0)-reg*c

        Z_prime = (Z>0) # for relu
        grad_b = ((delta@V.T)*Z_prime)
        grad_W = x_batch.T@grad_b-reg*W
        grad_b = grad_b.sum(axis=0)-reg*b

        
        # update m
        m_W = beta_1*m_W + (1-beta_1)*grad_W
        m_V = beta_1*m_V + (1-beta_1)*grad_V
        m_b = beta_1*m_b + (1-beta_1)*grad_b
        m_c = beta_1*m_c + (1-beta_1)*grad_c

        
        # update v
        v_W = beta_2*v_W + (1-beta_2)*grad_W**2
        v_V = beta_2*v_V + (1-beta_2)*grad_V**2
        v_b = beta_2*v_b + (1-beta_2)*grad_b**2
        v_c = beta_2*v_c + (1-beta_2)*grad_c**2
        
        # bias correct
        m_correction = 1-beta_1**t 
        m_W_hat = (m_W)/m_correction
        m_V_hat = (m_V)/m_correction
        m_b_hat = (m_b)/m_correction
        m_c_hat = (m_c)/m_correction
        
        v_correction = 1-beta_2**t 
        v_W_hat = (v_W)/v_correction
        v_V_hat = (v_V)/v_correction
        v_b_hat = (v_b)/v_correction
        v_c_hat = (v_c)/v_correction
        
        t+=1
        # now we update weights
        # effective learning rate
        
        b += m_b_hat*(lr/(np.sqrt(epsilon+v_b_hat)))
        c += m_c_hat*(lr/(np.sqrt(epsilon+v_c_hat)))
        W += m_W_hat*(lr/(np.sqrt(epsilon+v_W_hat)))
        V += m_V_hat*(lr/(np.sqrt(epsilon+v_V_hat)))

    #get test cost and accuracy
        if j%print_every == 0:
            #get test cost and accuracy
            Z_test = forward(Xtest,W,b,relu)
            y_test = forward(Z_test,V,c,softmax)
            co = cost(Ttest,y_test)
            costs_Adam.append(co)
            preds = np.argmax(y_test,axis=1)
            acc = accuracy(Ytest,preds)
            accuracies_Adam.append(acc)
            print("epoch: "+str(i)+" cost: "+str(co)+" accuracy: "+str(acc))

plt.plot(costs_Adam)
plt.title('Test data costs')
plt.show()
plt.plot(accuracies_Adam)
plt.title('Test data accuracies')
plt.show()
print("final test cost: ",co)
print("final test accuracy: ",acc)
```

```python
# finally lets try RMSProp + monetum
# there are multiple ways to implement this 
# we will procees normally by updating cahce
# this time cahce begins at 1 , since there is no bias correction (and we want fairest comparison to Adam)
# we can still start cache at 0 
# then we use momentum (weighted average) , replacing g by the diff in g we would have made in momentum
```

```python
lr = 0.001
decay = 0.999
epsilon = 1e-8
mu = 0.9

W = original_weights[0].copy()
b = original_weights[1].copy()
V = original_weights[2].copy()
c = original_weights[3].copy()

cache_W = np.ones(W.shape)
cache_b = np.ones(b.shape)
cache_V = np.ones(V.shape)
cache_c = np.ones(c.shape)

# this time v is that of momentum

v_W = np.zeros(W.shape)
v_b = np.zeros(b.shape)
v_V = np.zeros(V.shape)
v_c = np.zeros(c.shape)

accuracies_RMSProp_momentum = []
costs_RMSProp_momentum = []

for i in range(epochs): 
    #just to have a fair comparison we dont shuffle
    #tempX,tempT = shuffle(Xtrain,Ttrain)
    tempX,tempT = Xtrain,Ttrain
    for j in range(n_batches): 
        # feedforward
        x_batch = tempX[j*batch_sz : (j+1)*batch_sz]
        x_batch = np.atleast_2d(x_batch)
        t_batch = tempT[j*batch_sz : (j+1)*batch_sz]
        
        Z = forward(x_batch,W,b,relu)
        y_batch = forward(Z,V,c,softmax)

        #calculate gradients & add regularisation terms
        delta = t_batch-y_batch 
        grad_V = Z.T@(delta)-reg*V
        grad_c = delta.sum(axis=0)-reg*c

        Z_prime = (Z>0) # for relu
        grad_b = ((delta@V.T)*Z_prime)
        grad_W = x_batch.T@grad_b-reg*W
        grad_b = grad_b.sum(axis=0)-reg*b

        # update cache
        cache_W = decay*cache_W + (1-decay)*grad_W**2
        cache_V = decay*cache_V + (1-decay)*grad_V**2
        cache_b = decay*cache_b + (1-decay)*grad_b**2
        cache_c = decay*cache_c + (1-decay)*grad_c**2
        
        
        #update v
        # now replace g by delta g of RMSProp
        v_b = mu*v_b + (1-mu)*grad_b*(lr/(np.sqrt(epsilon+cache_b))) # instead if just grad_b
        v_c = mu*v_c + (1-mu)*grad_c*(lr/(np.sqrt(epsilon+cache_c)))
        v_W = mu*v_W + (1-mu)*grad_W*(lr/(np.sqrt(epsilon+cache_W)))
        v_V = mu*v_V + (1-mu)*grad_V*(lr/(np.sqrt(epsilon+cache_V)))

        # now we update weights as we would have done with momentum
    
        b += v_b
        c += v_c
        W += v_W
        V += v_V

    #get test cost and accuracy
        if j%print_every == 0:
            #get test cost and accuracy
            Z_test = forward(Xtest,W,b,relu)
            y_test = forward(Z_test,V,c,softmax)
            co = cost(Ttest,y_test)
            costs_RMSProp_momentum.append(co)
            preds = np.argmax(y_test,axis=1)
            acc = accuracy(Ytest,preds)
            accuracies_RMSProp_momentum.append(acc)
            print("epoch: "+str(i)+" cost: "+str(co)+" accuracy: "+str(acc))

plt.plot(costs_RMSProp_momentum)
plt.title('Test data costs')
plt.show()
plt.plot(accuracies_RMSProp_momentum)
plt.title('Test data accuracies')
plt.show()
print("final test cost: ",co)
print("final test accuracy: ",acc)
```

```python
plt.plot(costs_BGD,label='BGD')
plt.plot(costs_Adam,label='Adam')
plt.plot(costs_RMSProp_momentum,label='RMSProp + Momentum')
plt.title('Costs')
plt.legend()
plt.show()

plt.plot(accuracies_BGD,label='BGD')
plt.plot(accuracies_Adam,label='Adam')
plt.plot(accuracies_RMSProp_momentum,label='RMSProp + Momentum')
plt.title('Accuracy')
plt.legend()
plt.show()
```
