````python
# first we begin with an HMM for Discrete observations
import numpy as np
import matplotlib.pyplot as plt
````

````python
def coin_toss(N=50,T=30):
    # our data will be a sequence of coin tosses
    # N sampels each of size T
    # first we initialise variables that represent the true parameters of the HMM
    true_pi = np.array([0.5,0.5])
    true_A = np.array([[0.1,0.9],[0.8,0.2]])
    true_B = np.array([[0.6,0.4],[0.3,0.7]])
    # we will use these for sampling
    # hopefully after training our model learns these parameters
    
    sequences = []
    for _ in range(N): # sample a sequence
        sequence = []
        # first state
        s = np.random.choice(2,p=true_pi)
        # first observation
        o = np.random.choice(2,p=true_B[s])
        sequence.append(o)
        # the rest of the observations
        for _ in range(T-1):
            # go to next state
            s = np.random.choice(2,p=true_A[s])
            # get next observation
            o = np.random.choice(2,p=true_B[s])
            sequence.append(o)
        sequences.append(sequence)
    # our observations are 0s and 1s , 0 for H and 1 for T
    return sequences
````

````python
X = coin_toss(N=50)
````

````python
# we will attempt a more vectorised solution after we discuss scailing to counter the underflow problem
# for now for loops helps code look like equations :)

class HMM:

    def get_param(self,shape):
        # return a weight with each row summing up to 1
        w = np.random.random(shape)
        w = w/w.sum(axis=-1,keepdims=True)
        return w
    
    
    def forward(self,x):
        T = len(x)
        alpha = np.zeros((T,self.M))
        
        # initialisation step
        for i in range(self.M):
            alpha[0,i] = self.pi[i]*self.B[i,x[0]]
        
        # induction step , lets do this iteratively
        for t in range(T-1):
            for j in range(self.M):
                sum_ = 0
                for i in range(self.M):
                    sum_ += alpha[t,i]*self.A[i,j]
                alpha[t+1,j] = self.B[j,x[t+1]]*sum_
                
        # termination step
        p = alpha[T-1].sum()
        
        return alpha,p
    
    def backward(self,x):
        T = len(x)
        beta = np.zeros((T,self.M))
        
        # initialisation step
        for i in range(self.M):
            beta[T-1,i] = 1
            
        # induction step
        for t in range(T-2,-1,-1): # going backwards from second last element to first
            for i in range(self.M):
                beta_t_i = 0
                for j in range(self.M):
                    beta_t_i += self.A[i,j]*self.B[j,x[t+1]]*beta[t+1,j]
                beta[t,i] = beta_t_i

        return beta
        
    def viterbi(self,x): #  get best sequence of hidden states for observation sequence x
        M = self.M
        T = len(x)
        delta = np.zeros((T,M))
        # initialisation step
        for j in range(M):
            delta[0,j] = self.pi[j]*self.B[j,x[0]]
        # induction step

        for t in range(1,T):
            for j in range(M):
                delta[t,j] = np.max(delta[t-1,:]*self.A[:,j]) * self.B[j,x[t]] 
        
        # termination step : get the max probability
        # this will probably be zero , we will discuss the underflow problem later
        p = np.max(delta[T-1,:])
        
        # now time for Backtracking
        
        psi = np.zeros((T,self.M))
        
        # initialisation step
        for j in range(self.M):
            psi[0,j] = 0
        
        # induction step
        for t in range(1,T):
            for j in range(self.M):
                psi[t,j] = np.argmax(delta[t-1]*self.A[:,j])

        z = np.argmax(delta[T-1])
        
        Z = [] # sequence of best states
        
        Z.append(z)
        
        # backtrack and get best states
        
        for t in range(T-2,-1,-1): # from second last down to 0
            z = int(psi[t+1,z])
            Z.append(z)
            
        return [z for z in reversed(Z)] # reverse Z
    
    def fit(self,X,M,max_iter=30):
        N = len(X)
        V = K = len(set(obs for seq in X for obs in seq))
        
        self.M = M
        self.K = K
        
        # initialise parameters
        self.pi = self.get_param(M) 
        self.A = self.get_param((M,M))
        self.B = self.get_param((M,K))
        self.params = (self.pi.copy(),self.A.copy(),self.B.copy())
        costs = [] # log likelihoods
        
        for _ in range(max_iter):
            # now we will get the alpha , beta and P for each data sequence seperately
            alphas = []
            betas = []
            P = [] 
            for x in X:
                alpha,p = self.forward(x)
                beta = self.backward(x)
                
                alphas.append(alpha)
                betas.append(beta)
                P.append(p)
            
            # costs
            cost = np.sum(np.log(P))
            costs.append(cost)

            # then we can calculate the update of our parameters            

            # update pi

            for i in range(M):
                sum_ = 0
                for n in range(N):
                    sum_ += alphas[n][0,i]*betas[n][0,i]/P[n]
                self.pi[i] = sum_/N
            
            # update A
            
            for i in range(M):
                for j in range(M):
                    num = 0
                    denom = 0
                    for n in range(N):
                        T = len(X[n])
                        # numerator
                        sum_2 = 0
                        for t in range(T-1):
                            sum_2 += alphas[n][t,i]*self.A[i,j]*self.B[j,X[n][t+1]]*betas[n][t+1,j]
                        sum_1 = 1/P[n] * sum_2
                        num += sum_1
                        # denominator
                        sum_2 = 0
                        for t in range(T-1):
                            sum_2 += alphas[n][t,i]*betas[n][t,i]
                        sum_1 = 1/P[n] * sum_2
                        denom += sum_1
                    self.A[i,j] = num/denom
                    
            # update B
            for j in range(M):
                for k in range(K):
                    num = 0
                    denom = 0
                    for n in range(N):
                        T = len(X[n])
                        # numerator
                        sum_2 = 0
                        for t in range(T):
                            sum_2 += alphas[n][t,j]*betas[n][t,j] if X[n][t] == k else 0
                        sum_1 = sum_2/P[n]
                        num += sum_1
                        # denominator
                        sum_2 = 0
                        for t in range(T):
                            sum_2 += alphas[n][t,j]*betas[n][t,j]
                        sum_1 = 1/P[n] * sum_2
                        denom += sum_1
                
                    self.B[j,k] = num/denom
    

        print("pi: ",self.pi)
        print("A: ",self.A)
        print("B: ",self.B)
        print("final log-likelihood ",costs[-1])
        plt.plot(costs)
        plt.xlabel("iteration")
        plt.ylabel("log likelihood")
        plt.show()
````

````python
M=2
model = HMM()
model.fit(X,M)
````

````python
# lets calculate the log likelihood with original params
# also lets try the Viterbi Algorithm to get the most probable sequence
pi = np.array([0.5,0.5])
A = np.array([[0.1,0.9],[0.8,0.2]])
B = np.array([[0.6,0.4],[0.3,0.7]])

# we will use the functions inside model instead of rewriting them
model2 = HMM()
model2.M = M
model2.pi,model2.A,model2.B = pi,A,B

P = []
for x in X:
    _,p = model2.forward(x)
    P.append(p)

print("log-likelihood with true params ",np.log(P).sum())

Z = model2.viterbi(X[0])
print("For observation sequence ",X[0])
print("The best sequence of hidden states is ",Z)
````

<h1>Math</h1>

One problem that frequently arrises when we are working with HMMs is the underflow problem

Imagine we have a very long sequence , the forward algorithms says , that we should just keep multiplying $\alpha$ until the end of the sequence

But what happens when we keep multiplying small numbers together , they get even smaller and eventually go down to 0

Think of an audio signal , these are sampled 44100 HZ (44100 samples/second) , our HMM wont be that useful if it can only represent a few ms of data

How do we solvethis problem ?

---

<h4>Viterbi</h4>

Lets first consider the viterbi algorithm , this is an easy case since it only involves multiplication

we know that we can use the log probability instead of the actual probability because it is monotonically increasing

remember that $\log (AB) = \log A + \log B$

so viterbi is solved we just log everything

But what about the forward-backward Algorithm and the training algorithm which involve sums ?

---

<h4>Scaling Forward</h4>

we need to introduce the concept of scaling

To accomplish proper scaling , we are going to introduce a scale factor , $c(t)$ , t=1...T

with the scale factor we can restate the three steps of the forward Algorithm

<h4>Initialisation</h4>

$$\alpha'(1,i) = \pi_i B(i,x(1)) = \alpha(1,i)$$

so that the same expression we had before , the difference is that $\alpha'$ is a temporary variable

we use the prime variables to define the scale , which is just the sum of the prime variable over all the states at that time

$$c(t) = \sum^M_{i=1} \alpha'(t,i)$$

then we define $\hat\alpha$

$$\hat \alpha (t,i) = \alpha'(t,i)/c(t)$$

so for the initialisation step :

$$\hat \alpha(1,i) = \frac{\pi_i B(i,x(1))}{c(1)}$$

---

<h4>Indection step</h4>

Temporary variable

$$\alpha'(t,j) = \sum^M_{i=1} \hat \alpha (t-1,i) A(i,j) B(j,x(t))$$

so this is the same as before but the thing on the right gets hatted and the thing on the left gets primed

The scailing remains the same

$$\hat \alpha (t,i) = \alpha'(t,i)/c(t)$$

---

<h4>Termination</h4>

finally we can calculate the probability of the observation as follows :

$$p(x) = \prod^T_{t=1}c(t)$$

notice that this observation probability is also going to be a very small number , but it comes in the form of a product which means we can just calculate the log probability

---

<h4>Scailing backward</h4>

for this we dont have to calculate a new scale factor , we can just use the same one from before

<h4>Initialisation step</h4>

$$\hat \beta(T,i) = 1$$

<h4>Induction step</h4>

$$\hat \beta(t,i) = \frac{\sum\limits^M_{j=1} A(i,j)B(j,x(t+1))\hat \beta(t+1,j)}{c(t+1)}$$

So this is the old induction step scaled by $c(t+1)$

---

<h4>Idnetities</h4>

Using the previous equation we can prove the following three identities , which will be used to rewrite the update equations in terms of $\hat \alpha \ and \ \hat \beta$

$$\left( \prod_{\tau=1}^t c(\tau) \right) \hat \alpha(t,i) = \alpha (t,i)$$

$$\hat \alpha(t,i) \hat \beta(t,i) p(x) = \alpha(t,i) \beta(t,i)$$

$$\alpha(t,i)\beta(t+1,j) = \frac{p(x) \hat \alpha(t,i) \hat \beta(t+1,j)}{c(t+1)}$$

here is a quick guide for these proofs :

looking back at the equations for $\hat \alpha$ we see that the only difference from $\alpha$ is that we divide by $c(t)$ , $\alpha' $ doesnot change anything in the equations , its just a temporary variable used to calculate the scale $c$ , so for $\hat \alpha(t,i)$ we end up with the same expression as $\alpha(t,i)$ except that we divided $\hat \alpha(t,i)$ by $c(t),c(t-1),\ldots,c(1)$ , so we multiply by the inverse which is $\prod\limits_{\tau=1}^t c(\tau) = c(t),c(t-1),\ldots,c(t)$ to get back the $\alpha(t,i)$ , logically we divide $\alpha$ by scales to get $\hat \alpha$ so we need to multiply  $\hat \alpha$ by the scales to get back $\alpha$

The second one follows the same concept  except that for $\hat \beta$s we divide by scales for $c(T),c(T-1),\ldots,c(t+1)$ , so to unscale $\hat \alpha$ we need to multiply RHS by $c(1),\ldots,c(t)$ , to unscale $\hat \beta$ we need to multiply RHS by $c(t+1),\ldots,c(T)$ , so we need to multiply RHS by $c(1),\ldots,c(T)$ , that $\prod\limits_{t=1}^T c(t)$ which is the expression for $p(x)$

In the third equation since $\hat \beta$ is at time $t+1$ we exclude the scale $c(t+1)$ , other than this , this is the same as the second equation 

---

<h4>New updates</h4>

$$\pi_i = \frac{1}{N} \sum_{n=1}^{N} \hat \alpha_n(1,i) \hat \beta_n(1,i)$$

same as before but the probability $\left( \frac{1}{P(n)} \right)$ goes away (identity 2)

$$\large A(i,j) = \frac{\sum\limits^N_{n=1}\sum\limits^{T(n)-1}_{t=1} \frac{\hat \alpha_n (t,i) \hat \beta_n (t+1,j)}{c(t+1)}A(i,j)B(j,x_n(t+1))}{\sum\limits^N_{n=1} \sum\limits^{T(n)-1}_{t=1} \hat \alpha_n(t,i) \hat \beta_n (t,i)}$$

Also very similar to before (identity 3 for numerator and 2 for denominator)

$$\large B(j,k) = \frac{\sum\limits^N_{n=1} \sum\limits^{T(n)}_{t=1} \hat \alpha_n(t,j) \hat \beta_n(t,j) \ if \ x_n(t) = k, \ else \ 0}{\sum\limits^N_{n=1}\sum\limits^{T(n)}_{t=1} \hat \alpha_n(t,j) \hat \beta_n(t,j)}$$

Again very similar (identity 2)

<h1>code</h1>

````python
class HMMS:

    def get_param(self,shape):
        # return a weight with each row summing up to 1
        w = np.random.random(shape)
        w = w/w.sum(axis=-1,keepdims=True)
        return w
    
    def forward(self,x):
        T = len(x)
        alpha_prime = np.zeros((T,self.M))
        alpha_hat = np.zeros((T,self.M))
        c = np.zeros(T)
        
        # initialisation step
        alpha_prime[0,:] = self.pi[:]*self.B[:,x[0]]
        c[0] = alpha_prime[0,:].sum()
        alpha_hat[0,:] = alpha_prime[0,:] / c[0]
        
        #induction step
        
        for t in range(1,T):
            alpha_prime[t,:] = alpha_hat[t-1,:]@(self.A*self.B[:,x[t]])
            c[t] = alpha_prime[t,:].sum()
            alpha_hat[t,:] = alpha_prime[t,:]/c[t]
        
        # termination step , log probabilities , we dont need these
        p = np.log(c).sum()
        
        return alpha_hat,c
    
    def backward(self,x,c):
        T = len(x)
        beta_hat = np.zeros((T,self.M))
        
        # initialisation step
        beta_hat[T-1,:] = 1
        
        # induction step
        for t in range(T-2,-1,-1):
            beta_hat[t,:] = self.A@(self.B[:,x[t+1]]*beta_hat[t+1,:])/c[t+1]
        
        return beta_hat
        
    def viterbi(self,x): #  get best sequence of hidden states for observation sequence x
        # use log probabilities instead of actual probabilities
        # add log probabilities instead of multiplying actual ones
        
        M = self.M
        T = len(x)
        delta = np.zeros((T,M))
        # initialisation step
        for j in range(M):
            delta[0,j] = np.log(self.pi[j])+np.log(self.B[j,x[0]])
        # induction step

        for t in range(1,T):
            for j in range(M):
                delta[t,j] = np.max(delta[t-1,:]+np.log(self.A[:,j])) + np.log(self.B[j,x[t]]) 
        
        # termination step : get the max probability
        p = np.max(delta[T-1,:])
        
        # now time for Backtracking
        
        psi = np.zeros((T,self.M))
        
        # initialisation step
        for j in range(self.M):
            psi[0,j] = 0
        
        # induction step
        for t in range(1,T):
            for j in range(self.M):
                psi[t,j] = np.argmax(delta[t-1]+np.log(self.A[:,j]))

        z = np.argmax(delta[T-1])
        
        Z = [] # sequence of best states
        
        Z.append(z)
        
        # backtrack and get best states
        
        for t in range(T-2,-1,-1): # from second last down to 0
            z = int(psi[t+1,z])
            Z.append(z)
            
        return [z for z in reversed(Z)] # reverse Z
    
    def fit(self,X,M,max_iter=30):
        N = len(X)
        V = K = len(set(obs for seq in X for obs in seq))
        
        self.M = M
        self.K = K
        
        # initialise parameters
        self.pi = self.get_param(M) 
        self.A = self.get_param((M,M))
        self.B = self.get_param((M,K))
        costs = [] # log likelihoods
        
        for _ in range(max_iter):
            alpha_hats = []
            beta_hats = []
            C = [] # scales , we need these for A update
            # we no longer need the probabilities
            for x in X:
                alpha_hat,c = self.forward(x)
                beta_hat = self.backward(x,c)

                alpha_hats.append(alpha_hat)
                beta_hats.append(beta_hat)
                C.append(c)
        
            # costs
            cost = np.log(C).sum() 
            costs.append(cost)
            
            # update pi
            pi = np.zeros(M)
            for n in range(N):
                pi += alpha_hats[n][0,:]*beta_hats[n][0,:]
            
            self.pi = pi/N

            # update A
            den_A = np.zeros(M)
            num_A = np.zeros((M,M))
            for n in range(N):
                x = X[n]
                T = len(x)
                den_A += (alpha_hats[n][:T-1]*beta_hats[n][:T-1]).sum(axis=0)
                
                for i in range(M):
                    for j in range(M):
                        for t in range(T-1):
                            num_A[i,j] += alpha_hats[n][t,i]*beta_hats[n][t+1,j]*self.A[i,j]*self.B[j,x[t+1]]/C[n][t+1]
            self.A = num_A / den_A[:,np.newaxis]
        
            #update B
            den_B = np.zeros(M)
            num_B = np.zeros((M,K))
            for n in range(N):
                T = len(x)
                den_B += (alpha_hats[n][:T]*beta_hats[n][:T]).sum(axis=0)
                prod = alpha_hats[n]*beta_hats[n]
                for k in range(K):
                    num_B[:,k] += prod[np.array(X[n])==k].sum(axis=0)
            self.B = num_B/den_B[:,np.newaxis]
            
        print("pi: ",self.pi)
        print("A: ",self.A)
        print("B: ",self.B)
        print("final log-likelihood ",costs[-1])
        plt.plot(costs)
        plt.xlabel("iteration")
        plt.ylabel("log likelihood")
        plt.show()
````

````python
M=2
model = HMMS()
model.fit(X,M)
````

````python
# lets calculate the log likelihood with original params
# also lets try the Viterbi Algorithm to get the most probable sequence
pi = np.array([0.5,0.5])
A = np.array([[0.1,0.9],[0.8,0.2]])
B = np.array([[0.6,0.4],[0.3,0.7]])

# we will use the functions inside model instead of rewriting them
model2 = HMMS()
model2.M = M
model2.pi,model2.A,model2.B = pi,A,B

P = []
for x in X:
    _,p = model2.forward(x)
    P.append(p)

print("log-likelihood with true params ",np.log(P).sum())

Z = model2.viterbi(X[0])
print("For observation sequence ",X[0])
print("The best sequence of hidden states is ",Z)
````

<h1>Math</h1>

Next we will use HMMs for POS-tagging , recall that this is an example of when we can use counting

---

<h4>Data</h4>

The data can be found <a href="https://www.clips.uantwerpen.be/conll2000/chunking/">here</a> , we are interested in train.txt and test.txt

Here is an example of the data:

```
Confidence NN B-NP

in IN B-PP

the DT B-NP

pound NN I-NP

is VBZ B-VP

widely RB I-VP

expected VBN I-VP
```

each token is on a seperate line and the tag is beside it (we are interested only in the first 2 columns)

---

<h4>Baseline</h4>

A good thing to do when faced with a machnie learning problem is to find a baseline

we can use logistic regression , just one hot encode the words and the tags and use softmax

with just that we can obtain over 90% accuracy (that means 90% of the time the word has the same tag)

its very rare that one word can be used in two different ways , for ex:

<ul>
    <li>I drank milk this morning</li>
    <li>This Company is milking its customers' money</li>
</ul>

---

<h4>F1</h4>

we will also calcualte the F1 score , which is  the harmonic mean of the precision and recall which are common measures in information retreival and NLP

$$F1 = \frac{2 \times (precision \times recall)}{precision + recall}$$

$$precision = \frac{TP}{TP + FP}$$

$$recall = \frac{TP}{TP + FN}$$

---

<h4>Sequences</h4>

Can we use the fact that these words show in a sequence and therefore can we use the context to help us make better prediciton ?

A great tool for that might be RNNs (next notebook)

As for an HMM , we can think of the words as the observed sequence adn the tags as the sequence of hidden states

Again , here we are in an interesting situation where we actually know the hidden states of the training data , its one of these cases where the hidden states are not some abstract hyperparamater but an actual real thing , so we dont need the expectation maximisation algorithm , all the parameters can be directly calculated using maximum likelihood directly

HMM has three parameters:

<ul>
    <li>$P_i = $ frequency of start tags</li>
    <li>$A = p(tag(t)|tag(t-1))$</li>
    <li>$B = p(word(t)|tag(t))$</li>
</ul>

All these can be calculated by counting

<h1>code</h1>

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
````

````python
data_train = pd.read_csv("datasets/chunking/train.txt", sep=" ",header=None)
data_test = pd.read_csv("datasets/chunking/test.txt", sep=" ",header=None)
data_train.head()
````

````python
# merge both sets so we can one hot encode
Ntrain = len(data_train)
data = pd.concat([data_train,data_test])

X = pd.get_dummies(data.iloc[:,0]).to_numpy()
Y = pd.get_dummies(data.iloc[:,1]).to_numpy()

# we need Y to be targets , not one hot encoded
Y = Y.argmax(axis=1)
# a one hot encode X matrix is toooo slow
X = X.argmax(axis=1).reshape(len(X),1)

# now lets resplit data
Xtrain,Ytrain,Xtest,Ytest = X[:Ntrain],Y[:Ntrain],X[Ntrain:],Y[Ntrain:]
````

````python
# logistic regression is also too slow
# so we will use DescisionTree
model = DecisionTreeClassifier()
model.fit(Xtrain,Ytrain)
print("Decision Tree train accuracy : ",model.score(Xtrain,Ytrain))
print("Decision Tree test accuracy : ",model.score(Xtest,Ytest))
````

````python
# pandas skipped empty lines which represented the end of a sentence
# we need to re read the files to know where each sentence begins and ends

train_end = []
test_end = []

for i,line in enumerate(open("datasets/chunking/train.txt")):
    if line == '\n':
        train_end.append(i)


for i,line in enumerate(open("datasets/chunking/test.txt")):
    if line == '\n':
        test_end.append(i)
        


train_end =  np.array(train_end) - np.arange(len(train_end))
train_end = [0] + list(train_end) 

test_end = [0] + np.array(test_end) - np.arange(len(test_end))
test_end = [0] + list(test_end) 
````

````python
# now lets split Xs and Ys into sequnces of observations and states
Otrain = [] # trainining observation sequences
Ttrain = [] # trainining target sequences 

for i in range(len(train_end)-1):
    Otrain.append(Xtrain[train_end[i] : train_end[i+1]])
    Ttrain.append(Ytrain[train_end[i] : train_end[i+1]])
    
Otest = [] 
Ttest = []  

for i in range(len(test_end)-1):
    Otest.append(Xtest[test_end[i] : test_end[i+1]])
    Ttest.append(Ytest[test_end[i] : test_end[i+1]])
````

````python
# now we can start counting to estimate our parameters
# the number of hidden states M is the number of POS tags 
M = Y.max()+1
# K is the number of words we observed
K = X.max()+1

smoothing = 10e-2
````

````python
# lets start with pi , we will be adding smoothing too
pi = np.ones(M)*smoothing

for t in Ttrain:
    pi[t[0]] += 1

pi = pi/pi.sum()
````

````python
# next is A
A = np.ones((M,M))*smoothing
for tags in Ttrain:
    for i in range(len(tags)-1):
        tag_1 = tags[i]
        tag_2 = tags[i+1]
        A[tag_1,tag_2] +=1
        
A = A/A.sum(axis=1,keepdims=True)
````

````python
# next iis B
B = np.ones((M,K))*smoothing

for tag,obs in zip(Ttrain,Otrain):
    for t,o in zip(tag,obs):
        B[t,o]+=1
        
B = B/B.sum(axis=1,keepdims=True)
````

````python
# now we will create a HMM and feed in our parameters
# then for each test sequence we will use viterbi t calculate the sequence of hidden states with highest probability
# and this will be our prediction
model = HMMS()
model.M=M
model.pi = pi
model.A = A
model.B = B
````

````python
# get predictions
Ptrain = []
Ptest = []
for obs in Otrain:
    Ptrain.append(model.viterbi(obs))
    
for obs in Otest:
    Ptest.append(model.viterbi(obs))
````

````python
# now lets write a function to calculate accuracy
def accuracy(T,P): # both are sequences
    T = np.array(T)
    P = np.array(P)
    correct = 0
    count = 0
    for t,p in zip(T,P):
        correct += np.sum(p==t)
        count += len(t)
    return correct/count
````

````python
print("Train accuracy ",accuracy(Ttrain,Ptrain))
print("Test accuracy ",accuracy(Ttest,Ptest))
````

<h1>Math</h1>

Now we want to implement a HMM with gradient descent , we will be using tensorflow for this example

one of the challenges of this approach is that gradient descent doesnot conform to any constraints , it just goes into the direction of the gradient

This presents a problem for us since the paramaters of a HMM , $\pi,A,B$ , are all probabilities

$\pi$ must sum up to 1 and $A,B$ must sum up to 1 along the rows

one way to counter this is by renormalising the parameters after each gradient descent  step , however , this is not "real" gradient descent because by renormalising we are not exactly going in the direction of the gradient anymore

recall the softmax function :

$$softmax(x_i) = \frac{\exp(x_i)}{\sum\limits^K_{k=1}\exp(x_k)}$$

now take $\pi$ as an example , suppose we want to parametrise $\pi$ using the symbol $\theta$ , we can assign $\pi = softmax(\theta)$ , in this way $\pi$ is like an intermediate variable , and  theta is the actual parameter we will be updating

This ensure that $\pi$ is always between 0 and 1 and sums to 1 , at the same time the values of $\theta$ can be anything , so we can freely use gradient descent on $\theta$ without having to worry on any constraints

same goes for $A,B$ , these are matricies and we want their rows to sum up to 1 , we can also apply softmax on a matrix ,  and each row will be a seperate probability distribution

<h1>code</h1>

````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import Sequence
from tensorflow.keras import Model
from tensorflow.nn import softmax
from tensorflow.keras.optimizers import Adam,SGD
````

````python
def coin_toss(N=50,T=30):
    true_pi = np.array([0.5,0.5])
    true_A = np.array([[0.1,0.9],[0.8,0.2]])
    true_B = np.array([[0.6,0.4],[0.3,0.7]])
    sequences = []
    for _ in range(N):
        sequence = []
        s = np.random.choice(2,p=true_pi)
        o = np.random.choice(2,p=true_B[s])
        sequence.append(o)
        for _ in range(T-1):
            s = np.random.choice(2,p=true_A[s])
            o = np.random.choice(2,p=true_B[s])
            sequence.append(o)
        sequences.append(sequence)
    return sequences
````

````python
X = coin_toss(N=50)
M = 2
K = 2
````

````python
# now lets implement the model
# all we need to do is apply the forward algorithms and get the scales
# then we calcualte the cost , this the -ve log likelihood ,so we take the -ve log of the sccales
# then we leave tensorflow to do gradient descent for us

class HMMSST(Model): # HMM + Scaled + Softmax +Tensorflow
    def __init__(self,M,K):
        super(HMMSST, self).__init__()
        self.M = M
        self.K = K
        # these are not the true pi,A,B , but rather the ones we optimise
        self.pi_opt = tf.Variable(np.ones(M)/M,name="pi")
        self.A_opt = tf.Variable(np.random.randn(M,M),name="A")
        self.B_opt = tf.Variable(np.random.randn(M,K),name="B")
        
    
    def softmax(self,a):
        return np.exp(a)/np.exp(a).sum()

    @tf.function
    def call(self,x): # here we get one x sample and we return the log_prob
        log_cost = 0
        
        pi,A,B = softmax(self.pi_opt),softmax(self.A_opt),softmax(self.B_opt)
        T = len(x)
        
        alpha_prime = pi*B[:,x[0]]
        c = tf.reduce_sum(alpha_prime)
        log_cost += tf.math.log(c)
        alpha_hat = alpha_prime/c
        
        for t in range(1,T):
            alpha_prime = tf.matmul(alpha_hat[None,:],A) * B[:,x[t]]
            c = tf.reduce_sum(alpha_prime)
            log_cost += tf.math.log(c)
            alpha_hat = alpha_prime/c
            
        return log_cost

    def fit(self,X,epochs=10,lr=1e-3,plot=True):
        opt = Adam(learning_rate=lr)
        costs = []
        for epoch in range(epochs):
            cost = 0
            for x in X:
                with tf.GradientTape() as tape:
                    loss_value = -self(x, training=True)  
                    cost += loss_value
                    
                grads = tape.gradient(loss_value, self.trainable_weights)
                opt.apply_gradients(zip(grads, self.trainable_weights))
        
            costs.append(cost) # total cost for all samples
            print("epoch: ",epoch+1,'cost: ',cost.numpy())
        if plot:
            plt.plot(costs)
            plt.show()
````

````python
model = HMMSST(M,K)
model.fit(X,lr=1e-2,epochs=5)
````

````python
# now lets implement the model
# all we need to do is apply the forward algorithms and get the scales
# then we calcualte the cost , this the -ve log likelihood ,so we take the -ve log of the sccales
# then we leave tensorflow to do gradient descent for us

class HMMSNT(Model): # HMM + Scaled + Normalised +Tensorflow
    def __init__(self,M,K):
        super(HMMSNT, self).__init__()
        self.M = M
        self.K = K
        self.pi = tf.Variable(np.ones(M)/M,name="pi")  
        self.A = tf.Variable(self.get_param((M,M)),name="A")
        self.B = tf.Variable(self.get_param((M,K)),name="B")
        
    def get_param(self,shape):
        w = np.random.random(shape)
        w = w/w.sum(axis=-1,keepdims=True)
        return w
    
    def softmax(self,a):
        return np.exp(a)/np.exp(a).sum()

    @tf.function
    def call(self,x): # here we get one x sample and we return the log_prob
        log_cost = 0
        T = len(x)
        
        alpha_prime = self.pi*self.B[:,x[0]]
        c = tf.reduce_sum(alpha_prime)
        log_cost += tf.math.log(c)
        alpha_hat = alpha_prime/c
        
        for t in range(1,T):
            alpha_prime = tf.matmul(alpha_hat[None,:],self.A) * self.B[:,x[t]]
            c = tf.reduce_sum(alpha_prime)
            log_cost += tf.math.log(c)
            alpha_hat = alpha_prime/c
            
        return log_cost

    def fit(self,X,epochs=10,lr=1e-3,plot=True):
        #opt = Adam(learning_rate=lr)
        costs = []
        for epoch in range(epochs):
            cost = 0
            for x in X:
                with tf.GradientTape() as tape:
                    loss_value = -self(x, training=True)  
                    cost += loss_value
                    
                grads = tape.gradient(loss_value, self.trainable_weights)
                # now apply plain gradient descent + normalisation 
                self.pi.assign((self.pi-lr*grads[0])/tf.reduce_sum(self.pi))
                if grads[1] is not None:
                    self.A.assign((self.A-lr*grads[1])/tf.reduce_sum(self.A,axis=1,keepdims=True))
                self.B.assign((self.B-lr*grads[2])/tf.reduce_sum(self.B,axis=1,keepdims=True))
            costs.append(cost) # total cost for all samples
            print("epoch: ",epoch+1,'cost: ',cost.numpy())
        if plot:
            plt.plot(costs)
            plt.show()
````

````python
model = HMMSNT(M,K)
model.fit(X,lr=1e-4,epochs=5)
````

<h4>Generative vs Discriminative</h4>

we have discussed before that even though HMM is an unsupervised model , since it models a probability , we can turn it around and use it as a classifier using Bayes rule , these are called <strong>Generative</strong> models

Generative models are named as such because they take the probability density of a class and pretends it generates the data it observes , so these models model $P(X|Y)$

Compare this to <h4>Discriminative</h4> models like logisic regression and Deep neural networks ,  these models give us the probability of the class given the input directly ,they model $P(Y|X)$ , so there is no modelling of the likelihood probability

<img src="extras/26.1.PNG" width="400"></img>

<h4>Intersection</h4>

sometimes these two types of models intersect

In logistic regression , we saw how we cant solce for the weights directly , we needed a method like gradient descent

But we also saw that there is an exception too this rule , if we assume that each class has the same covariance but different means and they are both Gaussians , in this case we can solve for the weights directly (LDA)

If we know that both classes are Gaussians and we know there means and covariances , well , then we know their likelihoods , $p(X|Y) = N(X,\mu,\sigma)$, so this is a generative model also

In this particular setup , generative and discriminative models intersect

Discriminative models offer superior accuracy  but sometimes it is hard to interpret what is being learned

````python
# we are going to try to differentiate between poem by Robert Forst and Edgar Allan
# instead of having  sequence of words we will use a sequence of POS tags as our observations
# then we will use our HMM to get the likelihood
# and use Bayes rule to calculate the probability on which we make our prediction
# lets start by implelemnting our model

class HMMC: #HMM + Classifier
    def __init__(self,M,V):
        self.M = M
        self.V = V # number of POS tags in observations
    
    def fit(self,X,Y,lr=1e-5,epochs=10):
        # now we are going to create a seperate model for each class
        # and we train that model for observations of that class only
        # so if we have K classes we have K models
        self.K = max(Y)+1 # number of classes
        self.models = []
        self.priors = []
        for k in range(self.K):
            print("training model for class: ",k)
            X_k = X[np.array(Y)==k]
            model = HMMSST(self.M,self.V)
            model.fit(X_k,plot=True,epochs=epochs,lr=lr) # fit only on data that belongs to that class
            self.models.append(model)
            self.priors.append(len(X_k)/len(X))
            print("finished training model for class: ",k)
            
    def predict(self,X):
        probs = np.zeros((len(X),self.K))
        for i,x in enumerate(X):
            for k in range(self.K):
                log_likelihood = self.models[k](x).numpy()
                log_prior = np.log(self.priors[k])
                probs[i][k] = log_likelihood + log_prior
        return np.argmax(probs,axis=1)
    
    def score(self,X,Y):
        P = self.predict(X)
        return np.mean(P==Y)
        
````

<h1>Math</h1>

Finally , we want to look at how to use HMM for real-valued observations , this means that what we observe is a number on a scale rather than a symbol like H/T or words

This leads us to consider what is continuous and what is discrete and what we are approximating as discrete that is really continuous

Think of an audio signal , audio is just sound and sound is just air pressure that vibrates and by vibrating we mean that it is oscillating in time

when an audi is stored on a computer , obviously our computer cant store an infinite number of values , so the air pressure has to be quantised , it has to be given a number that the computer can represent

Now there are two things to quantise , the amplitude of the sound pressure and when that amplitude occured , so time itself has to be quantised

we mentioned before that sound is usually sampled at 44.1 KHz which is way too fast for the brain to realise is not continuous

So what happens when we use a HMM ? 

what we assume is that the observed variable is a mixture of Gaussians , that like saying , well one of the things we beleive is continuous we are going to treat asa continuous (ex:amplitude) , as of the other thing we are going to treat as discrete (time)

---

<h4>GMMs with HMMs</h4>

so what changes when we use HMMs with GMMS ?

remember that a HMM is defined by three things : $\pi,A,B$

it seems B is what has to change ,  since it used to be an MxV matrix (V is vocabulary) , but now there are no longer just V possible symbols

recall the three things we need for a GMM , R (responsibilities) , which tells us the probability of a specific Gaussian , and then for each individual Gaussian we need a $\mu,\Sigma$ , so we are replacing B by three new parameters , $R,\mu,\Sigma$ , we use $K$ to be the number of Gaussians

---

we will sotre R as and MxK matrix , so there is one row for each state , and then for each state there is K different probabilities , since this is a probability matrix each row has to sum to 1

each individual $\mu$ is D dimensional and there are K of them , and we need that many for the M states , sp $\mu$ will be an MxKxD matrix

similarly , $\Sigma$ will be MxKxDxD , a 4-dimensional matrix

---

so once we have chosen a state , call that j , our observation probability becomes :

$$B(j,t) = \sum\limits^K_{k=1}R(j,k)N \left( x(t),\mu(j,k),\Sigma(j,k) \right)$$

N here is the PDF for the multivariate Gaussian

now even though we dont technically have a B matrix anymore , we can still make one in this fashion just by calculating it for each sequence

lets also calculate the individual components too without doing the sum

$$comp(j,k,t) = R(j,k)N \left( x(t),\mu(j,k),\Sigma(j,k) \right)$$

---

<h4>E-step</h4>

now that we these two thing we can calculate a new $\gamma$ which is part of our E-step in the Expectation maximisation algorithm

$$\gamma(j,k,t) = \frac{\alpha(t,j)\beta(t,j)}{\sum\limits^M_{j'=1}\alpha(t,j')\beta(t,j')}\frac{R(j,k)N(x(t),\mu(j,k),\Sigma(j,k))}{\sum\limits^K_{k'=1}R(j,k')N(x(t),\mu(j,k'),\Sigma(j,k'))}$$

This can be rewritten as

$$\gamma(j,k,t) = \frac{\alpha(t,j)\beta(t,j)}{\sum\limits^M_{j'=1}\alpha(t,j')\beta(t,j')}\frac{Comp(j,k,t)}{B(j,t)}$$

once we have $\gamma$ we can define the updates for $R,\mu,sigma$

---

<h4>M-step</h4>

$$R(j,k) = \frac{\sum\limits^T_{t=1}\gamma(j,k,t)}{\sum\limits^T_{t=1}\sum^K_{k'=1}\gamma(j,k',t)}$$

$$\mu(j,k) = \frac{\sum\limits^T_{t=1}\gamma(j,k,t)x(t)}{\sum\limits^T_{t=1} \gamma(j,k,t)}$$

$$\Sigma(j,k) = \frac{\sum\limits^T_{t=1}\gamma(j,k,t) \left(x(t)-\mu(j,k) \right) \left( x(t) - \mu(j,k) \right)^T }{\sum\limits^T_{t=1}\gamma(j,k,t)}$$

````python
# we will follow the same pattern here
# first we need a function to generate contiinuous observations
# then we need to implement the model and take a look at our likelihood
# and finally we re-implement the model using gradient descent with tensorflow
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
````

````python
def init_1():
    M = 1
    K = 1
    D = 1
    pi = np.array([1])
    A = np.array([[1]])
    R = np.array([[1]])
    mu = np.array([[[0]]])
    sigma = np.array([[[[1]]]])
    return M,K,D,pi,A,R,mu,sigma

def init_2():
    M = 5
    K = 3
    D = 2
    
    pi = np.array([1,0,0,0,0])
    
    A = np.array([
        [0.9,0.025,0.025,0.025,0.025],
        [0.025,0.9,0.025,0.025,0.025],
        [0.025,0.025,0.9,0.025,0.025],
        [0.025,0.025,0.025,0.9,0.025],
        [0.025,0.025,0.025,0.025,0.9]
    ])
    
    R = np.ones((M,K))/K
    
    mu = np.array([
        [[0,0],[1,1],[2,2]],
        [[5,5],[6,6],[7,7]],
        [[10,10],[11,11],[12,12]],
        [[15,15],[16,16],[17,17]],
        [[20,20],[21,21],[22,22]]
    ])
    
    sigma = np.zeros((M,K,D,D))
    
    # circular gaussians
    
    for m in range(M):
        for k in range(K):
            sigma[m,k] = np.eye(D)

    return M,K,D,pi,A,R,mu,sigma
````

````python
# lets write a function to generate continuous signals
def generate_signals(N=20,T=100,init=init_1):
    
    M,K,D,pi,A,R,mu,sigma = init()
    signals = np.zeros((N,T,D))    
    for n in range(N):
        m = np.random.choice(M,p=pi)
        # sample mixture
        k = np.random.choice(K,p=R[m])
        # sample x from this mixture
        x = np.random.multivariate_normal(mu[m,k],sigma[m,k])
        signals[n][0] = x
        for t in range(1,T):
            m = np.random.choice(M,p=A[m])
            k = np.random.choice(K,p=R[m])
            x = np.random.multivariate_normal(mu[m,k],sigma[m,k])
            signals[n][t]=x
    return signals
````

````python
# now lets implement the model

class HMM_GMM: # running out of names :(
    
    def viterbi(self,x): #  get best sequence of hidden states for observation sequence x
        M = self.M
        T = len(x)
        delta = np.zeros((T,M))
        # initialisation step
        for j in range(M):
            delta[0,j] = self.pi[j]*self.B(j,x[0])
        # induction step

        for t in range(1,T):
            for j in range(M):
                delta[t,j] = np.max(delta[t-1,:]*self.A[:,j]) * self.B(j,x[t]) 
        
        # termination step : get the max probability
        # this will probably be zero , we will discuss the underflow problem later
        p = np.max(delta[T-1,:])
        
        # now time for Backtracking
        
        psi = np.zeros((T,self.M))
        
        # initialisation step
        for j in range(self.M):
            psi[0,j] = 0
        
        # induction step
        for t in range(1,T):
            for j in range(self.M):
                psi[t,j] = np.argmax(delta[t-1]*self.A[:,j])

        z = np.argmax(delta[T-1])
        
        Z = [] # sequence of best states
        
        Z.append(z)
        
        # backtrack and get best states
        
        for t in range(T-2,-1,-1): # from second last down to 0
            z = int(psi[t+1,z])
            Z.append(z)
            
        return [z for z in reversed(Z)] # reverse Z

        
    def B(self,j,t): # return the probability of t when in state j
        B = 0
        for k in range(self.K):
            g = multivariate_normal(mean=self.mu[j,k], cov=self.sigma[j,k]+ np.eye(self.D)*self.eps)
            N = g.pdf(t)
            B += self.R[j,k]*N
        return B
    
    def comp(self,j,k,t):
        g = multivariate_normal(mean=self.mu[j,k], cov=self.sigma[j,k] + np.eye(self.D)*self.eps)
        N = g.pdf(t)
        return self.R[j,k]*N
        
        
    def forward(self,x):
        T = len(x)
        alpha = np.zeros((T,self.M))
        
        # initialisation step
        for i in range(self.M):
            alpha[0,i] = self.pi[i]*self.B(i,x[0])
        
        # induction step 
        for t in range(T-1):
            for j in range(self.M):
                sum_ = alpha[t,:]@self.A[:,j]            
                alpha[t+1,j] = self.B(j,x[t+1])*sum_
                
        # termination step
        p = alpha[T-1].sum()
        
        return alpha,p
     
    def backward(self,x):
        T = len(x)
        beta = np.zeros((T,self.M))
        
        # initialisation step
        for i in range(self.M):
            beta[T-1,i] = 1
            
        # induction step
        for t in range(T-2,-1,-1): # going backwards from second last element to first
            for i in range(self.M):
                beta_t_i = 0
                for j in range(self.M):
                    beta_t_i += self.A[i,j]*self.B(j,x[t+1])*beta[t+1,j]
                beta[t,i] = beta_t_i

        return beta

        
    def get_param(self,shape):
        # return a weight with each row summing up to 1
        w = np.random.random(shape)
        w = w/w.sum(axis=-1,keepdims=True)
        return w
    
    def fit(self,X,M,K,max_iter=30,eps=1e-5):
        self.eps = eps
        # eps would be used to guarantee we dont divide by 0
        # since in the equation of gamma we divide by B(j,t) which may be 0
        # also we want to guarantee that cov matrix is not singular when calculating pdf
        N = len(X)
        D = len(X[0][0])
        self.M = M
        self.K = K
        self.D = D
        
        # initialise parameters
        self.pi = self.get_param(M) 
        self.A = self.get_param((M,M))
        
        # B is replaced by 3 parameters
        self.R = self.get_param((M,K))
        # mu and sigma dont need to sum to 1
        self.mu = np.zeros((M,K,D))
        # we set mu to a similar way we do with Kmeans
        for m in range(M):
            for k in range(K):
                n = np.random.choice(N)
                random_x = X[n]
                random_idx = np.random.choice(len(random_x))
                random_point = np.random.choice(random_x[random_idx])
                self.mu[m,k] = random_point
            
        self.sigma = np.zeros((M,K,D,D))
        # set the covariance to a circular Gaussian
        for m in range(M):
            for k in range(K):
                self.sigma[m,k] = np.eye(D)
        
        # same strategy as before
        # we loop over each x and get all alphas and betas
        # then we use these to update our weights
        costs = []
        for i in range(max_iter):
            alphas = []
            P = []
            betas = []
            for x in X:
                # now calculate alphas and betas
                alpha,p = self.forward(x)
                beta = self.backward(x)
                
                alphas.append(alpha)
                P.append(p)
                betas.append(beta)
            # calculate log likelihood
            cost = np.sum(np.log(P))
            costs.append(cost)
            # then we calculate updates for parameters
            # pi

            self.pi = np.sum([(alphas[n][0,:]*betas[n][0,:])/P[n] for n in range(N)],axis=0)/N

            # A
            den = np.zeros(M)
            num = np.zeros((M,M))
                        
            for n in range(N):
                x = X[n]
                T =len(x)
                            
                den += (alphas[n][:T-1]*betas[n][:T-1]).sum(axis=0)/P[n]
                for i in range(M):
                    for j in range(M):
                        for t in range(T-1):
                            num[i,j] += alphas[n][t,i]*self.A[i,j]*self.B(j,x[t+1])*betas[n][t+1,j]/P[n]
            self.A = num/den[:,np.newaxis]

            # R,mu,sigma all need gamma so lets make them in one step
            num_R = np.zeros((M,K))
            den_R = np.zeros((M))
            
            num_mu = np.zeros((M,K,D))
            den_mu = np.zeros((M,K))
            
            num_sigma = np.zeros((M,K,D,D))
            den_sigma = np.zeros((M,K))
            
            
            for n in range(N):
                x = X[n]
                T = len(x)
                
                gamma = np.zeros((M,K,T))
                for j in range(M):
                    for k in range(K):
                        # first we calculate gamma(j,k,t)
                        for t in range(T):
                            num = alphas[n][t,j]*betas[n][t,j]*self.comp(j,k,x[t])
                            denom = ((alphas[n][t]@betas[n][t])*(self.B(j,x[t])+self.eps))
                            gamma[j,k,t] = num/denom 
                            num_mu[j,k] += gamma[j,k,t]*x[t]
                            num_sigma[j,k] += gamma[j,k,t]*((x[t]-self.mu[j,k])@(x[t]-self.mu[j,k]).T)
                    
                        den_mu[j,k] += gamma[j,k].sum()
                        
                        den_sigma += gamma[j,k].sum()
                        
                        num_R[j,k] += gamma[j,k].sum()
                    
                    den_R[j] += gamma[j].sum()
            
            self.R = num_R/den_R[:,np.newaxis] 
            self.mu = num_mu / den_mu[:,:,np.newaxis]
            self.sigma = num_sigma / den_sigma[:,:,np.newaxis,np.newaxis]
            ################################
        print("pi: ",self.pi)
        print("A: ",self.A)
        print("R: ",self.R)
        print("mean: ",self.mu)
        print("covariance: ",self.sigma)
        print("intial log-likelihood ",costs[0])
        print("final log-likelihood ",costs[-1])
        plt.plot(costs)
        plt.xlabel("iteration")
        plt.ylabel("log likelihood")
        plt.show()
````

````python
# now again with tensorflow
# we will use the softmax strategy
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.nn import softmax
from tensorflow.keras.optimizers import Adam,SGD
import tensorflow_probability as tfp
mvn = tfp.distributions.MultivariateNormalDiag
````

````python
# we will use the softmax strategy
class HMM_GMM_T(Model): # HMM + GMM +  +Tensorflow
#    def __init__(self):
#        super(HMM_GMM_T, self).__init__()
        
    
    @tf.function
    def call(self,x): # here we get one x sample and we return the log_prob
        # only pi,A,R must follow the contraint of summing to1
        pi,A,R,mu,sigma = softmax(self.pi_opt),softmax(self.A_opt),softmax(self.R_opt),self.mu_opt,tf.exp(self.log_sigma_opt)
        # calculate B
        T = len(x)
        
        B = []
        for j in range(self.M):
            B.append([])
            for t in range(T):
                B[j].append(0)
                for k in range(self.K):
                    g = mvn(loc=mu[j,k], scale_diag=sigma[j,k])
                    N = g.prob(tf.cast(x[t],tf.float64))                    
                    B[j][t] += R[j,k]*N
        
        B = tf.stack(B)

        # forward
        
        # initialisation step


        alpha = [[]]
        
        # initialisation step
        for i in range(self.M):
            alpha[0].append(pi[i]*B[i][0])
        
        # induction step 
        for t in range(T-1):
            alpha.append([])
            for j in range(self.M):
                sum_ = alpha[t][0]*A[0,j]
                for i in range(1,self.M):
                    sum_ += alpha[t][i]*A[i,j]      
                alpha[t+1].append(B[j][t+1]*sum_)
                
        alpha = tf.stack(alpha)
        
        return tf.math.log(tf.reduce_sum(alpha[T-1]))
        

    def fit(self,X,M,K,epochs=10,lr=1e-3,plot=True):
        N = len(X)
        D = len(X[0][0])
        self.D = D
        self.M = M
        self.K = K
        # these are not the true parameters , but rather the ones we optimise
        
        # pi is uniformly distributed
        pi_opt = np.ones(M)/M
        A_opt = np.random.randn(M,M)
        R_opt = np.random.randn(M,K)
        mu_opt = np.zeros((M,K,D))
        for m in range(M):
            for k in range(K):
                n = np.random.choice(N)
                random_x = X[n]
                random_idx = np.random.choice(len(random_x))
                random_point = np.random.choice(random_x[random_idx])
                mu_opt[m,k] = random_point
        
        # another constraint is that the variance must always be positive
        # so we will optimise log sigma freely
        # and use exp(sigma) in our calculations
        log_sigma_opt = np.random.randn(M,K,D,D)


        self.pi_opt = tf.Variable(pi_opt,name="pi")
        self.A_opt = tf.Variable(A_opt,name="A")
        self.R_opt = tf.Variable(R_opt,name="R")
        self.mu_opt = tf.Variable(mu_opt,name="mean")
        self.log_sigma_opt = tf.Variable(log_sigma_opt,name="cov")

        opt = Adam(learning_rate=lr)
        costs = []
        for epoch in range(epochs):
            cost = 0
            for x in X:
                with tf.GradientTape() as tape:
                    loss_value = -self(x, training=True)  
                    cost += loss_value
                    
                grads = tape.gradient(loss_value, self.trainable_weights)
                opt.apply_gradients(zip(grads, self.trainable_weights))
        
            costs.append(cost) # total cost for all samples
            print("epoch: ",epoch+1,'cost: ',cost.numpy())
        if plot:
            plt.plot(costs)
            plt.show()
````

````python
tf.keras.backend.set_floatx('float64')
D = len(X[0][0])
model = HMM_GMM_T()
M = 1
K = 1
# may need to change lr (try 1e-1 but may give nans)  or increase number of epochs
model.fit(X,M,K,lr=1e-1)
````
