<h4>Whats wrong with counting ? </h4>

Nothing ! in fact counting and simple classifiers like Naive Bayes tend to perform extremely well

But it does have some properties that may be considered undesirable

for ex , consider the fact that words like "a","the","and","in","to",etc. probably appear many times in all documents , so they give us very high feature values even though they are totally meaningless words

no only this , the count for these words is probablly so large that they overshadow all the other words 

these are called stopwords and one common technique is to simply remove them from the dataset before doing any machine learning

 ---
 
<h4>TF-IDF</h4>
 
we know that words that appear in many documents are probably less meaningfull , and therefore we can weigh each vector component , which in this case is a word , by something related to how many documents that word appears in 

so intuitively we may do something like Raw word count / Document ccount

The numerator tells us how many times does this word appear in this document , and the denominator tells us how many documents does this word appear in in total

In practice we modify this to use log counts , or add smoothing , but we wont get into details now

---

<h4>Important Takeaway</h4>

The very important thing to remember is that , no matter the technique we use , we are always interested in a matrix (table) of size <strong><u>VxD</u></strong>

V = vocabulary size (total number of possible words)

D = vector dimensionality (ex: if we ware counting up how many times a word appears in a set of books , D = total number of books )

so we have V words , and each word is represented by a vector of length D

---

<h4>Word Embeddings</h4>

A word embedding is just a fancy name for a feature vector that represents a word , in other words , we can take a categorical object , a word in this case , and map this object to a list of numbers , a vector , then we say that we have embedded this word into a vector space , and hence the name word embeddings

so when we say word embedding what we mean is a vector that represents a word

<h1>Math</h1>

one of the most popular applications of word embeddings is Word Analogies

This is where the famous king - man ~= Queen - Woman comes from

<h4>Examples</h4>

first lets just talk about analogies in general 

<img src="extras/27.5.PNG" width="400"></img>

we have king as to man as queen is to woman , this makes sense since ling is the head of the royal family and is a male , queen is the head of the royal family and is a female , so if we subtract man from king , we migh think of this as a vector representing the head of the royal family without the gender component , sound complicated so lets take more examples

<h4>Visualising Analogies</h4>

How can we visualise and calculate word vectors ?

As usual it is very useful to think of these geometrically


first lets recall that a word embedding just means a word vector , in other words if we have a grid we can say that each word is represented by a dot on this grid


<img src="extras/27.6.PNG" width="200"></img>

now what happens when we subtract one vector from another vector , well this just gives us the vector between these two vectors

Woman - Man ~= Queen - King

LeftVector = Woman-Man

RightVector = Queen - King

LeftVector ~= RightVector

so when we say LeftVector ~= RightVector what are we really saying ?

we are saying that these two difference vectors are approximately the same , a vector has two components , a direction and a magnitude , saying that LeftVector ~= RightVector means that we are saying that the magnitude and direction of these vectors are very close to one another

later when we consider the cosine distance , we will see that we only consider directipn

---

<h4>How to find analogies</h4>

We know that there are 4 words in evey analogy , so what we can do is take 3 of these words and we try to find the 4th word

since everything is a vector we can rearrange this equation

King - Man = ? - Woman

King - Man + Woman = ?

in this example we are pretending that we dont know Queen , instead we use ?

now we know that

SomeVector = King - Man + Woman

our job is to find the word most associated with that vector , we can do that using distance

so what we want to do is find a word in our vocabulary whose word vector is closest to king - man + woman , here is a psudeo-code

```python
closest_distance = infinity
best_word = None
test_vector = king - man + woman
for word,vector in vocabulary:
    distance = get_distance(test,vector,vector)
    if distance < closest_distance:
        closest_distance = distance
        best_word = word
```

what we want to know is which word's word vector is closest to this vector

this can be accomplished by looping throgh every word in the vocabulary and checking the distance between the word vector and the test vector , and keepng track of which word had the closest vector

---

<h4>Distance</h4>

As of how to calculate the distance , the are multiple choices

Euclidean dist :

$$\vert a - b \vert ^2 = (a_1 - b_1)^2 + \ldots + (a_D - b_D)^2$$

cosine dist :  $$1 - \frac{a^Tb}{\vert a \vert \vert b \vert}$$

$since \ a^Tb = \vert a \vert \vert b \vert \cos(\theta)$

in practice , it is more common to use the cosine distance which is related to the cosine of the angle between two vectors

note that we need to invert the cosine of the angle because it returns 1 when the two vectors are close and -1 when the two vectors are far apart

so we subtract it from 1 and the result is that if the vectors are very clost together , we get $1-1=0$ , but if the vectors are far apart we get $1--1 = 2$

<h4>Remarkable</h4>

Algorithms like word2vec and GloVe (which we will look at later) have no concept of analogies , what they want to optimise is totally unrelated to word analogies , the very dact that word analogies emerged out of the model and training process is a very intriguing result , other methods however like Raw counts and TF-IDF dont give us good analogies

````python
import numpy as np
from glob import glob
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.manifold import TSNE
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
````

````python
# get file names
N_files = None
N_words = 2000
folder = "datasets/wiki"
wiki_files = glob(folder+"/*.txt")[:N_files]
````

````python
wiki_articles = []

article = ''

for file in wiki_files:
    for line in open(file,encoding = "utf8"):
        line = line.rstrip()
        # each file contains several documents
        # each document has a title that starts with [[ and ends with ]]
        
        if line.startswith("[[") and line.endswith("]]"):
            # previous article ended
            wiki_articles.append(article)
            article = ""
        
        # skip headers , ...
        elif line and line[0] not in ('[', '*', '-', '|', '=', '{', '}'):
            article += " " + line
    
del wiki_articles[0]
wiki_articles.append(article)
````

````python
# we will consider only top 2000 words
# helps tsne with dimensionality reduction
vectoriser = TfidfVectorizer(max_features=N_words)
# we want a term-document matrix so we take transpose
X = vectoriser.fit_transform(wiki_articles).T
````

````python
Z = TSNE(n_components=2).fit_transform(X)
````

````python
plt.scatter(Z[:,0],Z[:,1])
word2idx = vectoriser.vocabulary_
idx2word = {v:k for k,v in word2idx.items()}
for i in range(len(word2idx)):
    plt.annotate(s=idx2word[i],xy=(Z[i,0],Z[i,1]))
plt.show()
````

````python
def get_pos(word):
    idx = word2idx[word]
    pos = Z[idx]
    return pos
# this can be helpful to search for some words in cloud
get_pos('father')
# we see that some similar words are grouped together
# king,throne,reign,emperor,imperial
# father,son,daughter,married
# note that this code has some randomness to it due to t-sne
````

````python
# now we will use pretrained word vectors form GloVe and word2vec to find word analogies
# GloVe : http://nlp.stanford.edu/data/glove.6B.zip
# word2vec : https://drive.google.com/file/d/0B7XkCwpI5KDYNlNUTTlSS21pQmM/edit?usp=sharing
import numpy as np
from sklearn.metrics.pairwise import pairwise_distances
from gensim.models import KeyedVectors
````

````python
# there are multiple files dependiong on the dimensionality used
# we will use D = 300 , so each word is represented by a vector of length 300
# most frequent words come first in file so we can take the first V words

word2idx = {}
V = 50000
D = 300 #[50,100,200,300]

X = np.zeros((V,D))

for i,line in enumerate(open('datasets/glove/glove.6B.'+str(D)+'d.txt',encoding='utf8')):
    line = line.rstrip().split()
    word,vec = line[0],line[1:]
    word2idx[word] = i
    X[i] = np.array(vec)
    
    if i == V-1:
        break
````

````python
# now lets write a function to find analogies
# king - man = ? - woman

def analogy(w1,w2,w4):
    # first lets get our vector
    king = X[word2idx[w1]]
    man = X[word2idx[w2]]
    woman = X[word2idx[w4]]
    queen = king - man + woman
    
    # next we calculate the distance between our vector and all other vectors
    # once using euclidean distance then again using cosine
    idx2word = {v:k for k,v in word2idx.items()}
    metrics = ['euclidean','cosine']
    
    for metric in metrics:
        distances = pairwise_distances(queen.reshape(1,D),X,metric=metric)
        # now we need to consider the 4 closest neighbours to that point
        # not to return a word in [w1,w2,w4]
        closest =  np.argpartition(distances[0], 4)[:4]
        closest = [idx2word[i] for i in closest]
        
        for word in closest:
            if word not in [w1,w2,w4]:
                print(metric,' distance :',w1,"-",w2,'=',word,'-',w4)
                break
    
````

Now we want to discuss how we can directly apply pretrained word vectors in order to do text classification

<h4>Text Classification</h4>

lets start with : How are we going to represent our data ?

we will start by using a bag of words model (other methods will be discussed later)

what this means is , we dont consider the order of words in a sentence , for ex , the phrase "toy dog" would be represented in the same way as "dog toy" even though we know those are two toally different things

<img src="extras/27.7.PNG" width = "300"></img>

---

<h4>Bag of words</h4>

So how can we apply word vectors , such as what we would get from word2vec and GloVe to bag of words

simple , we know that we want to represent a sentence or a document as a single vector 

so first , for each word of the sentence ,  we can use our list of word vectors to ge ta vector

Then we can take the average of all these vectors

$$feature = \frac{1}{\vert sentence \vert} \sum_{w \in sentence} vec(w)$$

for ex : "I like eggs"

feature vector = [vec("I") + vec("like") + vec("eggs")]/3

---

<h4>Then ...</h4>

Now that we have converted every sentence to a vector , we can plug this into any machine learning classifier

---

<h4>whats the meaning of this ?</h4>

in order for the feature vectors to actually be "good" feature vectors , the individual word vectors have to make sense

so this is in gact a method of <strong><u>evaluating</u></strong> how good a word embedding is

other methods of evaluation : word analogies , similarity

<h1>code</h1>

````python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from gensim.models import KeyedVectors
````

````python
class BoW:
    def __init__(self,V=50000,D=300):
        word2idx = {}
        self.V = V
        self.D = D
        
    def fit_glove(self):
        V,D = self.V,self.D
        E = np.zeros((V,D))
        word2idx = {}
        for i,line in enumerate(open('datasets/glove/glove.6B.'+str(D)+'d.txt',encoding='utf8')):
            line = line.rstrip().split()
            word,vec = line[0],line[1:]
            word2idx[word] = i
            E[i] = np.array(vec)
            if i == V-1:
                break
        self.E = E      
        self.word2idx = word2idx
        
    def fit_word2vec(self):
        V = self.V
        word2vec = KeyedVectors.load_word2vec_format(
            'datasets/word2vec/GoogleNews-vectors-negative300.bin',binary=True)
        E = np.zeros((V,300)) # word2vec omes only with D = 300
        words = word2vec.index2word[:V]
        word2idx = {}
        for i in range(V):
            w = words[i]
            E[i] = word2vec[w]
            word2idx[w] = i
        self.E = E
        self.word2idx = word2idx

    def fit(self,source='glove'): # pass 'GloVe' or 'word2vec'
        if source == 'glove':
            self.fit_glove()
        if source == 'word2vec':
            self.fit_word2vec()
            

    def transform(self,sentences):
        N = len(sentences)
        X = np.zeros((N,self.D))
        for i,sentence in enumerate(sentences):
            idxs = []
            words = sentence.lower().split()
            for word in words:
                idx = self.word2idx.get(word,None)
                if idx is not None:
                    idxs.append(idx)
            if len(idxs)>0:
                X[i] = self.E[idxs].mean(axis=0)  
        return X

    def fit_transform(self,sentences):
        return self.transform(sentences)
````

````python
vectoriser = BoW()
vectoriser.fit('glove')
````

````python
# now lets read in the input and output sentences
Xtrain = []
Ytrain = []

Xtest = []
Ytest = []

i = 0

for line in open("datasets/r8/r8-train-all-terms.txt"):
    y,x = line.rstrip().split('\t')
    Ytrain.append(y)
    Xtrain.append(x)

for line in open("datasets/r8/r8-test-all-terms.txt"):
    y,x = line.rstrip().split('\t')
    Ytest.append(y)
    Xtest.append(x)
````

````python
# first we need to decrease vocab size or we run out of memory
# so we will take most frequent V words
words = [w.lower() for s in sents for w in s]
word2count = {}
for s in sents:
    for w in s:
        word2count[w] = word2count.get(w,0)+1
# decrease vocab size if it doesnot fit memory
V = 30000
# sort words in order of frequency
words = {k: v for k, v in sorted(word2count.items(), key=lambda item: item[1],reverse=True)}
words = list(words)[:V]
# all words that didnot make it to this list will be represented by a special token <None>
# we will also use 2 special tokens <start> and <end> 
# using <start> we do not have to calculate unigrams
# using <end> we get the probability of one word being the last word
words = set(words).union(["<start>","<end>","<None>"])
V = len(words) # now that V + our three new tokens
word2idx = {k:v for k,v in zip(words,range(V))}
````

````python
smoothing = 0.1
bigrams = np.ones((V,V))*smoothing
none = word2idx.get("<None>")
for s in sents:
    s = ["<start>"] + [w for w in s] + ["<end>"]
    for i in range(len(s)-1):
        w1,w2 = s[i].lower(),s[i+1].lower()
        # if the word is not in our word2idx
        # we return the idx of the <None> token
        # which is a representative of all other words
        i1,i2 = word2idx.get(w1,none),word2idx.get(w2,none)
        bigrams[i1,i2] += 1

# normalise
bigrams /= bigrams.sum(axis=1,keepdims=True)
# log
bigrams = np.log(bigrams)
````

````python
# now lets write a function to calculate the log probability of a sentence
def prob(s):
    p = 0
    s = ['<start>'] + [w for w in s.split()] + ['<end>']
    for i in range(len(s)-1):
        w1,w2 = s[i],s[i+1]
        i1,i2 = word2idx.get(w1,none),word2idx.get(w2,none)
        p+= bigrams[i1,i2]
    # remember to normalise
    # thinking about it we have add len(s)-1 probabilities
    return p/(len(s)-1)
````

<h1>Math</h1>

<h4>A neural model for bigrams</h4>

Now we want to extend our bigram model a little bit to use a neural model rather than just count each word to get probabilities

recall that the basic structure of a neuron is logistic regression , so that what we will focus on

---

<h4>How do we represent words ?</h4>

we know that logistic regression works on number not words

$$logits = w^Tx$$

here x muxt be a vector

---

<h4>words are categories</h4>

The trick is to think of words as categories , we know that the way we deal with categories in deep learning is one-hot encoding

so if we have V words in our vocabulary , then each word is represented by a V size vector which consists all zeros except a one in a unique spot specifically for that word

ex:

$"a" \rightarrow [1,0,0,0,\ldots]$

$"an" \rightarrow [0,1,0,0,\ldots]$

$\ldots$

$"zoo" \rightarrow [0,0,0,\ldots,1]$

---

<h4>How can logistic regression be used ?</h4>

lets first recall how logistic regression works

we have an input vector x and an output vector y and they are related by this equation

$$p(y|x) = softmax(W^Tx)$$

Interestingly , this looks exactly like a bigram where x is the first word and y is the following word

---

<h4>All data is the same</h4>

so how are we going to get our model into a format that logistic regression can understand ?

By extension , this would be the same for any machine learning model

In general , X is an NxD matrix of input samples , Y is an NxK matrix of output targets

D = number of input features

K = number of output classes

But in this case we want D = V , K = V , where V is the vocabulary size

This because the input and output are both one-hot encoded words of size V

we can create X&Y by looping over corpus and one-hot encoding and storing every bigram we encounter

---

<h4>How do we find W</h4>

we find W by doing gradient descent on the cost

cost function (categorical cross-entropy):

$$J = -\frac{1}{N}\sum^N_{n=1}\sum^V_{k=1}y_{n,k} \log (p(y_{n,k}|x_n))$$

gradient descent :

while not converged :

$$W \leftarrow W - \eta \nabla J , where \nabla J = X^T (p(Y|X)-Y)$$

---

<h4>Plan</h4>

<ul>
    <li>create a bigram model using logistic regression instead of counts</li>
    <li>plot the cost as a function of iteration</li>
</ul>

note how W (in this model) and bigrams (from previous code section) are both VxV matricies !

Are they somehow related ?

Yes , if we take the softmax of W , we should get back an approximation of the bigram probability matrix , therefore we can also do the opposite

we can replace W in the logistic regression model with $log(bigrams)$ (we can show that if W ~= softmax(bigrams) , then W ~= log(bigrams) , again all these are approximations)

from this we can calculate the cross-entropy using that version of W , then plot this cost also , we can then think of this as the ideal version of W that logistic regression is trying to find

<h1>code</h1>

````python
# lets try coding this now
# V = 30,000 already fills a lot of our memory
# lets retry everything with V = 2000
# we will be getting bigram again as we loop we will store words as indicies
# we will store each two consecutive words w1 w2 , w1 in X w2 in Y , so i1 in X , i2 in Y
# we cant store an NxV matrix thats too much so we will do batch gradient descent
# so we keep our words as indicies and one we require it we change them to one hot encoded vectors
import numpy as np
from nltk.corpus import brown
import string
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
````

````python
sents = brown.sents()
````

````python
# same code again with V = 2000
words = [w.lower() for s in sents for w in s]
word2count = {}
for s in sents:
    for w in s:
        word2count[w] = word2count.get(w,0)+1
V = 2000
words = {k: v for k, v in sorted(word2count.items(), key=lambda item: item[1],reverse=True)}
words = list(words)[:V]
words = set(words).union(["<start>","<end>","<None>"])
V = len(words) # now that V + our three new tokens
word2idx = {k:v for k,v in zip(words,range(V))}
````

````python
# same again , but this time as we loop , we populate X,Y
X = []
Y = []
smoothing = 0.1
bigrams = np.ones((V,V))*smoothing
none = word2idx.get("<None>")
for s in sents:
    s = ["<start>"] + [w for w in s] + ["<end>"]
    for i in range(len(s)-1):
        w1,w2 = s[i].lower(),s[i+1].lower()
        i1,i2 = word2idx.get(w1,none),word2idx.get(w2,none)
        X.append(i1)
        Y.append(i2)
        bigrams[i1,i2] += 1
bigrams /= bigrams.sum(axis=1,keepdims=True)
bigrams = np.log(bigrams)
````

````python
class neural_bigram:
    
    def __init__(self,V):
        self.W = np.random.randn(V,V)/np.sqrt(V)
        self.V = V
        self.eps = 1e-14
        
    def softmax(self,a):
        # this follows the definition but we also subtract the max to avoid numerical overflow
        a = a - a.max()
        return np.exp(a)/(np.exp(a).sum(axis=1,keepdims=True))
    
    def cost(self,Y,P):
        N = len(Y)
        cost =  -np.sum(Y*np.log(P+self.eps))
        return cost/(N-1)
    
    # calculate smoothed loss to reduce variability
    # this is not quite the normal classification problem
    # in a typical classification problem , we have one input that maps to a desired target
    # for bigrams , an input word has many possible next words
    # so we dont really have a notion of accuracy
    # because of this the loss is going to jump around a lot
    # but we can make it move around less by using exponential smoothing
    def smooth_loss(self,costs,decay=0.99):
        N = len(costs)
        smoothed = np.zeros(N)
        last = 0
        for t in range(N):
            z = decay * last + (1-decay) * costs[t]
            smoothed[t] = z/(1-decay**(t+1))
            last = z
        return smoothed
        
    def fit(self,X,Y,lr=1e-2,epochs=2,batch_size=2048,print_every=10,bigrams=None):
        # model almost converges in first epoch
        # so we will plot the loss across the batches
        N = len(X)
        V = self.V
        n_batches = np.ceil(N/batch_size).astype(int)
        costs = []
        epoch_costs = []
        bigram_costs = []
        for epoch in range(epochs):
            X,Y = shuffle(X,Y)
            epoch_costs = []
            for i in range(n_batches):
                X_idx = X[i*batch_size:(i+1)*batch_size]
                Y_idx = Y[i*batch_size:(i+1)*batch_size]
                # last batch may have size < batch_size
                size = len(X_idx)
                
                X_batch = np.zeros((size,V))
                Y_batch = np.zeros((size,V))
                # one hot encode
                rows = np.arange(size)
                X_batch[rows,X_idx] = 1 
                Y_batch[rows,Y_idx] = 1
                
                # now lets do gradient descent
                p = self.softmax(X_batch@self.W)
                self.W = self.W -lr*(X_batch.T@(p-Y_batch))
                cost = self.cost(Y_batch,p)
                epoch_costs.append(cost)
                
                # calculate bigram cost for this batch
                if bigrams is not None and epoch == 0:
                    p_ideal = self.softmax(X_batch@bigrams)
                    cost_ideal = self.cost(Y_batch,p_ideal)
                    bigram_costs.append(cost_ideal)
                
                
                if (i+1)%print_every == 0:
                    print('epoch :',epoch+1,"batches completed: ",i+1,'/',n_batches,'cost: ',np.array(epoch_costs).mean())
            print('epoch :',epoch+1,'cost',np.array(epoch_costs).mean())
            
            costs += epoch_costs
        print("Final cost : ",np.array(epoch_costs).mean())
        if bigrams is not None:
            bigram_cost = np.array(bigram_costs).mean()
            print("Bigram (ideal) cost : ",bigram_cost)
        smooth_loss = self.smooth_loss(costs)
        plt.plot(costs,label='batch costs')
        plt.plot(smooth_loss,label='exponentially smoothed costs')
        if bigrams is not None:
            plt.plot([bigram_cost]*len(costs),label = "Bigram (ideal) cost")
        plt.legend()
        plt.show()
````

````python
model = neural_bigram(V)
````

````python
model.fit(X,Y,bigrams=bigrams)
````

<h1>Math</h1>

now we want to extend our model by using a multi-layer neural network instead of just logistic regression

---

<h4>Neural Network Bigram Model</h4>

in a simple neural network , we now have two layers of weights , $W_1$ , $W_2$

<img src='extras/27.8.PNG' width='300'></img>

<h4>OLD:</h4>

$$p(y\vert x) = softmax(W^Tx)$$

<h4>NEW:</h4>

$$h = tanh(W_1^Tx)$$

$$p(y\vert x) = softmax(W_2^Th)$$

---

<h4>All machine learning interfaces are the same</h4>

One thing to keep in mind is that the way we represent our data X & Y dont change , this is due to the fact that all machine learning interfaces are the same , thats why we can plug-in the same data into any other classifier 

so we still get out $p(y\vert x)$ at the end which we can treat exactly the same as $p(y \vert x)$ we had before

---

<h4>Learning</h4>

Next step , how to find W1 and W2 ?

Algorithm is pretty much the same as before 

cost is exactly the sa me

$$J = - \frac{1}{N} \sum_{n=1}^{N}\sum_{k=1}^V y_{n,k}\log(p(y_{n,k})\vert x_n)$$

The only difference now is that we have to update both W1 and W2

$$W_2 \leftarrow W_2 - \eta \nabla_{W_2} J$$

$$W_1 \leftarrow W_1 - \eta \nabla_{W_1} J$$

we looked at how to calculate these gradient at the past :

$$\nabla_{W_2}J = H^T (p(Y|X) - Y)$$

$$\nabla_{W_1}J = X^T \left[ (p(Y|X) - Y) W_2^T \odot (1-H^2) \right]$$

note : here we assume using tanh

note : the gradient for $W^{(2)}$ is the same we get when we do binary logistic regression (for full derivation refer to notebook 5)

note : for the gradient for $W^{(1)}$ refer to notebook 18

---

<h4>Hyperparameters</h4>

We need to consider the hidden layer size , call it "D"

<img src = 'extras/27.9.PNG' width = '200'></img>

so :

shape($W_1$) == VxD

shape($W_2$) == DxV

where V is the vocabulary size , and D is an arbitary hyperparameter that we are allowed to choose

---

<h4> D $<<$ V </h4>

inour case we want to choose D to be much smaller than V , there are multiple reasons for this

Autoencoder pattern : these are neural networks that learn to reproduce their input , but importanly autoencoders always have this "Bow tie" shape , we call this a bottleneck , because we are trying to squeeze in the raw information (V dimensions) into much smaller space (D dimensions) , and we hope to do this without losing any data

<img src='extras/27.10.PNG' width='500'></img>

---

<h4>Logistic Regression</h4>

recall from the previous example that Logistic Regression was slow

we might be tempted to think that Logistic Regression would be faster than a neural network since it is linear , but here comes an exception

suppose V = 10,000 , then W contains VxV = $10^8$ numbers! (100 million)

suppose D = 100 , then W1 and W2 contain $10^210^4 = 10^6$ numbers each , thats only 2 million total 

so by using a neural network we get 50:1 compression ratio

---

<h4>Bigram Probabilities</h4>

Just like how the logistic regression weights where a model for the bigram probabilities using counting , so are the neural network weights , but how can this be when we have two matricies instead of one ?

The trick is to look at how they combine to give $p(y \vert x)$

if we write this out what we see that , if we look at this the right way , we can see how these combine to give us an approximation of W

<strong>Logistic Regression</strong>

$$p(y|x) = softmax(\boxed{W^T}x)$$

<strong>Neural Network</strong>

$$p(y\vert x) = softmax(\boxed{W_2^T tanh(W_1^T}x))$$

since x only contains 1s and 0s we can effectively ignore it

and so what we can see that with a neural network , our model of the bigram probability matrix is just :

$$\hat W = tanh(W_1)W_2$$

this makes sense since we know , as per the rules of matrix multiplication , if we multiple (VxD)(DxV) $\rightarrow$ (VxV)

<h1>code</h1>

````python
# this should look a lot like the previous code
# except that we change W to W1 and W2
# and modify the updates
import numpy as np
from nltk.corpus import brown
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
````

````python
sents = brown.sents()
````

````python
words = [w.lower() for s in sents for w in s]
word2count = {}
for s in sents:
    for w in s:
        word2count[w] = word2count.get(w,0)+1
V = 2000
words = {k: v for k, v in sorted(word2count.items(), key=lambda item: item[1],reverse=True)}
words = list(words)[:V]
words = set(words).union(["<start>","<end>","<None>"])
V = len(words) # now that V + our three new tokens
word2idx = {k:v for k,v in zip(words,range(V))}
````

````python
X = []
Y = []
smoothing = 0.1
bigrams = np.ones((V,V))*smoothing
none = word2idx.get("<None>")
for s in sents:
    s = ["<start>"] + [w for w in s] + ["<end>"]
    for i in range(len(s)-1):
        w1,w2 = s[i].lower(),s[i+1].lower()
        i1,i2 = word2idx.get(w1,none),word2idx.get(w2,none)
        X.append(i1)
        Y.append(i2)
        bigrams[i1,i2] += 1
bigrams /= bigrams.sum(axis=1,keepdims=True)
bigrams = np.log(bigrams)
````

````python
class neural_net_bigram:
    
    def __init__(self,V,D=100):
        self.W1 = np.random.randn(V,D)/np.sqrt(V)
        self.W2 = np.random.randn(D,V)/np.sqrt(D)
        self.V = V
        self.eps = 1e-14
        
    def softmax(self,a):
        a = a - a.max()
        return np.exp(a)/(np.exp(a).sum(axis=1,keepdims=True))
    
    def cost(self,Y,P):
        N = len(Y)
        cost =  -np.sum(Y*np.log(P+self.eps))
        return cost/(N-1)
    
    def smooth_loss(self,costs,decay=0.99):
        N = len(costs)
        smoothed = np.zeros(N)
        last = 0
        for t in range(N):
            z = decay * last + (1-decay) * costs[t]
            smoothed[t] = z/(1-decay**(t+1))
            last = z
        return smoothed
        
    def fit(self,X,Y,lr=1e-3,epochs=2,batch_size=2048,print_every=10,bigrams=None):
        # model almost converges in first epoch
        # so we will plot the loss across the batches
        N = len(X)
        V = self.V
        n_batches = np.ceil(N/batch_size).astype(int)
        costs = []
        epoch_costs = []
        bigram_costs = []
        for epoch in range(epochs):
            lr = lr/(10**(epoch))
            X,Y = shuffle(X,Y)
            epoch_costs = []
            for i in range(n_batches):
                X_idx = X[i*batch_size:(i+1)*batch_size]
                Y_idx = Y[i*batch_size:(i+1)*batch_size]
                # last batch may have size < batch_size
                size = len(X_idx)
                
                X_batch = np.zeros((size,V))
                Y_batch = np.zeros((size,V))
                # one hot encode
                rows = np.arange(size)
                X_batch[rows,X_idx] = 1 
                Y_batch[rows,Y_idx] = 1
                
                # now lets do gradient descent
                H = np.tanh(X_batch@self.W1)
                p = self.softmax(H@self.W2)
                
                gW2 = H.T@(p-Y_batch)
                self.W2 = self.W2 -lr*gW2
                
                gW1 = X_batch.T@((p-Y_batch)@self.W2.T*(1-H**2))
                self.W1 = self.W1 -lr*gW1
                
                cost = self.cost(Y_batch,p)
                epoch_costs.append(cost)
                
                # calculate bigram cost for this batch
                if bigrams is not None and epoch == 0:
                    p_ideal = self.softmax(X_batch@bigrams)
                    cost_ideal = self.cost(Y_batch,p_ideal)
                    bigram_costs.append(cost_ideal)
                
                
                if (i+1)%print_every == 0:
                    print('epoch :',epoch+1,"batches completed: ",i+1,'/',n_batches,'cost: ',np.array(epoch_costs).mean())
            print('epoch :',epoch+1,'cost',np.array(epoch_costs).mean())
            
            costs += epoch_costs
        print("Final cost : ",np.array(epoch_costs).mean())
        if bigrams is not None:
            bigram_cost = np.array(bigram_costs).mean()
            print("Bigram (ideal) cost : ",bigram_cost)
        smooth_loss = self.smooth_loss(costs)
        plt.plot(costs,label='batch costs')
        plt.plot(smooth_loss,label='exponentially smoothed costs')
        if bigrams is not None:
            plt.plot([bigram_cost]*len(costs),label = "Bigram (ideal) cost")
        plt.legend()
        plt.show()
````

````python
model = neural_net_bigram(V)
````

````python
model.fit(X,Y,bigrams=bigrams)
````

<h1>Math</h1>

We already saw how we can make our model more effecient by limiting the parameters of the model

Now we want to see how we can even go faster using some numpy indexing tricks , these will also be useful in further implementations

---

<h4>Inefficiency</h4>

the first thing we want to do is tackle some ineddiciencies in our previous code

for every batch , we needed to create a new one hot encoded input matrix and a new one hot encoded target matrix

if the batch is of size N , the nthe two matricies are of size NxV

if V = 10,000 , then we create 10,000x more data than we started with (thats each word being a single number (index) then becoming a V size vector)

Now is there a way to reduce this ineffeciency ?

---

<h4>Example</h4>

consider the result of the following multiplications

<img src='extras/27.11.PNG' width='500'></img>
 
<img src='extras/27.12.PNG' width='500'></img>

<img src='extras/27.13.PNG' width='500'></img>

what is the pattern that we notice , back to our first example [1,0,0]

<img src='extras/27.11.PNG' width='500'></img>
 
the question is , what is this a one hot encoding of ?

remember that one-hot encoding puts a 1 wherever the original input category is

so if the original input = 0 then we set one one_hot_vec[0] = 1

notice that the result of multiplying this one-hot vector by the weight matrix gave us the 0th rule of the weight matrix

lets look at the second example again

<img src='extras/27.12.PNG' width='500'></img>

[0,1,0] is the one-hot encoding of the input 1

as we can see the result of multiplying this one hot vector by the weight matrix gave us the row at index 1

the same pattern follows when we look at the third example

<img src='extras/27.13.PNG' width='500'></img>

how do we generalise this ?

if we have an input word k , where the one-hot vector contains all 0s except a 1 at index k

Then multiplying the one-hot vector by the weight matrix is the same as selecting the kth row of the weight matrix !

so these 2 code snippets are equivalent

OLD :

```python
one_hot = get_one_hot(k)
one_hot @ W
```
NEW :

```python
W[k]
```

but the second method is much faster than the first and much more memory effecient

That means that for a batch of length N , instead of creating a one-hot encoded matrix of size NxV we can just select the rows with the N indexes , for V = 10000 , thats 10,000x the savings

---

<h4>Enter Nupmy</h4>

numpy actually supports selecting multiple rows at the same time (it happens that we already did so)

ex: W[[0,1]] gives us both the row at index 0 and the row at index 1

ex : W[[0,0]] gives us a 2xD array with the row at index 0 repeated twice

so W[inputs] returns an NxD matrix , N = size of batch , this is equivalent (yet more effecient) to multiplying the one-hot encoded version of inputs by weight matrix

---

<h4>Apply this everywhere</h4>

<strong>predictions:</strong> 

```python
H = np.tanh(W1[inputs])
p = softmax(H@W2)
```

<strong>cost:</strong> 

OLD:
```python
one_hot_targets*log(predictions)
```

NEW:

```python
log(predictions[np.arange(N),targets])
```

in this way we can index both rows and columns at the same time , for the rows we index 0 to N-1 in that order essentially selecting one row at a time and for the columns we just select the targets

This is allowed since the one-hot targets contain mostly 0s so they dont actually contribute anything to result , and hece when we multiply them it wastes lots of time

---

<h4>Double indexing</h4>

we index both rows and columns with lists of equal length

each list contains the corresponding index for that dimension 

ex:

```python
X[[1,2,3],[4,5,6]] --> [X[1,4],X[2,5],X[3,6]]
```

---

<h4>Improving the updates</h4>

Another place to apply these improvements is the update equations

If we reason about this in terms of backprop , we can see which values get updated and which values do not

if we have 1 input word , only the weights from the kth row of $W_1$ will affect the hidden layer , but all the weights in $W_2$ will affect the output layer

<img src='extras/27.14.PNG' width='300'></img>

only the weights that contribute to the rror need to be updated

---

<h4>Updating $W_2$</h4>

now lets rellok at the equation of the updates , we find a problem

our update was :

```python
W2 = W2 - lr*hidden.T@(predictions-targets)
```

originally the targets where one-hot encoded so both predictions and targets where NxV matriciesnow only predicitons are NxV adn the targets are a vector of length N containing word indexes

now , we can use the double indexing trick , notice that when we subtract the one-hot encoded targets we are mostly subtracting 0s , in other words they have no effect

OLD:

```python
W2 = W2 -lr*hidden.T@(predictions-one_hot_targets)
```

we only need to subtract 1 from where target is :

NEW:

```python
dropout = predictions
dropout[np.arange(N),targets] -=1
```

---

<h4>Updating $W_1$</h4>

we only need to update the row of $W_1$ that corresponds to the input word

but we are doing batch gradient descent and we will have multiple input words

even more problematic is that we may have the same input word showing up twice

ex : "<u>the</u> quick brown fox jumps over <u>the</u> lazy dog"

but why is that a problem ?

this is not much of a problem if it shows on the right side of the equation because that is what is getting assigned to the left side

```python
new_var = do something to W[inputs]
```

but it is a problem if it shows on the left side :

```python
W[inputs] = ... some expresion ...
```
this might select same row of W more than once - the behaviour is undefined

maybe the first assignement supersedes the others or maybe the last assignment does or maybe they all sum together , we can test to see what happens , but generally speaking we dont like ambiguous code in programming so its best not to do it at all

---

<h4>Solution 1</h4>

fall back to the one-hot encoding method

```python
W1 = W1 - lr*onehot_inputs.T @ dhidden
```

actually when we look at the update equation we can see why it doesnot work using indexing method

the problem is that we need to transpose the one-hot encoded matrix , but if the input is just a list of word indexes , its not straight forward to transpose it

this solution is not preferable since our goal is to use the indexing method (we can compare this to next solution to see which works better)

---

<h4>Solution 2</h4>

pretend we are doing stochastic gradient descent and update one word at a time

not ideal since we like to make use of  vectorisation which is much faster than the slow python for loops

since we cant use vectorisation anyway , we will have to do this

```python
i = 0
for w in inputs:
    W1[w] = W1[w] - lr*dhidden[i]
    i+=1
```

here dhidden is $\nabla_{W_1}J$ , this is calculated in the same way as before with a simple exception

looking back at the equation of $\nabla_{W_1}J$ we see that without multiplying by $X^T$ we get an NxD matrix which is what we want to update the N rows (one for each word) each of size D , the $X^T$ was used to transform this into a VxD matrix which will have all 0s in rows corresponding to rows which wont get updated in W1 , and the appropriate values in rows corresponding to rows which would get updated ,  now that we only act on updatable weights NxD suffices , we can do a simple example to show this

assume N = 2 , V = 3 , D = 1

assume this is our input matrix size NxV

$$
    \begin{bmatrix}
    1 & 0 & 0 \\
    0 & 0 & 1
    \end{bmatrix}
$$

And assume that this is our NxD matrix of updates

$$
    \begin{bmatrix}
    1\\ 
    2
    \end{bmatrix}
$$

So that the row 0 (1 0 0 ) should get an update of 1 and row 2 (001) should get and update of 2 , and row 1 (010) should get an update of 0 since it is not found in inputs , lets see if multiplying by $X^T$ accomplishes this


$\begin{bmatrix}1 & 0\\ 0&0 \\0 & 1 \end{bmatrix}\begin{bmatrix}1\\ 2\end{bmatrix} = \begin{bmatrix}1\\0\\ 2\end{bmatrix} $

which is what we expect , an update of 1 for row 0 and update of 0 for row 1 and an update of 2 for row 2

if we use selection then there is no need for this multiplication

we can also see with this loop that the reason we need to do a loop at all is because the thing that indexes W1 is different from the thing that indexes dhidden

for W1 we want to index it using the current word W , for dhidden we want to index it using the corresponding index i which goes from 0 to N-1

now lets modify our neural net model so it used indexing everywhere :

    <li>training</li>

<h1>code</h1>

````python
# now lets rewrite the code , this time efficiency++
import numpy as np
from nltk.corpus import brown
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
````

````python
sents = brown.sents()
````

````python
words = [w.lower() for s in sents for w in s]
word2count = {}
for s in sents:
    for w in s:
        word2count[w] = word2count.get(w,0)+1
V = 2000
words = {k: v for k, v in sorted(word2count.items(), key=lambda item: item[1],reverse=True)}
words = list(words)[:V]
words = set(words).union(["<start>","<end>","<None>"])
V = len(words) # now that V + our three new tokens
word2idx = {k:v for k,v in zip(words,range(V))}
````

````python
X = []
Y = []
smoothing = 0.1
bigrams = np.ones((V,V))*smoothing
none = word2idx.get("<None>")
for s in sents:
    s = ["<start>"] + [w for w in s] + ["<end>"]
    for i in range(len(s)-1):
        w1,w2 = s[i].lower(),s[i+1].lower()
        i1,i2 = word2idx.get(w1,none),word2idx.get(w2,none)
        X.append(i1)
        Y.append(i2)
        bigrams[i1,i2] += 1
bigrams /= bigrams.sum(axis=1,keepdims=True)
bigrams = np.log(bigrams)
````

````python
class neural_net_bigram:
    
    def __init__(self,V,D=100):
        self.W1 = np.random.randn(V,D)/np.sqrt(V)
        self.W2 = np.random.randn(D,V)/np.sqrt(D)
        self.V = V
        self.eps = 1e-14
        
    def softmax(self,a):
        a = a - a.max()
        return np.exp(a)/(np.exp(a).sum(axis=1,keepdims=True))
    
    def cost(self,Y_idx,P):
        N = len(Y_idx)
        cost =  -np.sum(np.log(P[np.arange(N),Y_idx] + self.eps))
        return cost/(N-1)
    
    def smooth_loss(self,costs,decay=0.99):
        N = len(costs)
        smoothed = np.zeros(N)
        last = 0
        for t in range(N):
            z = decay * last + (1-decay) * costs[t]
            smoothed[t] = z/(1-decay**(t+1))
            last = z
        return smoothed
        
    def fit(self,X,Y,lr=1e-3,epochs=3,batch_size=2048,print_every=10,bigrams=None):
        # model almost converges in first epoch
        # so we will plot the loss across the batches
        N = len(X)
        V = self.V
        n_batches = np.ceil(N/batch_size).astype(int)
        costs = []
        epoch_costs = []
        bigram_costs = []
        for epoch in range(epochs):
            lr = lr/(10**(epoch))
            X,Y = shuffle(X,Y)
            epoch_costs = []
            for i in range(n_batches):
                X_idx = X[i*batch_size:(i+1)*batch_size]
                Y_idx = Y[i*batch_size:(i+1)*batch_size]
                
                size = len(X_idx)
                # no one hot encoding just jump to training
                
                # no matrix multiplication , use indexing
                H = np.tanh(self.W1[X_idx])
                p = self.softmax(H@self.W2)
                
                # lets calculate cost first since we will be modifying p
                # the cost uses indexes not one-hot encoded matrix
                cost = self.cost(Y_idx,p)
                epoch_costs.append(cost)

                # just subtract certain elements
                p[np.arange(size),Y_idx] -=1 # this is equivalent to p-Y_batch
                gW2 = H.T@p
                self.W2 = self.W2 -lr*gW2
                
                # we no longer need to multiply by X_batch.T
                gW1 = p@self.W2.T*(1-H**2)
                
                for j,w in enumerate(X_idx):
                    self.W1[w] = self.W1[w] -lr*gW1[j]
                                
                # calculate bigram cost for this batch
                if bigrams is not None and epoch == 0:
                    p_ideal = self.softmax(bigrams[X_idx])
                    cost_ideal = self.cost(Y_idx,p_ideal)
                    bigram_costs.append(cost_ideal)
                
                
                if (i+1)%print_every == 0:
                    print('epoch :',epoch+1,"batches completed: ",i+1,'/',n_batches,'cost: ',np.array(epoch_costs).mean())
            print('epoch :',epoch+1,'cost',np.array(epoch_costs).mean())
            
            costs += epoch_costs
        print("Final cost : ",np.array(epoch_costs).mean())
        if bigrams is not None:
            bigram_cost = np.array(bigram_costs).mean()
            print("Bigram (ideal) cost : ",bigram_cost)
        smooth_loss = self.smooth_loss(costs)
        plt.plot(costs,label='batch costs')
        plt.plot(smooth_loss,label='exponentially smoothed costs')
        if bigrams is not None:
            plt.plot([bigram_cost]*len(costs),label = "Bigram (ideal) cost")
        plt.legend()
        plt.show()
````

````python
model = neural_net_bigram(V)
````

````python
model.fit(X,Y,bigrams=bigrams)
````
