<h1>Math</h1>

we want to consider what to do if output is more than two categories

lets first reconsider binary output (as mentioned before) , all we needed was only 1-output node to represent p(Y=1|X) , and P(Y=0|X) = 1 - P(Y=1|X)

another way of doing this is having two output nodes

<img src="extras/17.6.PNG" width='300'></img>

this is how we calculate predictions

$$p(Y=1|X) = \frac{e^{a_1}}{e^{a_1} + e^{a_2}}$$

$$p(Y=2|X) = \frac{e^{a_2}}{e^{a_1} + e^{a_2}}$$

we exponentiate the activations to make sure we have positive numbers , then we normalise to get valid probabilities , if we add both together , clearly they sum up to 1

since every input node has to be connected to every output node , we have D input nodes and 2 output nodes so W is Dx2 , W = $[w_1 w_2]$

It should be easy to extend this for K-classes

Softmax for K classes:

$$P(Y=k|X) = \frac{e^{a_k}}{e^{a_1} + e^{a_1} + ... + e^{a_K}}$$

$W = [w_1 w_2 ... w_K]$ , (DxK matrix)

so a vectorised logistic regression with softmax :

$$A_{NxK} = X_{NxD} W_{DxK} \rightarrow Y_{NxK} = softmax \left( A_{NxK}\right)$$

<h1>Math</h1>

we want to discuss the relation between sigmoid and softmax , in fact we will show when do they become equivalent

lets write down the equations for softmax , for two classes

$$p(Y=1|X) = \frac{e^{w_1^Tx}}{e^{w_1^Tx} + e^{w_0^Tx}}$$

$$p(Y=0|X) = \frac{e^{w_1^Tx}}{e^{w_1^Tx} + e^{w_0^Tx}}$$

now lets divide top and bottom of p(Y=1|X) by $e^{w_1^Tx}$

$$ p(Y=1|X) = \frac{1}{\exp({{w_0 - w_1}^Tx)} } $$

if we take a weight $w = w_1 - w_0$ , then this takes the same form of the sigmoid  , this tells us that in the case of having only 2 classes , having 2 weights is redundant

<h1>code</h1>

```python
# now lets try to write a feedforward NN
import numpy as np
import matplotlib.pyplot as plt

Nclass = 500

# 3 gaussian clouds
X1 = np.random.randn(Nclass,2) + np.array([0,-2])
X2 = np.random.randn(Nclass,2) + np.array([2,2])
X3 = np.random.randn(Nclass,2) + np.array([-2,2])

X = np.concatenate((X1,X2,X3))
# labels
Y = np.array([0]*Nclass+[1]*Nclass+[2]*Nclass)

plt.scatter(X[:,0],X[:,1],c=Y,s=100,alpha=0.5)
plt.show()
```

```python
def softmax(output):
    output = np.exp(output)
    output = output / output.sum(axis=1,keepdims=True)
    return output

def sigmoid(a):
    return 1/(1+np.exp(-a))
```

```python
D = 2
M = 3
K = 3

W1 = np.random.randn(D,M)
b1 = np.random.randn(M)

W2 = np.random.randn(M,K)
b2 = np.random.randn(K)


def forward(X,W,b,act_fn):
    out = X@W+b
    act = act_fn(out)
    return act

layer_1 = forward(X,W1,b1,sigmoid)
final_layer = forward(layer_1,W2,b2,softmax)

preds = np.argmax(final_layer,axis=1)
```

```python
plt.scatter(X[:,0],X[:,1],c=preds,s=100,alpha=0.5)
plt.show()
```

```python
print('classification rate: ',np.mean(Y==preds))
```
