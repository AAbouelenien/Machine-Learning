So lets start by not doing any math, lets just look at convolution qualitatively

The output image is what we get when we <strong>convolve</strong> the input image with the filter

In other words, if we perform the convolution operation on the input image and the filter, we get the output image

The mathematical symbol for convolution is $*$, not to be used with the ```*```  that we use in computer programming for multiplication

So if we are writing convolution in code, it will not be an $*$, because in code we already use the ```*``` for multiplication

<h3>Exercise: Write Pseudocode</h3>

One helpful Excercise to get a better idea of how convolution works is to write pseudocode (maybe even real code) to implement the algorithm we just described

This will help us uncover a few hidden details we may have not considered

Just by starting the code we realise one thing we need to take care of, which is what size should we initialise our input array as

If we recall in our example, the input image had length 4 but the kernel had length 2, the output had length 3

So whats the pattern?

We claim that, if we have an array of length $N$, and a filter of length $K$, then there are $N-K+1$ distinct possible positions we can put the filter into

<img src='extras/38.4.PNG' width='200'></img>

We might want to draw this on paper if we dont see right away why this is true 

---

So the first step in our psuedocode is to initialise the output array

But here is another detail we might not have considered

```
Given : input_image, kernel
output_height = input_height - kernel_height + 1
output_width = input_width - kernel_width + 1
```

But in our example, both our image and kernel were square, so we might be wondering, is this always the case, what is the convention ?

And the answer is that, for iamges, usually these are not squares

This just has to do with cameras and screens

Most screens, like our conputer screen and our TV screen, are not square, and therefore cameras do not take square pictures

Therefore, images we find in the wild, that we want to use as data, are typically also not square

Some neural networks do use square images for convenience, so when we build a dataset, we make them square and one example of this is MNIST, which we have already seen

On the other hand, kernels are almost always square by convention

---

Lets move on

The next step is just to fill in the output array, by performing the convolution algorithm

```python
given is input_image,kernel
output_height = input_height - kernel_height + 1
output_width = input_width - kernel_width + 1

output_image = np.zeros((output_height,output_width))

# the pair i,j always index our output
for i in range(0,output_height):
    for j in range(0,output_width):
        # then we loop through each position of the kernel
        for ii in range(0,kernel_height):
            for jj in range(0,kernel_width):
                output_image[i,j] += input_image(i+ii,j+jj) * kernel[ii,jj]
        
```

$$\large (A * W)_{ij} = \sum^{K-1}_{i^\prime=0}\sum^{K-1}_{j^\prime=0} A(i+i^\prime,j+j^\prime)w(i^\prime,j^\prime)$$

<h3>What are we really doing ?</h3>

In fact, what we are doing in deep learning is actually known as the <strong>cross-correlation</strong>

This seems like a more helpful and descriptive name compared to convolution

Convolution, in and of itself, probably does not mean much to us

But the word correlation does, we probably think of the word correlated as the same as this

If we say "X and Y are correlated", that means to us that "There is some degree of similarity between X and Y"

Therefore, we might think of CNNs as a correlation neural network rather than a convolutional neural network

The only difference between correlation and convolution, is that convolution reverse the orientation of the filter, whereas correlation does not

<h3>A new perspective on convolution</h3>

In this section we are going to continue our discussion of convolution by looking at a totally different perspective on how it works

This is very helpful for understanding convolution, but it does not teach us anything new mechanically, so its not necessary for knowing how to make CNNs

---

<h3>Vectorisation</h3>

Something we mention very often is vectorisation

We like to vectorise operations in code, because using numpy functions is a lot more effecient than writing our own python for loops

The pattern we are looking for when we want to vectorise an operation is usually of the form of a dot product

A dot product is an element-wise multiplication and then the summation of those results

$$\large a^Tb = \sum^N_{i=1}a_ib_i$$

This also applies to matrix multiplication, since if $A$ and $B$ are matricies, then the matrix multiplication of $A$ and $B$ is just $A \cdot B$

We notice that for convolution, we yet again have something similar

$$\large (A * W)_{ij} = \sum^{K-1}_{i^\prime=0}\sum^{K-1}_{j^\prime=0} A(i+i^\prime,j+j^\prime)w(i^\prime,j^\prime)$$

The only difference is that there are two summations

However, this is not really a relevant detail, since the outcome is still the same

Instead of summing over one axis, we are summing over two axes, but its still an element-wise sum and add

---

<h3>Why is the dot product important?</h3>

The question is, why is the dot product important?

One definition of the dot product, other than the element-wise sum and add, is 

$$\large a \cdot b = \sum^N_{i=1} = \vert a \vert \vert b \vert \cos \theta_{ab}$$


We sometimes call this ($\cos \theta_{ab}$) the cosine similarity or cosine distance depending on the sign we use

---

<h3>Angle = $0^\circ$</h3>

So how does this work geometrically

Consider just the angle for a moment

<img src='extras/38.8.PNG' width='300'></img>

If the angle between the two vectors is zero, then the cosine for that angle is one, $\cos(0) = 1$, that the maximum value for the cosine

---

<h3>Angle = $90^\circ$</h3>

Now imagine that the angle between the two vectors is $90^\circ$

<img src='extras/38.9.PNG' width='300'></img>

Then $\cos(90)=0$

---

<h3>Angle = $180^\circ$</h3>

Finally, imagine that the angle between two vectors is $180^\circ$, this is basically as far apart as possible

<img src='extras/38.10.PNG' width='300'></img>

Then $\cos(180)=-1$, which is the minimum value of the cosine

---

<h3>Cosine Similarity</h3>

So if we are just using raw cosine, then its a similarity

$$\large cos \theta_{ab} = \frac{a \cdot b}{\vert a \vert \vert b \vert} = \text{ how similar are a and b}$$

The larger the number, the closer the 2 vectors are 

The smaller the number, the further away they are

The maximum value, when $a$ and $b$ are parallel, is $1$

The minimum value, when $a$ and $b$ are anti-parallel is $-1$

And if $a$ and $b$ are orthogonal, then the cosine similarity is just $0$

---

Okay, so why is that important?

consider now how we would find the cosine of the angle between two vectors

$$\large cos(a,b) = \frac{a \cdot b}{\vert a \vert \vert b \vert} = \frac{\sum\limits^N_{i=1} a_ib_i}{\sqrt{\sum\limits^N_{i=1}a_i^2}\sqrt{\sum\limits^N_{i=1}b_i^2}}$$

---

<h3>Pearson Correlation</h3>

Now, lets compare this to another popular measurement, the Pearson Correlation

$$\large \varrho = \frac{\sum\limits^N_{i=1}(a_i - \bar a)(b_i-\bar b)}{\sqrt{\sum\limits^N_{i=1}(a_i - \bar a)^2}\sqrt{\sum\limits^N_{i=1}(b_i - \bar b)^2}}$$

Notice how similar this is to cosine similarity

The only difference is that the peason correlation uses mean subtraction

And so now we have two hints that convolution is really correlation

The first one from before was that, we were actually doing what is called the cross-correlation

And second, now we have that, dot prodcut, is actually very similar to the pearson correlation

---

<h3>What is a dot product</h3>

So hopefully by this point, we are convinced that the dot product, although seems like an abstact convept, can be thought of as a correlation measure

It tells us, how correlated is the first thing with the second thing

If they are highly positively correlated, then the dot product should be large and positive, that means two vectors pointing in nearly the same direction

If they are highly negatively correlated, then the dot product should be large and negative, that means two vectors pointing in nearly opposite directions

Finally, if the two vectors are orthogonal (at right angles), then the dot prodcut should be zero

---

<h3>Why do tihs?</h3>

The reason wht this is important is, we dont need to think of a filter as an abstact concept, in fact is just a pattern finder

This actually makes the term filter make a lot more sense, it filters out everything not related to the pattern contained in the filter by setting them to zero, and keeps everything that is related to the pattern

<img src='extras/38.11.PNG'></img>

So what convolution is doing is, its passing this filter along each point on the original input image and sliding it along

At each point it asks, is the pattern here?, is the pattern here?, is the pattern here?, and so forth

Then it gives us a high number in the positions where the pattern is found and then a small number where the pattern is not found

And thus, our first alternative perspective on convolution is that, its just a sliding pattenr finder that passes through an entire image, looking for a particular pattern
